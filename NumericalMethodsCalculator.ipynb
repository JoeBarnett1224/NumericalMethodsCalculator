```python
import math as m
import mpmath as mp
import numpy as np
import tkinter as tk
from tkinter import ttk

#Nonbuttoh functions

def parse_function(function):
    
    #Remove spaces
    
    function = function.replace(" ", "")
    
    #Multiplication
    
    i = 48
    number_list = []
    while i < 58:
        number_list.append(chr(i))
        i += 1
    i = 65
    letter_list = []
    while i < 91:
        letter_list.append(chr(i))
        i += 1
    i = 97
    while i < 123:
        letter_list.append(chr(i))
        i += 1
    i = 0
    index_list = []
    while i < (len(function) - 1):
        if (function[i] in number_list and
         function[i + 1] in letter_list):
            index_list.append(i)
        i += 1
    i = 0
    j = 0
    break_function = []
    while i < len(function):
        if i in index_list:
            break_function.append(function[j:i + 1])
            j = i + 1
        i += 1
    break_function.append(function[j:])
    i = 0
    while i < len(break_function):
        if i == 0:
            function = break_function[i]
        else:
            function = function + "*" + break_function[i]
        i += 1
    if "x(" in function:
        function = function.replace("x(", "x*(")
    i = 0
    while i < 10:
        search_string = str(i) + "("
        if search_string in function:
            function = function.replace(search_string,
             str(i) + "*" + "(")
        i += 1
    if ")(" in function:
        function = function.replace(")(", ")*(")
        
    #Exponentiation
    
    function = function.replace("^", "**")
    
    #Absolute value
    
    split_function = function.split("|")
    i = 0
    temp_function = ""
    if "|" in function:
        while i < len(split_function):
            if i % 2 == 0 and (i != len(split_function) - 1):
                temp_function = temp_function + split_function[i] + \
                 "abs("
            if i % 2 == 1:
                temp_function = temp_function + split_function[i] +")"
            if i == (len(split_function) - 1):
                temp_function = temp_function + split_function[i]
            i += 1
        function = temp_function
    if "xabs" in function:
        function = function.replace("xabs", "x*abs")
    i = 0
    while i < 10:
        search_string = str(i) + "abs"
        if search_string in function:
            function = function.replace(search_string,
             str(i) + "*" + "abs")
        i += 1
    if ")abs" in function:
        function = function.replace(")abs", ")*abs")
    
    #Exponential and logarithmic function
    
    if "e**" in function:
        function = function.replace("e**", "m.exp")
    
    if "log" in function:
        function = function.replace("log", "m.log")
    
    i = 0
    argument_list = []
    while i < len(function) - 2:
        if function[i:i+3] == "log":
            j = i + 3
            log_argument = ""
            while function[j] != ")":
                log_argument = log_argument + function[j]
                j += 1
            log_argument = log_argument + ")"
            argument_list.append(log_argument)
        i += 1
    i = 0
    while i < len(argument_list):
        if "," not in argument_list[i]:
            function = function.replace(argument_list[i],
             argument_list[i][0:-1] + ",10)")
        i += 1
    
    if "ln" in function:
        function = function.replace("ln", "m.log")
        
    if "sqrt" in function:
        function = function.replace("sqrt", "m.sqrt")
        
    if "cbrt" in function:
        function = function.replace("cbrt", "mp.cbrt")
        
    if "root" in function:
        function = function.replace("root", "mp.root")
        
    if "W" in function:
        function = function.replace("W", "mp.lambertw")
        
    #Trigonometric and hyperbolic functions
    
    if "sin" in function:
        function = function.replace("sin", "m.sin")
        
    if "arcm.sin" in function:
        function = function.replace("arcm.sin", "m.asin")
        
    if "arm.sinh" in function:
        function = function.replace("arm.sinh", "m.asinh")
    
    if "cos" in function:
        function = function.replace("cos", "m.cos")
        
    if "arcm.cos" in function:
        function = function.replace("arcm.cos", "m.acos")
        
    if "arm.cosh" in function:
        function = function.replace("arm.cosh", "m.acosh")
        
    if "tan" in function:
        function = function.replace("tan", "m.tan")
        
    if "arcm.tan" in function:
        function = function.replace("arcm.tan", "m.atan")
        
    if "arm.tanh" in function:
        function = function.replace("arm.tanh", "m.atanh")
        
    if "sec" in function:
        function = function.replace("sec", "mp.sec")
        
    if "arcmp.sec" in function:
        function = function.replace("arcmp.sec", "mp.asec")
        
    if "armp.sech" in function:
        function = function.replace("armp.sech", "mp.asech")
        
    if "csc" in function:
        function = function.replace("csc", "mp.csc")
        
    if "arcmp.csc" in function:
        function = function.replace("arcmp.csc", "mp.acsc")
        
    if "armp.csch" in function:
        function = function.replace("armp.csch", "mp.acsch")
        
    if "cot" in function:
        function = function.replace("cot", "mp.cot")
        
    if "arcmp.cot" in function:
        function = function.replace("arcmp.cot", "mp.acot")
        
    if "armp.coth" in function:
        function = function.replace("armp.coth", "mp.acoth")
        
    #Special functions
        
    i = 0
    psi_function_list = []
    while i < len(function) - 2:
        if function[i:i+3] == "psi":
            j = i + 3
            psi_argument = ""
            while function[j] != ")":
                psi_argument = psi_argument + function[j]
                j += 1
            psi_argument = psi_argument + ")"
            psi_function = "psi" + psi_argument
            psi_function_list.append(psi_function)
        i += 1
    i = 0
    while i < len(psi_function_list):
        if "," in psi_function_list[i]:
            function = function.replace(psi_function_list[i],
             "mp." + psi_function_list[i])
        else:
            function = function.replace(psi_function_list[i],
             "mp.digamma" + psi_function_list[i])
        i += 1
    function = function.replace("digammapsi", "digamma")
    
    i = 0
    gamma_function_list = []
    while i < len(function) - 4:
        if function[i:i+5] == "Gamma":
            j = i + 5
            gamma_argument = ""
            while function[j] != ")":
                gamma_argument = gamma_argument + function[j]
                j += 1
            gamma_argument = gamma_argument + ")"
            gamma_function = "Gamma" + gamma_argument
            gamma_function_list.append(gamma_function)
        i += 1
    i = 0
    while i < len(gamma_function_list):
        if "," not in gamma_function_list[i]:
            function = function.replace(gamma_function_list[i],
             "m.g" + gamma_function_list[i])
        else:
            function = function.replace(gamma_function_list[i],
             "mp.gammainc" + gamma_function_list[i])
        i += 1
    function = function.replace("gGamma", "gamma")
    function = function.replace("gammaincGamma", "gammainc")
    
    if "G" in function:
        function = function.replace("G", "mp.barnesg")
        
    if "Ei" in function:
        function = function.replace("Ei", "mp.ei")
        
    if "En" in function:
        function = function.replace("En", "mp.exp-int")
    
    if "li" in function:
        function = function.replace("li", "mp.li")
        
    if "Ci" in function:
        function = function.replace("Ci", "mp.ci")
        
    if "Chi" in function:
        function = function.replace("Chi", "mp.chi")
        
    if "Si" in function:
        function = function.replace("Si", "mp.si")
        
    if "Shi" in function:
        function = function.replace("Shi", "mp.shi")
    
    if "erf" in function:
        function = function.replace("erf","m.erf")
        
    if "m.erfi" in function:
        function = function.replace("m.erfi","mp.erfi")
        
    if "S" in function:
        function = function.replace("S", "mp.fresnels")
        
    if "C" in function:
        function = function.replace("C", "mp.fresnelc")
        
    if "J" in function:
        function = function.replace("J", "mp.besselj")
        
    if "Y" in function:
        function = function.replace("Y", "mp.bessely")
        
    if "I" in function:
        function = function.replace("I", "mp.besseli")
        
    if "K" in function:
        function = function.replace("K", "mp.besselk")
        
    if "H" in function:
        function = function.replace("H", "mp.struveh")
        
    if "L" in function:
        function = function.replace("L", "mp.struvel")
        
    if "E" in function:
        function = function.replace("E", "mp.webere")
        
    if "s(" in function:
        function = function.replace("s(", "mp.lommels1(")
    if "abmp.lommels1(" in function:
        function = function.replace("abmp.lommels1(", "abs(")
    if "comp.lommels1(" in function:
        function = function.replace("comp.lommels1(", "cos(")
    if "fresnelmp.lommels1(" in function:
        function = function.replace("fresnelmp.lommels1(", "fresnels(")
        
    if "Ai" in function:
        function = function.replace("Ai", "mp.airyai")
        
    if "Bi" in function:
        function = function.replace("Bi", "mp.airybi")
        
    if "barnesgi" in function:
        function = function.replace("barnesgi", "scorergi")
        
    if "struvehi" in function:
        function = function.replace("struvehi", "scorerhi")
        
    if "pFq" in function:
        function = function.replace("pFq", "mp.hyper")
        
    if "Amp.struvehB" in function:
        function = function.replace("Amp.struvehB", "mp.bihyper")
        
    if "eta" in function:
        function = function.replace("eta", "mp.altzeta")
        
    if "thmp.altzeta" in function:
        function = function.replace("thmp.altzeta", "mp.jtheta")
        
    if "zmp.altzeta" in function:
        function = function.replace("zmp.altzeta", "mp.zeta")
        
    if "Phi" in function:
        function = function.replace("Phi", "mp.lerchPhi")
        
    if "mp.struveli" in function:
        function = function.replace("mp.struveli", "mp.polylog")
        
    #Constants
    
    if "pi" in function:
        function = function.replace("pi", "m.pi")
    if "pix" in function:
        function = function.replace("pix", "pi*x")
    if "piabs" in function:
        function = function.replace("piabs", "pi*abs")
    if "pim." in function:
        function = function.replace("pim.", "pi*m.")
    if "pi(" in function:
        function = function.replace("pi(", "pi*(")
        
    if "phi" in function:
        function = function.replace("phi", "mp.phi")
    if "phix" in function:
        function = function.replace("phix", "phi*x")
    if "phiabs" in function:
        function = function.replace("phiabs", "phi*abs")
    if "phim." in function:
        function = function.replace("phim.", "phi*m.")
    if "phi(" in function:
        function = function.replace("phi(", "phi*(")
        
    i = 0
    while i < len(function) - 4:
        if (function[i - 1] != "." and function[i-2:i] != "di" and
         function[i:i + 5] == "gamma"):
            function = function[:i] + "mp.euler" + function[i + 5:]
        i += 1
    if "eulerx" in function:
        function = function.replace("eulerx", "euler*x")
    if "eulerabs" in function:
        function = function.replace("eulerabs", "euler*abs")
    if "eulerm." in function:
        function = function.replace("eulerm.", "euler*m.")
    if "euler(" in function:
        function = function.replace("euler(", "euler*(")
        
    if "A" in function:
        function = function.replace("A", "mp.glaisher")
    if "glaisherx" in function:
        function = function.replace("glaisherx", "glaisher*x")
    if "glaisherabs" in function:
        function = function.replace("glaisherabs", "glaisher*abs")
    if "glaisherm." in function:
        function = function.replace("glaisherm.", "glaisher*m.")
    if "glaisher(" in function:
        function = function.replace("glaisher(", "glaisher*(")
        
    if "infinity" in function:
        function = function.replace("infinity", "m.inf")
        
    #Multiplication 2
    
    if "xm." in function:
        function = function.replace("xm.", "x*m.")
    
    if ")m." in function:
        function = function.replace(")m.", ")*m.")
        
    if "xmp." in function:
        function = function.replace("xmp.", "x*mp.")
    
    if ")mp." in function:
        function = function.replace(")mp.", ")*mp.")
        
    if "pimp." in function:
        function = function.replace("pimp.", "pi*mp.")
        
    if "phimp." in function:
        function = function.replace("phimp.", "phi*mp.")
        
    if "eulermp." in function:
        function = function.replace("eulermp.", "euler*mp.")
        
    if "glaishermp." in function:
        function = function.replace("glaishermp.", "glaisher*mp.")
        
    #Corrections
    
    if "mp.exp-int" in function:
        function = function.replace("mp.exp-int", "mp.expint")
        
    if "mp.lerchPhi" in function:
        function = function.replace("mp.lerchPhi", "mp.lerchphi")
    
    #Return parsed functiom
    
    return function

def split_by(function, operations):
    output_list = []
    string_segment = ""
    i = 0
    while i < len(function):
        if function[i] in operations:
            output_list.append(string_segment)
            output_list.append(function[i])
            string_segment = ""
        else:
            string_segment = string_segment + function[i]
        i += 1
    output_list.append(string_segment)
    
    i = 0
    while i < len(output_list):
        if output_list[i] == "":
            output_list.pop(i)
        i += 1
    
    return output_list

def evaluate_expression(function):
    
    expression_value = split_by(function, "+-*/")
    i = 0
    while i < len(expression_value):
        if expression_value[i][-1] == "e":
            expression_value[i:i + 3] = \
             ["". join(expression_value[i:i + 3])]
        i += 1
    i = 0
    while i < len(expression_value):
        if (expression_value[i - 1] == "*" and
         expression_value[i] == "-"):
            expression_value[i:i + 2] = \
             ["". join(expression_value[i:i + 2])]
        i += 1
    
    #Evaluate mathemetical constants
    
    constant_dictionary = {"m.pi": m.pi,
                           "mp.phi": mp.phi,
                           'mp.euler': mp.euler,
                           "mp.glaisher": mp.glaisher,
                           "m.inf": m.inf}
    i = 0
    while i < len(expression_value):
        if expression_value[i] in constant_dictionary:
            expression_value[i] = \
             str(constant_dictionary[expression_value[i]])
        i += 1
    
    #Evaluate mathematical functions
    
    function_dictionary = {"abs": m.fabs,
                           "m.exp": m.exp,
                           "m.log": m.log,
                           "m.sqrt": m.sqrt,
                           "mp.cbrt": mp.cbrt,
                           "mp.root": mp.root,
                           "mp.lambertw": mp.lambertw,
                           "m.sin": m.sin,
                           "m.asin": m.asin,
                           "m.sinh": m.sinh,
                           "m.asinh": m.asinh,
                           "m.cos": m.cos,
                           "m.acos": m.acos,
                           "m.cosh": m.cosh,
                           "m.acosh": m.acosh,
                           "m.tan": m.tan,
                           "m.atan": m.atan,
                           "m.tanh": m.tanh,
                           "m.atanh": m.atanh,
                           "mp.sec": mp.sec,
                           "mp.asec": mp.asec,
                           "mp.sech": mp.sech,
                           "mp.asech": mp.asech,
                           "mp.csc": mp.csc,
                           "mp.acsc": mp.acsc,
                           "mp.csch": mp.csch,
                           "mp.acsch": mp.acsch,
                           "mp.cot": mp.cot,
                           "mp.acot": mp.acot,
                           "mp.coth": mp.coth,
                           "mp.acoth": mp.acoth,
                           "mp.digamma": mp.digamma,
                           "mp.psi": mp.psi,
                           "m.gamma": mp.gamma,
                           "mp.gammainc": mp.gammainc,
                           "mp.barnesg": mp.barnesg,
                           "mp.ei": mp.ei,
                           "mp.expint": mp.expint,
                           "mp.li": mp.li,
                           "mp.ci": mp.ci,
                           "mp.chi": mp.chi,
                           "mp.si": mp.si,
                           "mp.shi": mp.shi,
                           "m.erf": m.erf,
                           "m.erfc": m.erfc,
                           "mp.erfi": mp.erfi,
                           "mp.fresnels": mp.fresnels,
                           "mp.fresnelc": mp.fresnelc,
                           "mp.besselj": mp.besselj,
                           "mp.bessely": mp.bessely,
                           "mp.besseli": mp.besseli,
                           "mp.besselk": mp.besselk,
                           "mp.struveh": mp.struveh,
                           "mp.struvel": mp.struvel,
                           "mp.webere": mp.webere,
                           "mp.lommels1": mp.lommels1,
                           "mp.airyai": mp.airyai,
                           "mp.airybi": mp.airybi,
                           "mp.scorergi": mp.scorergi,
                           "mp.scorerhi": mp.scorerhi,
                           "mp.hyper": mp.hyper,
                           "mp.bihyper": mp.bihyper,
                           "mp.altzeta": mp.altzeta,
                           "mp.jtheta": mp.jtheta,
                           "mp.zeta": mp.zeta,
                           "mp.lerchphi": mp.lerchphi,
                           "mp.polylog": mp.polylog}
    i = 0
    while i < len(expression_value):
        if "negative" in expression_value[i]:
            expression_value[i] = \
             expression_value[i].replace("negative", "-")
        if ("(" in expression_value[i] and
         "[" not in expression_value[i]):
            if "," not in expression_value[i]:
                sub_list = split_by(expression_value[i], "()")
                value = \
                 function_dictionary[sub_list[0]](float(sub_list[2]))
                expression_value[i] = str(value)
            else:
                sub_list = split_by(expression_value[i], "(,)")
                if len(sub_list) == 6:
                    value = \
                     function_dictionary[sub_list[0]](float(sub_list[2]),
                     float(sub_list[4]))
                    expression_value[i] = str(value)
                if len(sub_list) == 8:
                    value = \
                     mp.re(function_dictionary[sub_list[0]](float(sub_list[2]),
                     float(sub_list[4]),
                     float(sub_list[6])))
                    expression_value[i] = str(value)
        if "(" in expression_value[i] and "[" in expression_value[i]:
            sub_list = split_by(expression_value[i], "()[],")
            j = 0
            index_values = []
            while j < len(sub_list):
                if sub_list[j] == "[" or sub_list[j] == "]":
                    index_values.append(j)
                j += 1
            j = 0
            list1 = []
            list2 = []
            while j < len(sub_list):
                if (j > index_values[0] and j < index_values[1] and
                 sub_list[j] != ","):
                    list1.append(float(sub_list[j]))
                if (j > index_values[2] and j < index_values[3] and
                 sub_list[j] != ","):
                    list2.append(float(sub_list[j]))
                j += 1
            value = function_dictionary[sub_list[0]](list1, list2,
             float(sub_list[-2]))
            expression_value[i] = str(value)
        i += 1
    
    #Exponents
    
    i = 0
    while i < len(expression_value) - 2:
        if expression_value[i:i + 2] == ["*", "*"]:
            expression_value[i:i + 2] = \
             ["". join(expression_value[i:i + 2])]
        i += 1
    i = 0
    while i < len(expression_value) - 1:
        if expression_value[i] == "**":
            value = float(expression_value[i - 1]) ** \
             float(expression_value[i + 1])
            expression_value[i - 1:i + 2] = \
             ["".join(expression_value[i - 1:i + 2])]
            expression_value[i - 1] = str(value)
        i += 1
        
    #Multiplication & Division
    
    i = 0
    while i < len(expression_value) - 1:
        if expression_value[i] == "*":
            value = float(expression_value[i - 1]) * \
             float(expression_value[i + 1])
            expression_value[i - 1:i + 2] = \
             ["".join(expression_value[i - 1:i + 2])]
            expression_value[i - 1] = str(value)
            i = i
        elif expression_value[i] == "/":
            value = float(expression_value[i - 1]) / \
             float(expression_value[i + 1])
            expression_value[i - 1:i + 2] = \
             ["".join(expression_value[i - 1:i + 2])]
            expression_value[i - 1] = str(value)
            i = i
        else:
            i += 1
            
    #Addition & Subtraction
    
    i = 0
    while i < len(expression_value) - 1:
        if expression_value[i] == "+":
            value = float(expression_value[i - 1]) + \
             float(expression_value[i + 1])
            expression_value[i - 1:i + 2] = \
             ["".join(expression_value[i - 1:i + 2])]
            expression_value[i - 1] = str(value)
            i = i
        elif (expression_value[i] == "-" and i != 0 and
         ("e" not in expression_value[i])):
            value = float(expression_value[i - 1]) - \
             float(expression_value[i + 1])
            expression_value[i - 1:i + 2] = \
             ["".join(expression_value[i - 1:i + 2])]
            expression_value[i - 1] = str(value)
            i = i
        elif expression_value[i] == "-" and i == 0:
            expression_value[i:i + 2] = \
             ["".join(expression_value[i:i + 2])]
            i += 1
        else:
            i += 1
    
    return expression_value[0]

def evaluate_function(function, number):
    
    #Preliminary steps
    
    function = "(" + function + ")"
    function = function.replace("x", number)
    exception = "e" + str(number) + "p"
    if exception in function:
        function = function.replace(exception, "exp")
    
    #Split at ( and )
    
    evaluation_list = split_by(function, "(,)")
    
    #Evaluate parentheses
    
    i = 0
    j = len(evaluation_list)
    while j > 1:
        i = 0
        while i < j:
            if "start" in evaluation_list[i]:
                evaluation_list[i] = \
                evaluation_list[i].replace("start", "(")
                evaluation_list[i] = \
                evaluation_list[i].replace("end", ")")
            if "comma" in evaluation_list[i]:
                evaluation_list[i] = \
                evaluation_list[i].replace("comma", ",")
            i += 1
        i = 0
        while i < j - 2:
            if ((evaluation_list[i] == "(" and
             evaluation_list[i + 2] == ")") or
             (evaluation_list[i] == "(" and
             evaluation_list[i + 2] == ",") or
             (evaluation_list[i] == "comma" and
             evaluation_list[i + 2] == ",") or
             (evaluation_list[i] == "comma"
             and evaluation_list[i + 2] == ")")):
                evaluation_list[i + 1] = \
                 evaluate_expression(evaluation_list[i + 1])
                if (evaluation_list[i - 1][-1] == "+" or
                 evaluation_list[i - 1][-1] == "-" or
                 evaluation_list[i - 1][-1] == "*" or
                 evaluation_list[i - 1][-1] == "/" or
                 evaluation_list[i - 1][-1] == "("):
                    evaluation_list.pop(i)
                    evaluation_list.pop(i + 1)
                    evaluation_list.append("")
                    evaluation_list.append("")
                elif (evaluation_list[i] == "(" and
                 evaluation_list[i + 2] == ","):
                    evaluation_list[i] = "start"
                    evaluation_list[i + 2] = "comma"
                elif (evaluation_list[i] == "comma" and
                 evaluation_list[i + 2] == ","):
                    evaluation_list[i + 2] = "comma"
                elif (evaluation_list[i] == "comma" and
                 evaluation_list[i + 2] == ")"):
                    evaluation_list[i + 2] = "end"
                else:
                    evaluation_list[i] = "start"
                    evaluation_list[i + 2] = "end"
            i += 1
        i = 0
        function = ""
        while i < j:
            function = function + evaluation_list[i]
            i += 1
        if "start-" in function:
            function = function.replace("start-", "startnegative")
        if "comma-" in function:
            function = function.replace("comma-", "commanegative")
        evaluation_list = split_by(function, "()")
        j = len(evaluation_list)
    
    if "negative" in function:
        function = function.replace("negative", "-")
    function = function.replace("start", "")
    function = function.replace("end", "")
    
    #Return value
    
    return function

def string_to_list(string):
    
    #Remove spaces
    
    if " " in string:
        string = string.replace(" ", "")
        
    #Split String
    
    string_list = string.split(",")
    
    #Convert list items to floats
    
    i = 0
    while i < len(string_list):
        string_list[i] = float(string_list[i])
        i += 1
        
    #Return list
    
    return string_list

def multivariate_parse(function):
    
    function = parse_function(function)
    
    #Multiplication
    
    if "xy" in function:
        function = function.replace("xy", "x*y")
    if "y(" in function:
        function = function.replace("y(", "y*(")
    if "yabs" in function:
        function = function.replace("yabs", "y*abs")
    if "piy" in function:
        function = function.replace("piy", "pi*y")
    if "phiy" in function:
        function = function.replace("phiy", "phi*y")
    if "eulery" in function:
        function = function.replace("eulery", "euler*y")
    if "glaishery" in function:
        function = function.replace("glaishery", "glaisher*y")
    if "ym." in function:
        function = function.replace("ym.", "y*m.")
    if "ymp." in function:
        function = function.replace("ymp.", "y*mp.")
        
    #Corrections
    
    if "bessely*(" in function:
        function = function.replace("bessely*(", "bessely(")
        
    #Return parsed function
    
    return function

def evaluate_multivariate(function, number1, number2):
    
    #Preliminary steps
    
    function = function.replace("y", number2)
    exception = "bessel" + str(number2)
    if exception in function:
        function = function.replace(exception, "bessely")
        
    #Evaluate Function
    
    function = evaluate_function(function, number1)
    
    #Return value
    
    return function

#Button functions

def bisection_method_submit():
    
    #Get and parse function
    
    bisection_method_function = \
     str(bisection_method_function_entry_variable.get())
    bisection_method_function = \
     parse_function(bisection_method_function)
    
    #Get nterval
    
    a = str(bisection_method_interval_entry_variable1.get())
    b = str(bisection_method_interval_entry_variable2.get())
    
    #Get tolerance
    
    tolerance = bisection_method_tolerance_variable.get()
    
    #Check if conditions of intermediate value theorem are satisfied
    #and run bisection method.
    
    value_at_a = float(evaluate_function(bisection_method_function, a))
    value_at_b = float(evaluate_function(bisection_method_function, b))
    if value_at_a * value_at_b > 0:
        output = "Bisection method won't work as F(a) and f(b) have "
        output = output + "the same sign."
    else:
        error = abs(float(b) - float(a))
        while error > 2 * tolerance:
            c = (float(a) + float(b))/2
            value_at_c = \
            float(evaluate_function(bisection_method_function, str(c)))
            if value_at_a * value_at_c < 0:
                b = c
                error = abs(float(b) - float(a))
            if value_at_b * value_at_c < 0:
                a = c
                error = abs(float(b) - float(a))
        value = (float(a) + float(b)) / 2
        error = error / 2
        
        #Create and set output.
        
        output = "x = "
        output = output + str(value)
        output = output + ", Error: "
        output = output + str(error)
        output = output + "."
    bisection_method_output_string.set(output)

def fixed_point_iteration_submit():
    
    #Get and parse function
    
    fixed_point_iteration_function = \
     str(fixed_point_iteration_function_entry_variable.get())
    fixed_point_iteration_function = \
     parse_function(fixed_point_iteration_function)
    
    #Get initial value
    
    x0 = str(fixed_point_iteration_initial_value_entry_variable.get())
    
    #Get tolerance
    
    tolerance = fixed_point_iteration_tolerance_variable.get()
    
    #Run fixed point iteration
    
    x = evaluate_function(fixed_point_iteration_function, x0)
    error = abs(float(x0) - float(x))
    while error > tolerance:
        x0 = x
        x = evaluate_function(fixed_point_iteration_function, x0)
        error = abs(float(x0) - float(x))
        
    #Create and set output.
        
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    fixed_point_iteration_output_string.set(output)
    
def newtons_method_submit():
    
    #Get and parse function & derivative
    
    newtons_method_function1 = \
     str(newtons_method_function_entry_variable1.get())
    newtons_method_function1 = \
     parse_function(newtons_method_function1)
    newtons_method_function2 = \
     str(newtons_method_function_entry_variable2.get())
    newtons_method_function2 = \
     parse_function(newtons_method_function2)
    
    #Get initial value
    
    x0 = str(newtons_method_initial_value_entry_variable.get())
    
    #Get tolerance
    
    tolerance = newtons_method_tolerance_variable.get()
    
    #Run Newton's Method
    
    f = evaluate_function(newtons_method_function1, x0)
    df = evaluate_function(newtons_method_function2, x0)
    x = float(x0) - (float(f)/float(df))
    error = abs(float(x0) - float(x))
    while error > tolerance:
        x0 = x
        f = evaluate_function(newtons_method_function1, str(x0))
        df = evaluate_function(newtons_method_function2, str(x0))
        x = x0 - (float(f)/float(df))
        error = abs(float(x0) - float(x))
        
    #Create and set output.
        
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    newtons_method_output_string.set(output)
    
def secant_method_submit():
    
    #Get and parse function
    
    secant_method_function = \
     str(secant_method_function_entry_variable.get())
    secant_method_function = parse_function(secant_method_function)
    
    #Get initial values
    
    x0 = str(secant_method_initial_value_entry_variable1.get())
    x1 = str(secant_method_initial_value_entry_variable2.get())
    
    #Get tolerance
    
    tolerance = secant_method_tolerance_variable.get()
    
    #Run Secant Method
    
    f0 = evaluate_function(secant_method_function, x0)
    f1 = evaluate_function(secant_method_function, x1)
    x = float(x0) - (float(f0) / ((float(f0) - float(f1)) / \
     (float(x0) - float(x1))))
    error = abs(float(x0) - x)
    while error > tolerance:
        x0 = x1
        x1 = x
        f0 = evaluate_function(secant_method_function, str(x0))
        f1 = evaluate_function(secant_method_function, str(x1))
        x = float(x0) - (float(f0) / ((float(f0) - float(f1)) / \
         (float(x0) - float(x1))))
        error = abs(float(x0) - x)
        
    #Create and set output.
        
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    secant_method_output_string.set(output)
    
def method_of_false_position_submit():
    
    #Get and parse function
    
    method_of_false_position_function = \
     str(method_of_false_position_function_entry_variable.get())
    method_of_false_position_function = \
     parse_function(method_of_false_position_function)
    
    #Get nterval
    
    a = str(method_of_false_position_interval_entry_variable1.get())
    b = str(method_of_false_position_interval_entry_variable2.get())
    
    #Get tolerance
    
    tolerance = method_of_false_position_tolerance_variable.get()
    
    #Check if conditions of intermediate value theorem are satisfied 
    #and run method of false position.
    
    value_at_a = \
     float(evaluate_function(method_of_false_position_function, a))
    value_at_b = \
     float(evaluate_function(method_of_false_position_function, b))
    if value_at_a * value_at_b > 0:
        output = "Method of false position won't work as F(a) and "
        output = output + "f(b) have the same sign."
    else:
        error = abs(float(b) - float(a))
        while error > tolerance:
            c = (float(a) * value_at_b - float(b) * \
             value_at_a)/(value_at_b - value_at_a)
            value_at_c = \
             float(evaluate_function(method_of_false_position_function,
             str(c)))
            if value_at_a * value_at_c < 0:
                b = c
                error = abs(float(b) - float(a))
            if value_at_b * value_at_c < 0:
                a = c
                error = abs(float(b) - float(a))
                
        output = "x = "
        output = output + str(c)
        output = output + ", Error: "
        output = output + str(error)
        output = output + "."
    method_of_false_position_output_string.set(output)
    
def modified_newtons_method_submit():
    
    #Get and parse function & derivatives
    
    modified_newtons_method_function1 = \
     str(modified_newtons_method_function_entry_variable1.get())
    modified_newtons_method_function1 = \
     parse_function(modified_newtons_method_function1)
    modified_newtons_method_function2 = \
     str(modified_newtons_method_function_entry_variable2.get())
    modified_newtons_method_function2 = \
     parse_function(modified_newtons_method_function2)
    modified_newtons_method_function3 = \
     str(modified_newtons_method_function_entry_variable3.get())
    modified_newtons_method_function3 = \
     parse_function(modified_newtons_method_function3)
    
    #Get initial value
    
    x0 = \
    str(modified_newtons_method_initial_value_entry_variable.get())
    
    #Get tolerance
    
    tolerance = modified_newtons_method_tolerance_variable.get()
    
    #Run Modified Newton's Method
    
    f = evaluate_function(modified_newtons_method_function1, x0)
    df = evaluate_function(modified_newtons_method_function2, x0)
    d2f = evaluate_function(modified_newtons_method_function3, x0)
    x = float(x0) - (float(f) * float(df))/((float(df)) ** 2 - \
     float(f) * float(d2f))
    error = abs(float(x0) - float(x))
    while error > tolerance:
        x0 = x
        f = evaluate_function(modified_newtons_method_function1,
         str(x0))
        df = evaluate_function(modified_newtons_method_function2,
         str(x0))
        d2f = evaluate_function(modified_newtons_method_function3,
         str(x0))
        x = float(x0) - (float(f) * float(df)) / ((float(df)) ** 2 - \
         float(f) * float(d2f))
        error = abs(float(x0) - float(x))
        
    #Create and set output.
        
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    modified_newtons_method_output_string.set(output)
    
def steffensens_method_submit():
    
    #Get and parse function
    
    steffensens_method_function = \
     str(steffensens_method_function_entry_variable.get())
    steffensens_method_function = \
     parse_function(steffensens_method_function)
    
    #Get initial value
    
    x0 = str(steffensens_method_initial_value_entry_variable.get())
    
    #Get tolerance
    
    tolerance = steffensens_method_tolerance_variable.get()
    
    #Run Steffensen's Method
    
    x1 = float(x0) + \
     float(evaluate_function(steffensens_method_function, x0))
    x = float(x0) - \
     ((float(evaluate_function(steffensens_method_function, x0))) ** \
     2) / (float(evaluate_function(steffensens_method_function,
     str(x1))) - \
     float(evaluate_function(steffensens_method_function, x0)))
    error = abs(float(x0) - float(x))
    while error > tolerance:
        x0 = x
        x1 = float(x0) + \
         float(evaluate_function(steffensens_method_function, str(x0)))
        x = x0 - ((float(evaluate_function(steffensens_method_function,
         str(x0)))) ** 2) / \
         (float(evaluate_function(steffensens_method_function,
         str(x1))) - \
         float(evaluate_function(steffensens_method_function,
         str(x0))))
        error = abs(float(x0) - float(x))
    
    #Create and set output.
    
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    steffensens_method_output_string.set(output)
    
def mullers_method_submit():
    
    #Get and parse function
    
    mullers_method_function = \
     str(mullers_method_function_entry_variable.get())
    mullers_method_function = \
     parse_function(mullers_method_function)
    
    #Get initial values
    
    x0 = str(mullers_method_initial_value_entry_variable1.get())
    x1 = str(mullers_method_initial_value_entry_variable2.get())
    x2 = str(mullers_method_initial_value_entry_variable3.get())
    
    #Get tolerance
    
    tolerance = mullers_method_tolerance_variable.get()
    
    #Run Muller's Method
    
    f0 = float(evaluate_function(mullers_method_function, x0))
    f1 = float(evaluate_function(mullers_method_function, x1))
    f2 = float(evaluate_function(mullers_method_function, x2))
    x0 = float(x0)
    x1 = float(x1)
    x2 = float(x2)
    a = ((x1 - x2) * (f0 - f2) - (x0 - x2) * (f1 - f2)) / \
     ((x0 - x2) * (x1 - x2) * (x0 - x1))
    b = ((x0 - x2) ** 2 * (f1 - f2) - (x1 - x2) ** 2 * (f0 - f2)) / \
     ((x0 - x2) * (x1 - x2) * (x0 - x1))
    c = f2
    x = x2 - (2 * c)/(b + (b/abs(b)) * m.sqrt(b**2 - 4 * a * c))
    error = abs(x2 - x)
    while error > tolerance:
        x0 = x1
        x1 = x2
        x2 = x
        f0 = float(evaluate_function(mullers_method_function, str(x0)))
        f1 = float(evaluate_function(mullers_method_function, str(x1)))
        f2 = float(evaluate_function(mullers_method_function, str(x2)))
        a = ((x1 - x2) * (f0 - f2) - (x0 - x2) * (f1 - f2)) / \
         ((x0 - x2) * (x1 - x2) * (x0 - x1))
        b = ((x0 - x2) ** 2 * (f1 - f2) - (x1 - x2) ** 2 * \
         (f0 - f2)) / ((x0 - x2) * (x1 - x2) * (x0 - x1))
        c = f2
        x = x2 - (2 * c)/(b + (b/abs(b)) * m.sqrt(b**2 - 4 * a * c))
        error = abs(x2 - x)
        
    #Create and set output.
    
    output = "x = "
    output = output + str(x)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    mullers_method_output_string.set(output)
    
def lagrange_interpolating_polynomials_submit():
    
    #Get x values and create list
    
    x_list = \
     str(lagrange_interpolating_polynomials_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values
    
    y_list = \
     str(lagrange_interpolating_polynomials_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get interpolation point
    
    x = \
     float(lagrange_interpolating_polynomials_interpolation_point_entry_variable.get())
    
    #Run Lagrange Interpolation
    
    y = 0
    i = 0
    while i < len(x_list):
        p = 1
        j = 0
        while j < len(x_list):
            if j != i:
                p *= (x - x_list[j])/(x_list[i] - x_list[j])
            j += 1
        y += y_list[i] * p
        i += 1
        
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    lagrange_interpolating_polynomials_output_string.set(output)
    
def nevilles_method_submit():
    
    #Get x values and create list
    
    x_list = str(nevilles_method_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values
    
    y_list = str(nevilles_method_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get interpolation point
    
    x = float(nevilles_method_interpolation_point_entry_variable.get())
    
    #Run Neville's Method
    
    i = 1
    y = y_list
    while i < len(x_list):
        j = 0
        while j < len(x_list) - i:
            y[j] = ((x - x_list[i + j]) * y[j] + (x_list[j] - x) * \
             y[j + 1]) / (x_list[j] - x_list[i + j])
            j += 1
        i += 1
    y = y[0]
    
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    nevilles_method_output_string.set(output)
    
def newtons_divided_difference_method_submit():
    
    #Get x values and create list
    
    x_list = \
     str(newtons_divided_difference_method_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values
    
    y_list = \
     str(newtons_divided_difference_method_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get interpolation point
    
    x = \
     float(newtons_divided_difference_method_interpolation_point_entry_variable.get())
    
    #Run Newton's Divided Difference Method
    
    matrix = []
    i = 0
    while i < len(x_list):
        matrix.append([])
        j = 0
        while j < len(x_list):
            matrix[i].append(0)
            j += 1
        i += 1
    i = 0
    while i < len(y_list):
        matrix[i][0] = y_list[i]
        i += 1
    i = 1
    while i < len(x_list):
        j = 1
        while j <= i:
            matrix[i][j] = (matrix[i][j - 1] - matrix[i - \
             1][j - 1]) / (x_list[i] - x_list[i - j])
            j += 1
        i += 1
    i=0
    sum = 0
    while i < len(x_list):
        j = 0
        product = 1
        while i - j > 0:
            product *= x - x_list[i - j - 1]
            j += 1
        sum += matrix[i][i] * product
        i += 1
        
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(sum)
    output = output + "."
    newtons_divided_difference_method_output_string.set(output)
    
def hermite_interpolation_submit():
    
    #Get x values and create list
    
    x_list = str(hermite_interpolation_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values and create list
    
    y_list = str(hermite_interpolation_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get y' values and create list
    
    y_prime_list = \
     str(hermite_interpolation_y_prime_values_entry_variable.get())
    y_prime_list = string_to_list(y_prime_list)
    
    #Get interpolation point
    
    x = \
     float(hermite_interpolation_interpolation_point_entry_variable.get())
    
    #Run Hermite Interpolation
    
    matrix = []
    i = 0
    while i < 2 * len(x_list):
        matrix.append([])
        j = 0
        while j < 2 * len(x_list):
            matrix[i].append(0)
            j += 1
        i += 1
    i = 0
    while i < 2 * len(y_list):
        matrix[i][0] = y_list[i // 2]
        i += 1
    i = 0
    x_list_temp = []
    while i < 2 * len(x_list):
        x_list_temp.append(x_list[i // 2])
        i += 1
    i = 1
    while i < 2 * len(x_list):
        j = 1
        while j <= i:
            if j == 1 and i % 2 == 1:
                matrix[i][j] = y_prime_list[i // 2]
            else:
                matrix[i][j] = (matrix[i][j - 1] - \
                matrix[i - 1][j - 1]) / (x_list_temp[i] - \
                x_list_temp[i - j])
            j += 1
        i += 1
    i = 0
    sum = 0
    while i < 2 * len(x_list):
        j = 0
        product = 1
        while i - j > 0:
            product *= x - x_list_temp[i - j - 1]
            j += 1
        sum += matrix[i][i] * product
        i += 1
        
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(sum)
    output = output + "."
    hermite_interpolation_output_string.set(output)
    
def natural_cubic_spline_submit():
    
    #Get x values and create list
    
    x_list = str(natural_cubic_spline_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values
    
    y_list = str(natural_cubic_spline_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get interpolation point
    
    x = \
     float(natural_cubic_spline_interpolation_point_entry_variable.get())
    
    #Evaluate natural cubic spline
    
    n = 4 * (len(x_list) - 1)    
    A = np.zeros((n, n))
    i = 0
    while i < n / 2:
        j = 0
        while j < n:
            if i % 2 == 0 and j > 2 * i - 1 and j < 2 * i + 4:
                A[i][j] = x_list[int(i / 2)] ** (j % 4)
            if i % 2 == 1 and j > 2 * i - 3 and j < 2 * i + 2:
                A[i][j] = x_list[int(i // 2 + 1)] ** (j % 4)
            j += 1
        i += 1
    i = int(n / 2)
    while i < (3 * n - 4) / 4:
        j = 0
        while j < n:
            if j > (4 * i) - (2 * n) - 1 and j < (4 * i) - (2 * n) + 8:
                if j % 4 == 0:
                    A[i][j] = 0
                elif i % 2 == 0 and j % 8 == 1:
                    A[i][j] = x_list[i - int(n / 2) + 1]
                elif i % 2 == 0 and j % 8 == 2:
                    A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                elif i % 2 == 0 and j % 8 == 3:
                    A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                elif i % 2 == 0 and j % 8 == 5:
                    A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 0 and j % 8 == 6:
                    A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 0 and j % 8 == 7:
                    A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
                elif i % 2 == 1 and j % 8 == 5:
                    A[i][j] = x_list[i - int(n / 2) + 1]
                elif i % 2 == 1 and j % 8 == 6:
                    A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                elif i % 2 == 1 and j % 8 == 7:
                    A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                elif i % 2 == 1 and j % 8 == 1:
                    A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 1 and j % 8 == 2:
                    A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                else:
                    A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
            j += 1
        i += 1
    i = int((3 * n - 4) / 4)
    while i < n - 2:
        j = 0
        while j < n:
            if (j > (4 * i) - (3 * n) + 3 and
             j < (4 * i) - (3 * n) + 12):
                if j % 4 == 0 or j % 4 == 1:
                    A[i][j] = 0
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 2:
                    A[i][j] = -2
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 3:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                    4) + 1])
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 6:
                    A[i][j] = 2
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 7:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 2:
                    A[i][j] = 2
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 3:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 6:
                    A[i][j] = -2
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 7:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                    4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 2:
                    A[i][j] = 2
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 3:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 6:
                    A[i][j] = -2
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 7:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 2:
                    A[i][j] = -2
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 3:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 6:
                    A[i][j] = 2
                else:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
            j += 1
        i += 1
    i = n - 2
    while i < n:
        j = 0
        while j < n:
            if i % 2 == 0 and j < 4:
                if j % 4 == 0 or j % 4 == 1:
                    A[i][j] = 0
                elif j % 4 == 2:
                    A[i][j] = 2
                else:
                    A[i][j] = 6 * x_list[0]
            if i % 2 == 1 and j > n - 5:
                if j % 4 == 0 or j % 4 == 1:
                    A[i][j] = 0
                elif j % 4 == 2:
                    A[i][j] = 2
                else:
                    A[i][j] = 6 * x_list[-1]
            j += 1
        i += 1
    B = np.zeros(n)
    i = 0
    while i < n/2:
        if i % 2 == 0:
            B[i] = y_list[int(i / 2)]
        else:
            B[i] = y_list[int(i // 2) + 1]
        i += 1
    C = np.linalg.solve(A, B.transpose())
    x_list.sort(key = float)
    n = 0
    while x_list[n] < x:
        n += 1
    n = n - 1
    y = C[4 * n] + C[4 * n + 1] * x + C[4 * n + 2] * x ** 2 + C[4 * \
     n + 3] * x ** 3
    
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    natural_cubic_spline_output_string.set(output)

def clamped_cubic_spline_submit():
    
    #Get x values and create list
    
    x_list = str(clamped_cubic_spline_x_values_entry_variable.get())
    x_list = string_to_list(x_list)
    
    #Get y values
    
    y_list = str(clamped_cubic_spline_y_values_entry_variable.get())
    y_list = string_to_list(y_list)
    
    #Get initial conditions
    
    ic1 = \
     float(clamped_cubic_spline_initial_condition_entry_variable1.get())
    ic2 = \
     float(clamped_cubic_spline_initial_condition_entry_variable2.get())
    
    #Get interpolation point
    
    x = \
     float(clamped_cubic_spline_interpolation_point_entry_variable.get())
    
    #Evaluate clamped cubic spline
    
    n = 4 * (len(x_list) - 1)    
    A = np.zeros((n, n))
    i = 0
    while i < n / 2:
        j = 0
        while j < n:
            if i % 2 == 0 and j > 2 * i - 1 and j < 2 * i + 4:
                A[i][j] = x_list[int(i / 2)] ** (j % 4)
            if i % 2 == 1 and j > 2 * i - 3 and j < 2 * i + 2:
                A[i][j] = x_list[int(i // 2 + 1)] ** (j % 4)
            j += 1
        i += 1
    i = int(n / 2)
    while i < (3 * n - 4) / 4:
        j = 0
        while j < n:
            if j > (4 * i) - (2 * n) - 1 and j < (4 * i) - (2 * n) + 8:
                if j % 4 == 0:
                    A[i][j] = 0
                elif i % 2 == 0 and j % 8 == 1:
                    A[i][j] = x_list[i - int(n / 2) + 1]
                elif i % 2 == 0 and j % 8 == 2:
                    A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                elif i % 2 == 0 and j % 8 == 3:
                    A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                elif i % 2 == 0 and j % 8 == 5:
                    A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 0 and j % 8 == 6:
                    A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 0 and j % 8 == 7:
                    A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** \
                     2
                elif i % 2 == 1 and j % 8 == 5:
                    A[i][j] = x_list[i - int(n / 2) + 1]
                elif i % 2 == 1 and j % 8 == 6:
                    A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                elif i % 2 == 1 and j % 8 == 7:
                    A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                elif i % 2 == 1 and j % 8 == 1:
                    A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                elif i % 2 == 1 and j % 8 == 2:
                    A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                else:
                    A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
            j += 1
        i += 1
    i = int((3 * n - 4) / 4)
    while i < n - 2:
        j = 0
        while j < n:
            if (j > (4 * i) - (3 * n) + 3 and
             j < (4 * i) - (3 * n) + 12):
                if j % 4 == 0 or j % 4 == 1:
                    A[i][j] = 0
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 2:
                    A[i][j] = -2
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 3:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 6:
                    A[i][j] = 2
                elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 7:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 2:
                    A[i][j] = 2
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 3:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 6:
                    A[i][j] = -2
                elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 7:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 2:
                    A[i][j] = 2
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 3:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 6:
                    A[i][j] = -2
                elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 7:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 2:
                    A[i][j] = -2
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 3:
                    A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / \
                     4) + 1])
                elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 6:
                    A[i][j] = 2
                else:
                    A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
            j += 1
        i += 1
    i = n - 2
    while i < n:
        j = 0
        while j < n:
            if i % 2 == 0 and j < 4:
                if j % 4 == 0:
                    A[i][j] = 0
                elif j % 4 == 1:
                    A[i][j] = 1
                elif j % 4 == 2:
                    A[i][j] = 2 * x_list[0]
                else:
                    A[i][j] = 3 * x_list[0] ** 2
            if i % 2 == 1 and j > n - 5:
                if j % 4 == 0:
                    A[i][j] = 0
                elif j % 4 == 1:
                    A[i][j] = 1
                elif j % 4 == 2:
                    A[i][j] = 2 * x_list[-1]
                else:
                    A[i][j] = 3 * x_list[-1] ** 2
            j += 1
        i += 1
    B = np.zeros(n)
    i = 0
    while i < n/2:
        if i % 2 == 0:
            B[i] = y_list[int(i / 2)]
        else:
            B[i] = y_list[int(i // 2) + 1]
        i += 1
    B[-2] = ic1
    B[-1] = ic2
    C = np.linalg.solve(A, B.transpose())
    x_list.sort(key = float)
    n = 0
    while x_list[n] < x:
        n += 1
    n = n - 1
    y = C[4 * n] + C[4 * n + 1] * x + C[4 * n + 2] * x ** 2 + C[4 * \
     n + 3] * x ** 3
    
    #Create and set output.
    
    output = "f("
    output = output + str(x)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    clamped_cubic_spline_output_string.set(output)
    
def forward_difference_submit():
    
    #Get and parse function
    
    forward_difference_function = \
     str(forward_difference_function_entry_variable.get())
    forward_difference_function = \
     parse_function(forward_difference_function)
    
    #Get x value
    
    x = float(forward_difference_x_value_entry_variable.get())
    
    #Get h value
    
    h = float(forward_difference_h_value_entry_variable.get())
    
    #Evaluate forward difference
    
    df = (float(evaluate_function(forward_difference_function,
     str(x + h))) - \
     float(evaluate_function(forward_difference_function, str(x))))/h
    d2f = (float(evaluate_function(forward_difference_function,
     str(x + 2 * h))) - 2 * \
     (float(evaluate_function(forward_difference_function,
     str(x + h)))) + \
     float(evaluate_function(forward_difference_function, str(x)))) / \
     (h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    forward_difference_output_string.set(output)

def backward_difference_submit():
    
    #Get and parse function
    
    backward_difference_function = \
     str(backward_difference_function_entry_variable.get())
    backward_difference_function = \
     parse_function(backward_difference_function)
    
    #Get x value
    
    x = float(backward_difference_x_value_entry_variable.get())
    
    #Get h value
    
    h = float(backward_difference_h_value_entry_variable.get())
    
    #Evaluate backward difference
    
    df = (float(evaluate_function(backward_difference_function,
     str(x))) - float(evaluate_function(backward_difference_function,
     str(x - h))))/h
    d2f = (float(evaluate_function(backward_difference_function,
     str(x))) - 2 * \
     (float(evaluate_function(backward_difference_function,
     str(x - h)))) + \
     float(evaluate_function(backward_difference_function,
     str(x - 2 * h))))/(h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    backward_difference_output_string.set(output)

def central_difference_submit():
    
    #Get and parse function
    
    central_difference_function = \
     str(central_difference_function_entry_variable.get())
    central_difference_function = \
     parse_function(central_difference_function)
    
    #Get x value
    
    x = central_difference_x_value_entry_variable.get()
    
    #Get h value
    
    h = central_difference_h_value_entry_variable.get()
    
    #Evaluate central difference
    
    df = (float(evaluate_function(central_difference_function,
     str(x + h))) - \
     float(evaluate_function(central_difference_function,
     str(x - h)))) / (2 * h)
    d2f = (float(evaluate_function(central_difference_function,
     str(x + h))) - 2 * \
     (float(evaluate_function(central_difference_function,
     str(x)))) + float(evaluate_function(central_difference_function,
     str(x - h))))/(h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    central_difference_output_string.set(output)
    
def three_point_midpoint_formula_submit():
    
    #Get and parse function
    
    three_point_midpoint_formula_function = \
     str(three_point_midpoint_formula_function_entry_variable.get())
    three_point_midpoint_formula_function = \
     parse_function(three_point_midpoint_formula_function)
    
    #Get x value
    
    x = \
     float(three_point_midpoint_formula_x_value_entry_variable.get())
    
    #Get h value
    
    h = \
     float(three_point_midpoint_formula_h_value_entry_variable.get())
    
    #Evaluate three-point midpoint formula
    
    df = (float(evaluate_function(three_point_midpoint_formula_function,
     str(x + h))) - \
     float(evaluate_function(three_point_midpoint_formula_function,
     str(x - h)))) / (2 * h)
    d2f = (float(evaluate_function(three_point_midpoint_formula_function,
     str(x + h))) - 2 * \
     (float(evaluate_function(three_point_midpoint_formula_function,
     str(x)))) + \
     float(evaluate_function(three_point_midpoint_formula_function,
     str(x - h))))/(h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    three_point_midpoint_formula_output_string.set(output)

def three_point_endpoint_formula_submit():
    
    #Get and parse function
    
    three_point_endpoint_formula_function = \
     str(three_point_endpoint_formula_function_entry_variable.get())
    three_point_endpoint_formula_function = \
     parse_function(three_point_endpoint_formula_function)
    
    #Get x value
    
    x = \
     float(three_point_endpoint_formula_x_value_entry_variable.get())
    
    #Get h value
    
    h = \
     float(three_point_endpoint_formula_h_value_entry_variable.get())
    
    #Evaluate three-point endpoint formula
    
    df = (-3 * \
     (float(evaluate_function(three_point_endpoint_formula_function,
     str(x)))) + 4 * \
     (float(evaluate_function(three_point_endpoint_formula_function,
     str(x + h)))) - \
     float(evaluate_function(three_point_endpoint_formula_function,
     str(x + 2 * h))))/(2 * h)
    d2f = \
     (float(evaluate_function(three_point_endpoint_formula_function,
     str(x))) - 2 * \
     (float(evaluate_function(three_point_endpoint_formula_function,
     str(x + h)))) + \
     float(evaluate_function(three_point_endpoint_formula_function,
     str(x + 2 * h))))/(h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    three_point_endpoint_formula_output_string.set(output)

def five_point_midpoint_formula_submit():
    
    #Get and parse function
    
    five_point_midpoint_formula_function = \
     str(five_point_midpoint_formula_function_entry_variable.get())
    five_point_midpoint_formula_function = \
     parse_function(five_point_midpoint_formula_function)
    
    #Get x value
    
    x = float(five_point_midpoint_formula_x_value_entry_variable.get())
    
    #Get h value
    
    h = float(five_point_midpoint_formula_h_value_entry_variable.get())
    
    #Evaluate five-point midpoint formula
    
    df = (float(evaluate_function(five_point_midpoint_formula_function,
     str(x - 2 * h))) - 8 * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x - h)))) + 8 * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x + h)))) - \
     float(evaluate_function(five_point_midpoint_formula_function,
     str(x + 2 * h))))/(12 * h)
    d2f = ((-1) * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x - 2 * h)))) + 16 * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x - h)))) - 30 * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x)))) + 16 * \
     (float(evaluate_function(five_point_midpoint_formula_function,
     str(x + h)))) - \
     float(evaluate_function(five_point_midpoint_formula_function,
     str(x + 2 * h))))/(12 * h ** 2)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + ", f''("
    output = output + str(x)
    output = output + ") = "
    output = output + str(d2f)
    output = output + "."
    five_point_midpoint_formula_output_string.set(output)

def five_point_endpoint_formula_submit():
    
    #Get and parse function
    
    five_point_endpoint_formula_function = \
     str(five_point_endpoint_formula_function_entry_variable.get())
    five_point_endpoint_formula_function = \
     parse_function(five_point_endpoint_formula_function)
    
    #Get x value
    
    x = float(five_point_endpoint_formula_x_value_entry_variable.get())
    
    #Get h value
    
    h = float(five_point_endpoint_formula_h_value_entry_variable.get())
    
    #Evaluate five-point endpoint formula
    
    df = ((-25) * \
     (float(evaluate_function(five_point_endpoint_formula_function,
     str(x)))) + 48 * \
     (float(evaluate_function(five_point_endpoint_formula_function,
     str(x + h)))) - 36 * \
     (float(evaluate_function(five_point_endpoint_formula_function,
     str(x + 2 * h)))) + 16 * \
     (float(evaluate_function(five_point_endpoint_formula_function,
     str(x + 3 * h)))) - 3 * \
     float(evaluate_function(five_point_endpoint_formula_function,
     str(x + 4 * h))))/(12 * h)
    
    #Create and set output.
    
    output = "f'("
    output = output + str(x)
    output = output + ") = "
    output = output + str(df)
    output = output + "."
    five_point_endpoint_formula_output_string.set(output)
    
def midpoint_rule_submit():
    
    #Get and parse function
    
    midpoint_rule_function = \
     str(midpoint_rule_function_entry_variable.get())
    midpoint_rule_function = parse_function(midpoint_rule_function)
    
    #Get nterval
    
    a = float(midpoint_rule_interval_entry_variable1.get())
    b = float(midpoint_rule_interval_entry_variable2.get())
    
    #Get number of subintervals
    
    n = midpoint_rule_n_entry_variable.get()
    
    #Run midpoint rule
    
    h = (b - a) / n
    x = a + (h / 2)
    s = 0
    while x < b:
        s += float(evaluate_function(midpoint_rule_function, str(x)))
        x += h
    s = h * s
    
    #Create and set output.
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(a)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(b)
    output = output.translate(superscripts)
    output = output + str(midpoint_rule_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(s)
    output = output + "."
    midpoint_rule_output_string.set(output)
    
def trapezoid_rule_submit():
    
    #Get and parse function
    
    trapezoid_rule_function = \
     str(trapezoid_rule_function_entry_variable.get())
    trapezoid_rule_function = \
     parse_function(trapezoid_rule_function)
    
    #Get nterval
    
    a = float(trapezoid_rule_interval_entry_variable1.get())
    b = float(trapezoid_rule_interval_entry_variable2.get())
    
    #Get number of subintervals
    
    n = trapezoid_rule_n_entry_variable.get()
    
    #Run trapezoid rule
    
    h = (b - a) / n
    x = a + h
    s = 0
    while x < b:
        s += 2 * float(evaluate_function(trapezoid_rule_function,
        str(x)))
        x += h
    s = s + float(evaluate_function(trapezoid_rule_function, str(a)))
    s = s + float(evaluate_function(trapezoid_rule_function, str(b)))
    s = (h / 2) * s
    
    #Create and set output.
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(a)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(b)
    output = output.translate(superscripts)
    output = output + str(trapezoid_rule_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(s)
    output = output + "."
    trapezoid_rule_output_string.set(output)

def simpsons_rule_submit():
    
    #Get and parse function
    
    simpsons_rule_function = \
     str(simpsons_rule_function_entry_variable.get())
    simpsons_rule_function = parse_function(simpsons_rule_function)
    
    #Get nterval
    
    a = float(simpsons_rule_interval_entry_variable1.get())
    b = float(simpsons_rule_interval_entry_variable2.get())
    
    #Get number of subintervals
    
    n = simpsons_rule_n_entry_variable.get()
    
    #Run Simpson's rule
    
    h = (b - a) / n
    x = a + h
    i = 0
    s = 0
    while x < b:
        i += 1
        if i % 2 == 0:
            s += 2 * float(evaluate_function(simpsons_rule_function,
             str(x)))
        else:
            s += 4 * float(evaluate_function(simpsons_rule_function,
             str(x)))
        x += h
    s = s + float(evaluate_function(simpsons_rule_function, str(a)))
    s = s + float(evaluate_function(simpsons_rule_function, str(b)))
    s = (h / 3) * s
    
    #Create and set output.
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(a)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(b)
    output = output.translate(superscripts)
    output = output + str(simpsons_rule_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(s)
    output = output + "."
    simpsons_rule_output_string.set(output)
    
def romberg_integration_submit():
    
    #Get and parse function
    
    romberg_integration_function = \
     str(romberg_integration_function_entry_variable.get())
    romberg_integration_function = \
     parse_function(romberg_integration_function)
    
    #Get nterval
    
    a = float(romberg_integration_interval_entry_variable1.get())
    b = float(romberg_integration_interval_entry_variable2.get())
    
    #Get tolerance
    
    tolerance = romberg_integration_tolerance_variable.get()
    
    #Run Romberg Integration
    
    matrix = [[0, 0], [0, 0]]
    matrix[0][0] = ((b - a) / 2) * \
     (float(evaluate_function(romberg_integration_function,
     str(a))) + float(evaluate_function(romberg_integration_function,
     str(b))))
    matrix[1][0] = ((b - a) / 4) * \
     (float(evaluate_function(romberg_integration_function,
     str(a))) + 2 * \
     float(evaluate_function(romberg_integration_function,
     str((a + b) / 2))) + \
     float(evaluate_function(romberg_integration_function, str(b))))
    matrix[1][1] = matrix[1][0] + (1 / 3) * (matrix[1][0] - \
     matrix[0][0])
    error = abs(matrix[1][1] - matrix[0][0])
    while error > tolerance:
        i = 0
        while i < len(matrix):
            matrix[i].append(0)
            i += 1
        matrix.append([])
        i = len(matrix) - 1
        j = 0
        while j < len(matrix):
            if j == 0:
                n = 2 ** i
                h = (b - a) / n
                x = a + h
                s = 0
                while x < b:
                    s += 2 * \
                     float(evaluate_function(romberg_integration_function,
                     str(x)))
                    x += h
                s = s + \
                float(evaluate_function(romberg_integration_function,
                 str(a)))
                s = s + \
                 float(evaluate_function(romberg_integration_function,
                 str(b)))
                s = (h / 2) * s
            else:
                s = matrix[i][j - 1] + (1 / ((4 ** j) - 1)) * \
                 (matrix[i][j - 1] - matrix[i - 1][j - 1])
            matrix[i].append(s)
            j += 1
        error = abs(matrix[i][i] - matrix[i - 1][i - 1])
    I = matrix[-1][-1]
    
    #Create and set output
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(a)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(b)
    output = output.translate(superscripts)
    output = output + \
    str(romberg_integration_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(I)
    output = output + ", Error: "
    output = output + str(error)
    output = output + "."
    romberg_integration_output_string.set(output)
    
def adaptive_quadrature_submit():
    
    #Get and parse function
    
    adaptive_quadrature_function = \
     str(adaptive_quadrature_function_entry_variable.get())
    adaptive_quadrature_function = \
     parse_function(adaptive_quadrature_function)
    
    #Get nterval
    
    a = float(adaptive_quadrature_interval_entry_variable1.get())
    b = float(adaptive_quadrature_interval_entry_variable2.get())
    
    #Get tolerance
    
    tolerance = adaptive_quadrature_tolerance_variable.get()
    
    #Run adaptive quadrature
    
    I = 0
    e = 0
    point_list = [a, b]
    while len(point_list) > 0:
        I1 = ((b - a) / 6) * \
         (float(evaluate_function(adaptive_quadrature_function,
         str(point_list[0]))) + 4 * \
         (float(evaluate_function(adaptive_quadrature_function,
         str((point_list[0] + point_list[1]) / 2)))) + \
         float(evaluate_function(adaptive_quadrature_function,
         str(point_list[1]))))
        I2 = ((b - a) / 12) * \
         (float(evaluate_function(adaptive_quadrature_function,
         str(point_list[0]))) + 4 * \
         (float(evaluate_function(adaptive_quadrature_function,
         str((point_list[0] + point_list[1]) / 4)))) + \
         float(evaluate_function(adaptive_quadrature_function,
         str((point_list[0] + point_list[1]) / 2))))
        I3 = ((b - a) / 12) * \
         (float(evaluate_function(adaptive_quadrature_function,
         str((point_list[0] + point_list[1]) / 2))) + 4 * \
         (float(evaluate_function(adaptive_quadrature_function,
         str((3 * (point_list[0] + point_list[1])) / 4)))) + \
         float(evaluate_function(adaptive_quadrature_function,
         str(point_list[1]))))
        error = (abs(I1 - I2 - I3)) / 15
        if error < tolerance:
            I += I2
            I += I3
            e += error
            point_list.pop(0)
            point_list.pop(0)
        else:
            tolerance = tolerance / 2
            m = (point_list[0] + point_list[1]) / 2
            first = point_list.pop(0)
            point_list.insert(0, m)
            point_list.insert(0, m)
            point_list.insert(0, first)
        
    #Create and set output
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(a)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(b)
    output = output.translate(superscripts)
    output = output + \
     str(adaptive_quadrature_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(I)
    output = output + ", Error: "
    output = output + str(e)
    output = output + "."
    adaptive_quadrature_output_string.set(output)
    
def gaussian_quadrature_submit():
    
    #Get and parse function
    
    gaussian_quadrature_function = \
     str(gaussian_quadrature_function_entry_variable.get())
    gaussian_quadrature_function = \
     parse_function(gaussian_quadrature_function)

    #Get number of points and weights
    
    n = gaussian_quadrature_n_entry_variable.get()
    
    #Run Gaussian Quadrature
    
    nodes = mp.taylor(lambda x: mp.legendre(n, x), 0, n)
    nodes.reverse()
    nodes = mp.polyroots(nodes)
    coefficients = mp.taylor(lambda x: mp.legendre(n - 1, x), 0, n - 1)
    coefficients.reverse()
    weights = []
    i = 0
    while i < len(nodes):
        weight = (2 * (1 - (nodes[i]) ** 2)) / \
         ((n * (mp.polyval(coefficients, nodes[i]))) ** 2)
        weights.append(weight)
        i += 1
    i = 0
    I = 0
    while i < len(nodes):
        I += weights[i] * \
         float(evaluate_function(gaussian_quadrature_function,
         str(nodes[i])))
        i += 1
        
    #Create and set output
    
    output = ""
    subscripts = str.maketrans("0123456789", "")
    output = output + str(-1)
    output = output.translate(subscripts)
    superscripts = str.maketrans("0123456789", "")
    output = output + str(1)
    output = output.translate(superscripts)
    output = output + \
     str(gaussian_quadrature_function_entry_variable.get())
    output = output + "dx = "
    output = output + str(I)
    output = output + "."
    gaussian_quadrature_output_string.set(output)
    
def eulers_method_submit():
    
    #Get and parse function
    
    eulers_method_function = \
     str(eulers_method_function_entry_variable.get())
    eulers_method_function = multivariate_parse(eulers_method_function)
    
    #Get nterval
    
    a = float(eulers_method_interval_entry_variable1.get())
    b = float(eulers_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = float(eulers_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = eulers_method_n_entry_variable.get()
    
    #Run Euler's Method
    
    h = (b - a) / n
    x = a
    y = ic
    while x < b:
        df = float(evaluate_multivariate(eulers_method_function,
         str(x), str(y)))
        y = y + h * df
        x = x + h
    
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    eulers_method_output_string.set(output)
    
def midpoint_method_submit():
    
    #Get and parse function
    
    midpoint_method_function = \
     str(midpoint_method_function_entry_variable.get())
    midpoint_method_function = \
     multivariate_parse(midpoint_method_function)
    
    #Get nterval
    
    a = float(midpoint_method_interval_entry_variable1.get())
    b = float(midpoint_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = float(midpoint_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = midpoint_method_n_entry_variable.get()
    
    #Run Midpoint Method
    
    h = (b - a) / n
    x = a
    y = ic
    while x < b - h:
        df = float(evaluate_multivariate(midpoint_method_function,
         str(x), str(y)))
        y = y + h * \
         float(evaluate_multivariate(midpoint_method_function,
         str(x + (h / 2)), str(y + ((h / 2) * df))))
        x = x + h
    
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    midpoint_method_output_string.set(output)

def modified_euler_method_submit():
    
    #Get and parse function
    
    modified_euler_method_function = \
     str(modified_euler_method_function_entry_variable.get())
    modified_euler_method_function = \
     multivariate_parse(modified_euler_method_function)
    
    #Get nterval
    
    a = float(modified_euler_method_interval_entry_variable1.get())
    b = float(modified_euler_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = \
     float(modified_euler_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = modified_euler_method_n_entry_variable.get()
    
    #Run Modified Euler Method
    
    h = (b - a) / n
    x = a
    y = ic
    while x < b - h:
        df = \
         float(evaluate_multivariate(modified_euler_method_function,
         str(x), str(y)))
        y = y + (h / 2) * \
         float(df + \
         float(evaluate_multivariate(modified_euler_method_function,
         str(x + h), str(y + (h * df)))))
        x = x + h
        
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    modified_euler_method_output_string.set(output)

def heuns_method_submit():
    
    #Get and parse function
    
    heuns_method_function = \
     str(heuns_method_function_entry_variable.get())
    heuns_method_function = multivariate_parse(heuns_method_function)
    
    #Get nterval
    
    a = float(heuns_method_interval_entry_variable1.get())
    b = float(heuns_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = float(heuns_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = heuns_method_n_entry_variable.get()
    
    #Run Heun's Method
    
    h = (b - a) / n
    x = a
    y = ic
    while x < b - h:
        df = \
         float(evaluate_multivariate(heuns_method_function,
         str(x), str(y)))
        y = y + (h / 4) * \
         float(df + 3 * \
         float(evaluate_multivariate(heuns_method_function,
         str(x + ((2 * h) / 3)), str(y + ((2 * h) / 3) * \
         float(evaluate_multivariate(heuns_method_function,
         str(x + (h / 3)), str(y + (h / 3) * df)))))))
        x = x + h
        
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    heuns_method_output_string.set(output)

def runge_kutta_method_submit():
    
    #Get and parse function
    
    runge_kutta_method_function = \
     str(runge_kutta_method_function_entry_variable.get())
    runge_kutta_method_function = \
     multivariate_parse(runge_kutta_method_function)
    
    #Get nterval
    
    a = float(runge_kutta_method_interval_entry_variable1.get())
    b = float(runge_kutta_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = \
     float(runge_kutta_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = runge_kutta_method_n_entry_variable.get()
    
    #Run Runge Kutta Method
    
    h = (b - a) / n
    x = a
    y = ic
    while x < b - h:
        k1 = h * \
         float(evaluate_multivariate(runge_kutta_method_function,
         str(x), str(y)))
        k2 = h * \
         float(evaluate_multivariate(runge_kutta_method_function,
         str(x + (h / 2)), str(y + (k1 / 2))))
        k3 = h * \
         float(evaluate_multivariate(runge_kutta_method_function,
         str(x + (h / 2)), str(y + (k2 / 2))))
        k4 = h * \
         float(evaluate_multivariate(runge_kutta_method_function,
         str(x + h), str(y + k3)))
        y = y + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
        x = x + h
        
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    runge_kutta_method_output_string.set(output)
    
def runge_kutta_fehlberg_method_submit():
    
    #Get and parse function
    
    runge_kutta_fehlberg_method_function = \
     str(runge_kutta_fehlberg_method_function_entry_variable.get())
    runge_kutta_fehlberg_method_function = \
     multivariate_parse(runge_kutta_fehlberg_method_function)
    
    #Get nterval
    
    a = \
     float(runge_kutta_fehlberg_method_interval_entry_variable1.get())
    b = \
     float(runge_kutta_fehlberg_method_interval_entry_variable2.get())
    
    #Get step sizes
    
    h1 = \
     float(runge_kutta_fehlberg_method_h_value_entry_variable1.get())
    h2 = \
     float(runge_kutta_fehlberg_method_h_value_entry_variable2.get())
    
    #Get initial condition
    
    ic = \
     float(runge_kutta_fehlberg_method_initial_condition_entry_variable.get())
    
    #Get tolerance
    
    tolerance = \
     float(runge_kutta_fehlberg_method_tolerance_variable.get())
    
    #Run #Runge-Kutta-Fehlberg Method
    
    x = a
    y = ic
    h = h2
    flag = 1
    output = ""
    while flag == 1:
        k1 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x), str(y)))
        k2 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x + (h / 4)), str(y + (k1 / 4))))
        k3 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x + ((3 * h) / 8)), str(y + ((3 * k1) / 32) + \
         ((9 * k2) / 32))))
        k4 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x + ((12 * h) / 13)), str(y + ((1932 * k1) / 2197) - \
         ((7200 * k2) / 2197) + ((7296 * k3) / 2197))))
        k5 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x + h), str(y + ((439 * k1) / 216) - (8 * k2) + \
         ((3680 * k3) / 513) - ((845 * k4) / 4104))))
        k6 = h * \
         float(evaluate_multivariate(runge_kutta_fehlberg_method_function,
         str(x + (h / 2)), str(y - ((8 * k1) / 27) + (2 * k2) - \
         ((3544 * k3) / 2565) + ((1859 * k4) / 4104) - \
         ((11 * k5) / 40))))
        r = (1 / h) * (abs((k1 / 360) - ((128 * k3) / 4275) - \
         ((2197 * k4) / 75240) + (k5 / 50) + ((2 * k6) / 55)))
        if r <= tolerance:
            x = x + h
            y = y + ((25 * k1) / 216) + ((1408 * k3) / 2565) + \
             ((2197 * k4) / 4104) - (k5 / 5)
        delta = 0.84 * ((tolerance / r) ** 0.25)
        if delta <= 0.1:
            h = 0.1 * h
        elif delta >= 4:
            h = 4 * h
        else:
            h = delta * h
        if h > h2:
            h = h2
        if x >= b:
            flag = 0
        else:
            if x + h > b:
                h = b - x
            else:
                if h < h1:
                    output = "Minimum step size is too high."
                    flag = 0
    
    #Create and set output
    
    if output == "":
        output = "y("
        output = output + str(b)
        output = output + ") = "
        output = output + str(y)
        output = output + "."
    runge_kutta_fehlberg_method_output_string.set(output)
    
def adams_bashforth_technique_submit():
    
    #Get and parse function
    
    adams_bashforth_technique_function = \
     str(adams_bashforth_technique_function_entry_variable.get())
    adams_bashforth_technique_function = \
     multivariate_parse(adams_bashforth_technique_function)
    
    #Get nterval
    
    a = float(adams_bashforth_technique_interval_entry_variable1.get())
    b = float(adams_bashforth_technique_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = \
     float(adams_bashforth_technique_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = adams_bashforth_technique_n_entry_variable.get()
    
    #Run Adams-Bashforth Technique
    
    h = (b - a) / n
    x = a
    y = ic
    i = 0
    y_list = []
    while i < n + 1:
        if i == 0:
            y_list.append(y)
        elif i < 4:
            k1 = h * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x), str(y_list[-1])))
            k2 = h * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x + (h / 2)), str(y_list[-1] + (k1 / 2))))
            k3 = h * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x + (h / 2)), str(y_list[-1] + (k2 / 2))))
            k4 = h * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x + h), str(y_list[-1] + k3)))
            y = y_list[-1] + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
            y_list.append(y)
        else:
            y = y_list[3] + (h / 24) * ((55 * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x), str(y_list[3])))) - (59 * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x - h), str(y_list[2])))) + (37 * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x - (2 * h)), str(y_list[1])))) - (9 * \
             float(evaluate_multivariate(adams_bashforth_technique_function,
             str(x - (3 * h)), str(y_list[0])))))
            y_list.pop(0)
            y_list.append(y)
        if i > 0:
            x = x + h
        i += 1
    y = y_list[-1]
    
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    adams_bashforth_technique_output_string.set(output)
    
def milues_method_submit():
    
    #Get and parse function
    
    milues_method_function = \
     str(milues_method_function_entry_variable.get())
    milues_method_function = multivariate_parse(milues_method_function)
    
    #Get nterval
    
    a = float(milues_method_interval_entry_variable1.get())
    b = float(milues_method_interval_entry_variable2.get())
    
    #Get initial condition
    
    ic = float(milues_method_initial_condition_entry_variable.get())
    
    #Get number of steps
    
    n = milues_method_n_entry_variable.get()
    
    #Run Milues Method
    
    h = (b - a) / n
    x = a
    y = ic
    i = 0
    y_list = []
    while i < n + 1:
        if i == 0:
            y_list.append(y)
        elif i < 4:
            k1 = h * \
             float(evaluate_multivariate(milues_method_function,
             str(x), str(y_list[-1])))
            k2 = h * \
             float(evaluate_multivariate(milues_method_function,
             str(x + (h / 2)), str(y_list[-1] + (k1 / 2))))
            k3 = h * \
             float(evaluate_multivariate(milues_method_function,
             str(x + (h / 2)), str(y_list[-1] + (k2 / 2))))
            k4 = h * \
             float(evaluate_multivariate(milues_method_function,
             str(x + h), str(y_list[-1] + k3)))
            y = y_list[-1] + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
            y_list.append(y)
        else:
            y = y_list[0] + ((4 * h) / 3) * ((2 * \
             float(evaluate_multivariate(milues_method_function,
             str(x), str(y_list[3])))) - \
             (float(evaluate_multivariate(milues_method_function,
             str(x - h), str(y_list[2])))) + (2 * \
             float(evaluate_multivariate(milues_method_function,
             str(x - (2 * h)), str(y_list[1])))))
            y_list.pop(0)
            y_list.append(y)
        if i > 0:
            x = x + h
        i += 1
    y = y_list[-1]
    
    #Create and set output
    
    output = "y("
    output = output + str(b)
    output = output + ") = "
    output = output + str(y)
    output = output + "."
    milues_method_output_string.set(output)
    
def extrapolation_submit():
    
    #Get and parse function
    
    extrapolation_function = \
     str(extrapolation_function_entry_variable.get())
    extrapolation_function = multivariate_parse(extrapolation_function)
    
    #Get nterval
    
    a = float(extrapolation_interval_entry_variable1.get())
    b = float(extrapolation_interval_entry_variable2.get())
    
    #Get step sizes
    
    h1 = float(extrapolation_h_value_entry_variable1.get())
    h2 = float(extrapolation_h_value_entry_variable2.get())
    
    #Get initial condition
    
    ic = float(extrapolation_initial_condition_entry_variable.get())
    
    #Get tolerance
    
    tolerance = float(extrapolation_tolerance_variable.get())
    
    #Run Extrapolation
    
    denominators = [2, 4, 6, 8, 12, 16, 24, 32]
    matrix = [[0, 0], [0, 0]]
    h = h2
    h = h / denominators[0]
    x = a
    y = ic
    output = ""
    while x < b:
        y_list = []
        i = 0
        while i < denominators[0] + 1:
            if i == 0:
                y_list.append(y)
            elif i == 1:
                y = y_list[0] + (h * \
                 float(evaluate_multivariate(extrapolation_function,
                 str(x), str(y_list[0]))))
                y_list.append(y)
            else:
                y = y_list[0] + (2 * h * \
                 float(evaluate_multivariate(extrapolation_function,
                 str(x + h), str(y_list[-1]))))
                y_list.pop(0)
                y_list.append(y)
            i += 1
        y = (1 / 2) * (y_list[0] + y_list[-1] + (h * \
         float(evaluate_multivariate(extrapolation_function,
         str(x + (2 * h)), str(y_list[-1])))))
        matrix[0][0] = y
        h = h2
        h = h / denominators[1]
        x = a
        y = ic
        y_list = []
        i = 0
        while i < denominators[1] + 1:
            if i == 0:
                y_list.append(y)
            elif i == 1:
                y = y_list[0] + (h * \
                 float(evaluate_multivariate(extrapolation_function,
                 str(x), str(y_list[0]))))
                y_list.append(y)
            else:
                y = y_list[0] + (2 * h * \
                 float(evaluate_multivariate(extrapolation_function,
                 str(x + (i - 1) * h), str(y_list[-1]))))
                y_list.pop(0)
                y_list.append(y)
            i += 1
        y = (1 / 2) * (y_list[0] + y_list[-1] + (h * \
         float(evaluate_multivariate(extrapolation_function,
         str(x + (4 * h)), str(y_list[-1])))))
        matrix[1][0] = y
        coefficient = ((1/4) ** 2) / (((1 / 2) ** 2) - ((1 / 4) ** 2))
        y = matrix[1][0] + coefficient * (matrix[1][0] - matrix[0][0])
        matrix[1][1] = y
        error = abs(matrix[1][1] - matrix[0][0])
        while error > tolerance:
            i = 0
            while i < len(matrix):
                matrix[i].append(0)
                i += 1
            matrix.append([])
            i = len(matrix) - 1
            if i == 8:
                break
            j = 0
            while j < len(matrix):
                if j == 0:
                    h = h2
                    h = h / denominators[i]
                    x = a
                    y = ic
                    y_list = []
                    k = 0
                    while k < denominators[i] + 1:
                        if k == 0:
                            y_list.append(y)
                        elif k == 1:
                            y = y_list[0] + (h * \
                             float(evaluate_multivariate(extrapolation_function,
                             str(x), str(y_list[0]))))
                            y_list.append(y)
                        else:
                            y = y_list[0] + (2 * h * \
                             float(evaluate_multivariate(extrapolation_function,
                             str(x + (k - 1) * h), str(y_list[-1]))))
                            y_list.pop(0)
                            y_list.append(y)
                        k += 1
                    y = (1 / 2) * (y_list[0] + y_list[-1] + (h * \
                     float(evaluate_multivariate(extrapolation_function,
                     str(x + (denominators[i] * h)),
                     str(y_list[-1])))))
                else:
                    coefficient = ((1 / denominators[i]) ** 2) / \
                     (((1 / denominators[i - j]) ** 2) - \
                     ((1 / denominators[i]) ** 2))
                    y = matrix[i][j - 1] + coefficient * \
                     (matrix[i][j - 1] - matrix[i - 1][j - 1])
                matrix[i].append(y)
                j += 1
            error = abs(matrix[i][i] - matrix[i - 1][i - 1])
        if i == 8:
            x = x
            y = y
            h = h / 2
            matrix = [[0, 0], [0, 0]]
            if h < h1:
                output = "Minimum step size is too high."
                break
        else:
            x = x + h
            y = matrix[-1][-1]
            if x + h2 > b:
                h = b - x
            else:
                h = h2
            matrix = [[0, 0], [0, 0]]
    
    #Create and set output
    
    if output == "":
        output = "y("
        output = output + str(b)
        output = output + ") = "
        output = output + str(y)
        output = output + "."
    extrapolation_output_string.set(output)
    
#Help & Acknowledgements windows

def help1():
    
    #Set up window
    
    global window2
    window2 = tk.Toplevel()
    window2.title("Help")
    window2.geometry("1250x500")
    
    #Title
    
    help_title = tk.Label(window2, text = "Help",
     font = "Calibri 24 bold")
    help_title.pack()
    
    #Scrollbar
    
    main_frame = tk.Frame(window2)
    main_frame.pack(fill = "both", expand = True)
    help_canvas = tk.Canvas(main_frame)
    help_canvas.pack(side = "left", fill = "both", expand = True)
    help_scrollbar = ttk.Scrollbar(main_frame, orient = "vertical",
     command = help_canvas.yview)
    help_scrollbar.pack(side = "right", fill = "y")
    help_canvas.configure(yscrollcommand = help_scrollbar.set)
    help_canvas.bind("<Configure>",
     lambda e: help_canvas.configure(scrollregion = \
     help_canvas.bbox("all")))
    help_frame = tk.Frame(help_canvas)
    help_canvas.create_window((0, 0), window = help_frame,
     anchor = "nw")
    
    #Introduction
    
    label1 = '''Below are instructions on how to use this app,
including installation, supported functions, and implemented numerical
methods.'''
    help_label = tk.Label(help_frame, text = label1, font = "Calibri 14")
    help_label.pack()
    
    #Required Packages
    
    required_packages_label = tk.Label(help_frame,
     text = "Required Packages", font = "Calibri 20 bold")
    required_packages_label.pack()
    required_packages_label1 = tk.Label(help_frame, text = "math",
     font = "Calibri 14")
    required_packages_label1.pack()
    required_packages_label2 = tk.Label(help_frame, text = "mpmath",
     font = "Calibri 14")
    required_packages_label2.pack()
    required_packages_label3 = tk.Label(help_frame, text = "numpy",
     font = "Calibri 14")
    required_packages_label3.pack()
    required_packages_label4 = tk.Label(help_frame, text = "tkinter",
     font = "Calibri 14")
    required_packages_label4.pack()
    required_packages_label5 = tk.Label(help_frame, text = "ttk",
     font = "Calibri 14")
    required_packages_label5.pack()
    label2 = '''All of these packages can be installed via the terminal
in Jupyter Notebook using the pip install command.'''
    required_packages_label6 = tk.Label(help_frame, text = label2,
     font = "Calibri 14")
    required_packages_label6.pack()
    
    #Supported constants and functions
    
    constants_and_functions_label = tk.Label(help_frame,
     text = "Supported Constants and Functions", font = "Calibri 20 bold")
    constants_and_functions_label.pack()
    label3 = '''Many of the numerical methods implemented in this app
require the user to input one or more functions, which are interpreted
as Python strings.

The supported mathematical operations are addition (+), subtraction
(-), multiplication (*), division (/), and exponentiation (^). The app
follows the order of operations taught in grade school.

This app implements various mathematical constants and functions. See
the Constants and Functions sections for more information on which
constants and functions are supported and how they are implemented.
When using a mathematical function be sure to enclose the argument(s)
in parenthesis. For example, use sin(x), not sinx.'''
    constants_and_functions_label2 = tk.Label(help_frame,
     text = label3, font = "Calibri 14")
    constants_and_functions_label2.pack()
    
    #Constants
    
    constants_label = tk.Label(help_frame, text = "Constants",
     font = "Calibri 18 bold")
    constants_label.pack()
    constants_table = ttk.Treeview(help_frame, columns = ("column1",
     "column2", "column3"), show = "headings", height = 7)
    constants_table.heading("column1", text = "Mathematical Constant")
    constants_table.heading("column2", text = "Implementation")
    constants_table.heading("column3", text = "Python Function")
    constants_table.insert(parent = "", index = 0,
     values = ("Pi", "pi", "math.pi"))
    constants_table.insert(parent = "", index = "end",
     values = ("Euler's Number", "e^(1)", "math.exp(1)"))
    constants_table.insert(parent = "", index = "end",
     values = ("Golden Ratio", "phi", "mpmath.phi"))
    constants_table.insert(parent = "", index = "end",
     values = ("Euler-Mascheroni Constant", "gamma", "mpmath.euler"))
    constants_table.insert(parent = "", index = "end",
     values = ("Glaisher-Kinkelin Constant", "A", "mpmath.glaisher"))
    constants_table.insert(parent = "", index = "end",
     values = ("Catalan's Constant", "(Phi(-1,2,1/2))/4",
     "(mpmath.lerchphi(-1,2,1/2))/4"))
    constants_table.insert(parent = "", index = "end",
     values = ("Infinity", "infinity", "math.inf"))
    constants_table.pack()
    
    #Functions
    
    functions_label = tk.Label(help_frame, text = "Functions",
     font = "Calibri 18 bold")
    functions_label.pack()
    functions_table = ttk.Treeview(help_frame, columns = ("column1",
     "column2", "column3"), show = "headings", height = 70)
    functions_table.heading("column1", text = "Mathematical Function")
    functions_table.column("column1", width = 300)
    functions_table.heading("column2", text = "Implementation")
    functions_table.heading("column3", text = "Python Function")
    functions_table.column("column3", width = 300)
    functions_table.insert(parent = "", index = 0,
     values = ("Absolute Value", "|x|", "math.fabs(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Exponential", "e^(x)", "math.exp(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Natural Logarith", "ln(x)", "math.log(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Common Logarithm", "log(x)", "math.log(x,10)"))
    functions_table.insert(parent = "", index = "end",
     values = ("General Logarithm", "log(x,a)", "math.log(x,a)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Square Root", "sqrt(x)", "math.sqrt(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Cube Root", "cbrt(x)", "mpmath.cbrt(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("N-th Root", "root(x,n)", "mpmath.root(x,n)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Lambert W Function", "W(x)", "mpmath.lambertw(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Sine", "sin(x)", "math.sin(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Sine", "arcsin(x)", "math.asin(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Sine", "sinh(x)", "math.sinh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Sine", "arsinh(x)",
     "math.asinh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Cosine", "cos(x)", "math.cos(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Cosine", "arccos(x)", "math.acos(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Cosine", "cosh(x)", "math.cosh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Cosine", "arcosh(x)",
     "math.acosh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Tangent", "tan(x)", "math.tan(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Tangent", "arctan(x)", "math.atan(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Tangent", "tanh(x)", "math.tanh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Tangent", "artanh(x)",
     "math.atanh(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Secant", "sec(x)", "mpmath.sec(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Secant", "arcsec(x)", "mpmath.asec(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Secant", "sech(x)", "mpmath.sech(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Secant", "arsech(x)",
     "mpmath.asech(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Cosecant", "csc(x)", "mpmath.csc(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Cosecant", "arccsc(x)", "mpmath.acsc(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Cosecant", "csch(x)", "mpmath.csch(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Cosecant", "arcsch(x)",
     "mpmath.acsch(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Cotangent", "cot(x)", "mpmath.cot(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Cotangent", "arccot(x)", "mpmath.acot(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Cotangent", "coth(x)", "mpmath.coth(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Inverse Hyperbolic Cotangent", "arcoth(x)",
     "mpmath.acoth(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Gamma Function", "Gamma(x)", "math.gamma(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Generalized Incomplete Gamma Function", "Gamma(x,a,b)",
     "mpmath.gammainc(x,a,b)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Digamma Function", "psi(x)", "mpmath.digamma(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Polygamma Function", "psi(n,x)", "mpmath.psi(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Barnes G Function", "G(x)", "mpmath.barnesg(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Exponential Integral Function", "Ei(x)",
     "mpmath.ei(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Generalized Exponential Integral Function", "En(n,x)",
     "mpmath.expint(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Logarithmic Integral Function", "li(x)",
     "mpmath.li(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Cosine Integral Function", "Ci(x)", "mpmath.ci(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Cosine Integral Function", "Chi(x)",
     "mpmath.chi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Sine Integral Function", "Si(x)", "mpmath.si(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hyperbolic Sine Integral Function", "Shi(x)",
     "mpmath.shi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Error Function", "erf(x)", "math.erf(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Complementary Error Function", "erfc(x)",
     "math.erfc(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Imaginary Error Function", "erfi(x)",
     "mpmath.erfi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Fresnel Sine Integral Function", "S(x)",
     "mpmath.fresnels(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Fresnel Cosine Integral Function", "C(x)",
     "mpmath.fresnelc(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Bessel Function of the First Kind", "J(n,x)",
     "mpmath.besselj(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Bessel Function of the Second Kind", "Y(n,x)",
     "mpmath.bessely(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Modified Bessel Function of the First Kind", "I(n,x)",
     "mpmath.besseli(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Modified Bessel Function of the Second Kind", "K(n,x)",
     "mpmath.besselk(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Struve Function", "H(n,x)", "mpmath.struveh(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Modified Struve Function", "L(n,x)",
     "mpmath.struvel(n,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Weber Function", "E(v,x)", "mpmath.webere(v,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Lommel Function", "s(u,v,x)",
     "mpmath.lommels1(u,v,x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Airy Ai Function", "Ai(x)", "mpmath.airyai(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Airy Bi Function", "Bi(x)", "mpmath.airybi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Scorer Gi Function", "Gi(x)", "mpmath.scorergi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Scorer Hi Function", "Hi(x)", "mpmath.scorerhi(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Generalized Hypergeometric Function",
     "pFq([a_1,...,a_p],[b_1,...,b_q],x)",
     "mpmath.hyper([a_1,...,a_p],[b_1,...,b_q],x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Bilateral Hypergeometric Function",
     "AHB([a_1,...,a_A],[b_1,...,b_B],x)",
     "mpmath.bihyper([a_1,...,a_A],[b_1,...,b_B],x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Dirichlet Eta Function", "eta(x)",
     "mpmath.altzeta(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Jacobi Theta Function", "theta(n,x,q)",
     "mpmath.jtheta(n,x,q)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Riemann Zeta Function", "zeta(x)", "mpmath.zeta(x)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Hurwitz Zeta Function", "zeta(x,a)",
     "mpmath.zeta(x,a)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Lerch Transcendent Function", "Phi(x,s,a)",
     "mpmath.lerchphi(x,s,a)"))
    functions_table.insert(parent = "", index = "end",
     values = ("Polylogarithm", "Li(s,x)", "mpmath.polylog(s,x)"))
    functions_table.pack()
    
    #Numerical Algorithms
    
    numerical_algorithms_label = tk.Label(help_frame,
     text = "Numerical Algorithms", font = "Calibri 20 bold")
    numerical_algorithms_label.pack()
    label4 = '''This app implements many of the most commonly used
numerical methods. It includes a root finder, an interpolation
calculator, a numerical differentiator, a numerical integrator, and a
numerical solver. See the respective sections for instructions on how
to use these tools.'''
    numerical_algorithms_label1 = tk.Label(help_frame, text = label4,
     font = "Calibri 14")
    numerical_algorithms_label1.pack()
    
    #Root Finder
    
    root_finder_label1 = tk.Label(help_frame, text = "Root Finder",
     font = "Calibri 18 bold")
    root_finder_label1.pack()
    label5 = '''Root finding is the process of finding solutions,
also called roots or zeros, of the equation f(x) = 0. A similar
procedure is determining fixed points of a function, or values of x
which satisfy the equation f(x) = x. Some equations, such as x^2 - 7x
+ 12 = 0, can be easily solved analytically. However, equations like
tanx = x cannot be solved analytically. In these cases, numerical
techniques must be used to approximate a solution. The techniques
include bracketing methods and iterative methods.

Bracketing methods start with a closed interval [a,b] such that the
function is continuous over the interval and f(a) and f(b) have
opposite signs. By the Intermediate Value Theorem, it is then known
that a root exists between a and b. Bracketing methods shorten the
size of this interval while still maintaining the conditions of the
Intermediate Value Theorem, and proceeds until the interval is small
enough to approximate the root within a given tolerance. Bracketing
methods implemented in this app include the bisection method and the
method of false position.

Iterative methods begin with one or more initial approximations. These
data are then used to derive a better approximation which is then used
to derive an even better approximation. The process continues until a
root is found within a given tolerance. Fixed-point iteration, Newton's
method, the secant method, modified Newton's method, Steffensen's
method, and Muller's method are the iterative methods implemented in
this app.'''
    root_finder_label2 = tk.Label(help_frame, text = label5,
     font = "Calibri 14")
    root_finder_label2.pack()
    
    #Bisection method

    bisection_method_label1 = tk.Label(help_frame,
     text = "Bisection Method", font = "Calibri 16 bold")
    bisection_method_label1.pack()
    label6 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

a (float): Left endpoint of interval. The function value at f(a) must
have a different sign than at f(b).

b (float): Right endpoint of interval. The function value at f(b) must
have a different sign than at f(a).

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of f(x) = 0 and estimated error derived via the
bisection method, or message that the conditions of the Intermediate
Value Theorem are not met.'''
    bisection_method_label2 = tk.Label(help_frame, text = label6,
     font = "Calibri 14")
    bisection_method_label2.pack()
    
    #Fixed point iteration

    fixed_point_iteration_label1 = tk.Label(help_frame,
     text = "Fixed Point Iteration", font = "Calibri 16 bold")
    fixed_point_iteration_label1.pack()
    label7 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

x_0 (float): Initial approximation for fixed point.

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of f(x) = x and estimated error derived via
fixed-point interation.'''
    fixed_point_iteration_label2 = tk.Label(help_frame, text = label7,
     font = "Calibri 14")
    fixed_point_iteration_label2.pack()
    
    #Newton's Method

    newtons_method_label1 = tk.Label(help_frame,
     text = "Newton's Method", font = "Calibri 16 bold")
    newtons_method_label1.pack()
    label8 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

Derivative (string): Derivative of given function with respect to x.
See the Supported Constants and Functions section.

x_0 (float): Initial approximation for root.

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of f(x) = 0 and estimated error derived via
Newton's method.'''
    newtons_method_label2 = tk.Label(help_frame, text = label8,
     font = "Calibri 14")
    newtons_method_label2.pack()

    #Secant Method

    secant_method_label1 = tk.Label(help_frame, text = "Secant Method",
     font = "Calibri 16 bold")
    secant_method_label1.pack()
    label9 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions sectipn.

x_0 (float): First initial approximation for root.

x_1 (float): Second initial approximation for root.

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of f(x) = 0 and estimated error derived via the
secant method.'''
    secant_method_label2 = tk.Label(help_frame, text = label9,
     font = "Calibri 14")
    secant_method_label2.pack()

    #Method of False Position

    method_of_false_position_label1 = tk.Label(help_frame,
     text = "Method of False Position", font = "Calibri 16 bold")
    method_of_false_position_label1.pack()
    label10 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

a (float): Left endpoint of interval. The function value at f(a) must
have a different sign than at f(b).

b (float): Right endpoint of interval. The function value at f(b) must
have a different sign than at f(a).

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of f(x) = 0 and estimated error derived via the
method of false position, or message that the conditions of the
Intermediate Value Theorem are not met.'''
    method_of_false_position_label2 = tk.Label(help_frame,
     text = label10, font = "Calibri 14")
    method_of_false_position_label2.pack()

    #Modified Newton's Method

    modified_newtons_method_label1 = tk.Label(help_frame,
     text = "Modified Newton's Method", font = "Calibri 16 bold")
    modified_newtons_method_label1.pack()
    label11 = ''''''
    modified_newtons_method_label2 = tk.Label(help_frame,
     text = label11, font = "Calibri 14")
    modified_newtons_method_label2.pack()

    #Steffensen's Method

    steffensens_method_label1 = tk.Label(help_frame,
     text = "Steffensen's Method", font = "Calibri 16 bold")
    steffensens_method_label1.pack()
    label12 = ''''''
    steffensens_method_label2 = tk.Label(help_frame, text = label12,
     font = "Calibri 14")
    steffensens_method_label2.pack()

    #Muller's Method

    mullers_method_label1 = tk.Label(help_frame,
     text = "Mllers Method", font = "Calibri 16 bold")
    mullers_method_label1.pack()
    label13 = ''''''
    mullers_method_label2 = tk.Label(help_frame, text = label13,
     font = "Calibri 14")
    mullers_method_label2.pack()
    
    #Interpolation Calculator
    
    interpolation_calculator_label1 = tk.Label(help_frame,
     text = "Interpolation Calculator", font = "Calibri 18 bold")
    interpolation_calculator_label1.pack()
    label14 = '''Given any set of data there will be gaps in the data.
As a result, there are often missing data points for which information
needs to be determined. The process of estimating these missing data
points based on given data is called interpolation. Included in the
techniques of interpolation are polynomial interpolation and spline
interpolation.

Polynomial interpolation is based on the idea that given n data points
there is a unique polynomial of maximum degree n - 1 that agrees with
the data. Polynomial interpolation determines this polynomial and
evaluates it at the point(s) to be determined. Four methods of
polynomial interpolation implemented in this app include the use of
Lagrange interpolating polynomials, Neville's method, Newton's divided
difference method, and Hermite Interpolation.

Given n data points, spline interpolation fits a polynomial to the data
in each of the n - 1 subintervals. This app uses cubic splines,
meaning each subinterval approximates the data using a cubic
polynomial. These polynomials agree at each of the data points and
have continuous first and second order derivatives at the interior
data points. In order to obtain a unique solution, conditions must be
given at the two boundary points which are called boundary conditions.
For a natural cubic spline, the second derivative is equal to zero at
the boundary points. In a clamped cubic spline, the value of the first
derivative is specified at each boundary point. Natural cubic splines
and clamped cubic splines are implemented in this app.'''
    interpolation_calculator_label2 = tk.Label(help_frame,
     text = label14, font = "Calibri 14")
    interpolation_calculator_label2.pack()
    
    #Lagrange interpolating polynomials

    lagrange_interpolating_polynomials_label1 = tk.Label(help_frame,
     text = "Lagrange Interpolating Polynomials",
     font = "Calibri 16 bold")
    lagrange_interpolating_polynomials_label1.pack()
    label15 = \
'''Inputs:

x-values (string): x-coordinates of data set entered as floats
separated by commas.

y-values (string): y-coordinates of data set entered as floats
separated by commas.

x (float): Interpolation point.

    Output:

Estimated value of y-coordinate at interpolation point derived via
Lagrange interpolating polynomial.'''
    lagrange_interpolating_polynomials_label2 = tk.Label(help_frame,
     text = label15, font = "Calibri 14")
    lagrange_interpolating_polynomials_label2.pack()

    #Neville's method

    nevilles_method_label1 = tk.Label(help_frame,
     text = "Neville's Method", font = "Calibri 16 bold")
    nevilles_method_label1.pack()
    label16 = \
'''Inputs:

x-values (string): x-coordinates of data set entered as floats
separated by commas.

y-values (string): y-coordinates of data set entered as floats
separated by commas.

x (float): Interpolation point.

    Output:

Estimated value of y-coordinate at interpolation point derived via
Neville's method.'''
    nevilles_method_label2 = tk.Label(help_frame, text = label16,
     font = "Calibri 14")
    nevilles_method_label2.pack()

    #Newton's Divided Difference Method

    newtons_divided_difference_method_label1 = tk.Label(help_frame,
     text = "Newton's Divided Difference Method",
     font = "Calibri 16 bold")
    newtons_divided_difference_method_label1.pack()
    label17 = ''''''
    newtons_divided_difference_method_label2 = tk.Label(help_frame,
     text = label17, font = "Calibri 14")
    newtons_divided_difference_method_label2.pack()

    #Hermite interpolation

    hermite_interpolation_label1 = tk.Label(help_frame,
     text = "Hermite Interpolation", font = "Calibri 16 bold")
    hermite_interpolation_label1.pack()
    label18 = ''''''
    hermite_interpolation_label2 = tk.Label(help_frame, text = label18,
     font = "Calibri 14")
    hermite_interpolation_label2.pack()

    #Natural Cubic Spline

    natural_cubic_spline_label1 = tk.Label(help_frame,
     text = "Natural Cubic Spline", font = "Calibri 16 bold")
    natural_cubic_spline_label1.pack()
    label19 = ''''''
    natural_cubic_spline_label2 = tk.Label(help_frame, text = label19,
     font = "Calibri 14")
    natural_cubic_spline_label2.pack()

    #Clamped Cubic Spline

    clamped_cubic_spline_label1 = tk.Label(help_frame,
     text = "Clamped Cubic Spline", font = "Calibri 16 bold")
    clamped_cubic_spline_label1.pack()
    label20 = ''''''
    clamped_cubic_spline_label2 = tk.Label(help_frame, text = label20,
     font = "Calibri 14")
    clamped_cubic_spline_label2.pack()
    
    #Numerical Differentiator
    
    numerical_differentiator_label1 = tk.Label(help_frame,
     text = "Numerical Differentiator", font = "Calibri 18 bold")
    numerical_differentiator_label1.pack()
    label21 = '''Differentiation, along with integration, is one of
the two major operators taught in calculus. The chain rule, product
rule, and quotient rule are among the techniques used to compute
derivatives analytically. Due to the wide applicability of the
techniques the set of elementary functions is closed under
differentiation, that is every elementary function has an elementary
derivative. Even though any derivative can be found analytically, the
more complicated the function becomes the less practical it becomes to
compute the derivative by hand. In these cases, it is more practical
to compute the derivative numerically. The finite difference methods,
such as those included in this app, are the most common methods used
to compute derivatives numerically.

Finite difference methods use a given step size and the given
function evaluated at one or more points to approximate the derivative
at a given point. The smaller the step size and the more points used
in the evaluation, the more accurate the approximation. The
definition of the derivative in terms of the difference quotient is
used to derive the forward, backward, and central difference formulas.
Higher order methods, including the three and five point formulas can
be derived using Taylor series. The forward difference method, the
backward difference method, the central difference method, the
three-point midpoint formula, the three-point endpoint formula, the
five-point midpoint formula, and the five-point endpoint formula are
the finite difference methods implemented in this app.'''
    numerical_differentiator_label2 = tk.Label(help_frame,
     text = label21, font = "Calibri 14")
    numerical_differentiator_label2.pack()
    
    #Forward Difference

    forward_difference_label1 = tk.Label(help_frame,
     text = "Forward Difference", font = "Calibri 16 bold")
    forward_difference_label1.pack()
    label22 = ''''''
    forward_difference_label2 = tk.Label(help_frame, text = label22,
     font = "Calibri 14")
    forward_difference_label2.pack()

    #Backward Difference

    backward_difference_label1 = tk.Label(help_frame,
     text = "Backward Difference", font = "Calibri 16 bold")
    backward_difference_label1.pack()
    label23 = ''''''
    backward_difference_label2 = tk.Label(help_frame, text = label23,
     font = "Calibri 14")
    backward_difference_label2.pack()

    #Central Difference

    central_difference_label1 = tk.Label(help_frame,
     text = "Central Difference", font = "Calibri 16 bold")
    central_difference_label1.pack()
    label24 = ''''''
    central_difference_label2 = tk.Label(help_frame, text = label24,
     font = "Calibri 14")
    central_difference_label2.pack()

    #Three-point midpoint formula

    three_point_midpoint_formula_label1 = tk.Label(help_frame,
     text = "Three-Point Midpoint Formula", font = "Calibri 16 bold")
    three_point_midpoint_formula_label1.pack()
    label25 = ''''''
    three_point_midpoint_formula_label2 = tk.Label(help_frame,
     text = label25, font = "Calibri 14")
    three_point_midpoint_formula_label2.pack()

    #Three-point endpoint formula

    three_point_endpoint_formula_label1 = tk.Label(help_frame,
     text = "Three-Point Endpoint Formula", font = "Calibri 16 bold")
    three_point_endpoint_formula_label1.pack()
    label26 = ''''''
    three_point_endpoint_formula_label2 = tk.Label(help_frame,
     text = label26, font = "Calibri 14")
    three_point_endpoint_formula_label2.pack()

    #Five-point midpoint formula

    five_point_midpoint_formula_label1 = tk.Label(help_frame,
     text = "Five-Point Midpoint Formula", font = "Calibri 16 bold")
    five_point_midpoint_formula_label1.pack()
    label27 = ''''''
    five_point_midpoint_formula_label2 = tk.Label(help_frame,
     text = label27, font = "Calibri 14")
    five_point_midpoint_formula_label2.pack()

    #Five-point endpoint formula

    five_point_endpoint_formula_label1 = tk.Label(help_frame,
     text = "Five-Point Endpoint Formula", font = "Calibri 16 bold")
    five_point_endpoint_formula_label1.pack()
    label28 = ''''''
    five_point_endpoint_formula_label2 = tk.Label(help_frame,
     text = label28, font = "Calibri 14")
    five_point_endpoint_formula_label2.pack()
    
    #Numerical Integrator
    
    numerical_integrator_label1 = tk.Label(help_frame,
     text = "Numerical Integrator", font = "Calibri 18 bold")
    numerical_integrator_label1.pack()
    label29 = '''Integration, along with differentiation, is one of
the two major operators taught in calculus. Substitution, integration
by parts, and partial fraction decomposition are just a few of the
many techniques used to compute integrals. Techniques such as
substitution and integration by parts are general techniques and can
be applied in many circumstances. However, techniques such as partial
fraction decomposition and trigonometric substitution are more
specialized and apply in specific cases. As a result, the set of
elementary functions is not closed under integration, that is not
every elementary function has an elementary antiderivative. In cases
where an integrand does not have an elementary antiderivative, more
sophisticated techniques such as Taylor series, differentiation under
the integral sign, and contour integration can on occasion be used to
obtain a solution analytically. For these reasons, integration is
often a very difficult process, if even possible, and it is often more
practical to use numerical methods to approximate integrals. These
methods include the Newton-Cotes formulas, extrapolation, and
quadrature. All of these techniques make use of the fact that the
integral is just the signed area under a curve.

The Newton-Cotes formulas break up the interval of integration into
subintervals, and then uses polynomial interpolation and equally
spaced points to approximate the area over each subinterval. It then
sums up all these areas to get the total area. The more subintervals,
the more accurate the approximation. The Newton-Cotes formulas
implemented in this app include the midpoint rule, the trapezoid rule,
and Simpson's rule.

The combination of a Newton-Cotes formula along with the technique of
extrapolation is called Romberg integration. An advantage of this
method is that it can be used to approximate an integral within a
given tolerance. Romberg integration is implemented in this app.

Quadrature is similar to the use of the Newton-Cotes formulas, but
instead of using predetermined weights and having the user specify the
number of subintervals, it chooses the grid spacing and the weights at
each point such that the integral is approximated as accurately as
possible. Adaptive quadrature and Gaussian quadrature are implemented
in this app.'''
    numerical_integrator_label2 = tk.Label(help_frame, text = label29,
     font = "Calibri 14")
    numerical_integrator_label2.pack()
    
    #Midpoint rule

    midpoint_rule_label1 = tk.Label(help_frame, text = "Midpoint Rule",
     font = "Calibri 16 bold")
    midpoint_rule_label1.pack()
    label30 = ''''''
    midpoint_rule_label2 = tk.Label(help_frame, text = label30,
     font = "Calibri 14")
    midpoint_rule_label2.pack()

    #Trapezoid rule

    trapezoid_rule_label1 = tk.Label(help_frame,
     text = "Trapezoid Rule", font = "Calibri 16 bold")
    trapezoid_rule_label1.pack()
    label31 = ''''''
    trapezoid_rule_label2 = tk.Label(help_frame, text = label31,
     font = "Calibri 14")
    trapezoid_rule_label2.pack()

    #Simpson's rule

    simpsons_rule_label1 = tk.Label(help_frame,
     text = "Simpson's Rule", font = "Calibri 16 bold")
    simpsons_rule_label1.pack()
    label32 = ''''''
    simpsons_rule_label2 = tk.Label(help_frame, text = label32,
     font = "Calibri 14")
    simpsons_rule_label2.pack()

    #Romberg Integration

    romberg_integration_label1 = tk.Label(help_frame,
     text = "Romberg Integration", font = "Calibri 16 bold")
    romberg_integration_label1.pack()
    label33 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

a (float): Lower limit of integration.

b (float): Upper limit of integration.

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of the integral and estimated error derived via
Romberg integration.

    Note: This algorithm uses the trapezoid rule as the base method.'''
    romberg_integration_label2 = tk.Label(help_frame, text = label33,
     font = "Calibri 14")
    romberg_integration_label2.pack()

    #Adaptive quadrature

    adaptive_quadrature_label1 = tk.Label(help_frame,
     text = "Adaptive Quadrature", font = "Calibri 16 bold")
    adaptive_quadrature_label1.pack()
    label34 = \
'''Inputs:

Function (string): Function with independent variable x. See the
Supported Constants and Functions section.

a (float): Lower limit of integration.

b (float): Upper limit of integration.

Tolerance (float): Acceptable error for approximation.

    Output:

Approximate solution of the integral and estimated error derived via adaptive quadrature.

    Note: This algorithm uses Simpson's rule as the base method.'''
    adaptive_quadrature_label2 = tk.Label(help_frame, text = label34,
     font = "Calibri 14")
    adaptive_quadrature_label2.pack()

    #Gaussian quadrature

    gaussian_quadrature_label1 = tk.Label(help_frame,
     text = "Gaussian Quadrature", font = "Calibri 16 bold")
    gaussian_quadrature_label1.pack()
    label35 = ''''''
    gaussian_quadrature_label2 = tk.Label(help_frame, text = label35,
     font = "Calibri 14")
    gaussian_quadrature_label2.pack()
    
    #Numerical Solver
    
    numerical_solver_label1 = tk.Label(help_frame,
     text = "Numerical Solver", font = "Calibri 18 bold")
    numerical_solver_label1.pack()
    label36 = '''Differential equations are equations which contain
one or more independent variables, a function and one or more of its
derivatives. An ordinary differential equation (ODE) contains only
ordinary derivatives, whereas a partial differential equation (PDE)
contains partial derivatives. The order of a differential equation is
the order of its highest order derivative. Analytical solution
techniques for solving first order ODEs include direct integration,
integrating factors, separation of variables, potential functions, and
substitution. However, these techniques can only be applied in very
specific cases. In the vast majority of cases, one must resort to the
use of numerical methods to approximate solutions to ODEs. This app
implements various techniques for approximating solutions of first
order ODEs including linear multistep methods and Runge-Kutta methods.

Linear multistep methods use a given step size to estimate the value
of the solution to an ODE until it achieves an estimate for the
solution at the desired point. At each step, an estimate for the
solution is obtained using the value of the function and the
derivative at previous points. Euler's method, Adams-Bashforth
technique, Milue's method, and extrapolation are the linear multistep
methods implemented in this app.

Runge-Kutta methods are similar to the linear multistep methods except
the Rung-Kutta methods also use the values of the function and
derivative at half step sizes to compute an estimate at each point.
The Runge-Kutta methods implemented in this app are the midpoint
method, the modified Euler method, Heun's method, the Runge-Kutta
method, and the Runge-Kutta-Fehlberg method.'''
    numerical_solver_label2 = tk.Label(help_frame, text = label36,
     font = "Calibri 14")
    numerical_solver_label2.pack()
    
    #Euler's Method

    eulers_method_label1 = tk.Label(help_frame,
     text = "Euler's Method", font = "Calibri 16 bold")
    eulers_method_label1.pack()
    label37 = ''''''
    eulers_method_label2 = tk.Label(help_frame, text = label37,
     font = "Calibri 14")
    eulers_method_label2.pack()

    #Midpoint Method

    midpoint_method_label1 = tk.Label(help_frame,
     text = "Midpoint Method", font = "Calibri 16 bold")
    midpoint_method_label1.pack()
    label38 = ''''''
    midpoint_method_label2 = tk.Label(help_frame, text = label38,
     font = "Calibri 14")
    midpoint_method_label2.pack()

    #Modified Euler Method

    modified_euler_method_label1 = tk.Label(help_frame,
     text = "Modified Euler Method", font = "Calibri 16 bold")
    modified_euler_method_label1.pack()
    label39 = ''''''
    modified_euler_method_label2 = tk.Label(help_frame, text = label39,
     font = "Calibri 14")
    modified_euler_method_label2.pack()

    #Heun's Method

    heuns_method_label1 = tk.Label(help_frame, text = "Heun's Method",
     font = "Calibri 16 bold")
    heuns_method_label1.pack()
    label40 = ''''''
    heuns_method_label2 = tk.Label(help_frame, text = label40,
     font = "Calibri 14")
    heuns_method_label2.pack()

    #Runge-Kutta Method

    runge_kutta_method_label1 = tk.Label(help_frame,
     text = "Runge-Kutta Method", font = "Calibri 16 bold")
    runge_kutta_method_label1.pack()
    label41 = ''''''
    runge_kutta_method_label2 = tk.Label(help_frame, text = label41,
     font = "Calibri 14")
    runge_kutta_method_label2.pack()

    #Runge-Kutta-Fehlberg Method

    runge_kutta_fehlberg_method_label1 = tk.Label(help_frame,
     text = "Runge-Kutta-Fehlberg Method", font = "Calibri 16 bold")
    runge_kutta_fehlberg_method_label1.pack()
    label42 = ''''''
    runge_kutta_fehlberg_method_label2 = tk.Label(help_frame,
     text = label42, font = "Calibri 14")
    runge_kutta_fehlberg_method_label2.pack()

    #Adams-Bashforth Technique

    adams_bashforth_technique_label1 = tk.Label(help_frame,
     text = "Adams-Bashforth Technique", font = "Calibri 16 bold")
    adams_bashforth_technique_label1.pack()
    label43 = ''''''
    adams_bashforth_technique_label2 = tk.Label(help_frame,
     text = label43, font = "Calibri 14")
    adams_bashforth_technique_label2.pack()

    #Milue's Method

    milues_method_label1 = tk.Label(help_frame, text = "Milue's Method",
     font = "Calibri 16 bold")
    milues_method_label1.pack()
    label44 = ''''''
    milues_method_label2 = tk.Label(help_frame, text = label44,
     font = "Calibri 14")
    milues_method_label2.pack()

    #Extrapolation

    extrapolation_label1 = tk.Label(help_frame, text = "Extrapolation",
     font = "Calibri 16 bold")
    extrapolation_label1.pack()
    label45 = ''''''
    extrapolation_label2 = tk.Label(help_frame, text = label45,
     font = "Calibri 14")
    extrapolation_label2.pack()
    
    #Close
    
    help_close_button = tk.Button(window2, text = "Close",
     command = close_help)
    help_close_button.pack()

def acknowledgements():
    
    #Set up window
    
    global window3
    window3 = tk.Toplevel()
    window3.title("Acknowledgements")
    window3.geometry("1250x500")
    
    #Title
    
    acknowledgements_title = tk.Label(window3,
     text = "Acknowledgements", font = "Calibri 24 bold")
    acknowledgements_title.pack()
    
    #Scrollbar
    
    main_frame = tk.Frame(window3)
    main_frame.pack(fill = "both", expand = True)
    acknowledgements_canvas = tk.Canvas(main_frame)
    acknowledgements_canvas.pack(side = "left", fill = "both",
     expand = True)
    acknowledgements_scrollbar = ttk.Scrollbar(main_frame,
     orient = "vertical", command = acknowledgements_canvas.yview)
    acknowledgements_scrollbar.pack(side = "right", fill = "y")
    acknowledgements_canvas.configure(yscrollcommand = \
     acknowledgements_scrollbar.set)
    acknowledgements_canvas.bind("<Configure>",
     lambda e: acknowledgements_canvas.configure(scrollregion = \
     acknowledgements_canvas.bbox("all")))
    acknowledgements_frame = tk.Frame(acknowledgements_canvas)
    acknowledgements_canvas.create_window((0, 0),
     window = acknowledgements_frame, anchor = "nw")
    
    #Acknowledgements
    
    label = \
    '''I acknowledge the extensive use of the math and mpmath Python
modules in this project and am grateful for the work done by the
creators of these packages. Without these libraries this project would
have been much harder.

The book Numerical Analysis by Burden & Faires, 9th edition was used
as a reference over the course of the project, and I thank the authors
of this book which served as a valuable resource.

Furthermore, Wolfram MathWorld allowed me to learn about the various
special functions I did not know about prior to this project. I thank
the creators of this website, for the material I referred to during
this project. As well as serving as a great resource for nearly every
area of mathematics.

I thank the owner of the YouTube channel Clear Code. His tkinter
tutorial helped me learn how to create user interfaces in Python.
Including the one for this project.

I thank Liu Zuo Lin for his Medium article "Another Python Question
That Took Me Days To Solve As A Beginner" on how to evaluate Python
strings as mathematical expressions. This helped me avoid the use of
the eval and exec functions, which have been shown to be insecure.'''
    acknowledgements_label = tk.Label(acknowledgements_frame,
     text = label, font = "Calibri 14")
    acknowledgements_label.pack()
    
    #Close
    
    acknowledgements_close_button = tk.Button(window3, text = "Close",
     command = close_acknowledgements)
    acknowledgements_close_button.pack()
    
def close_help():
    window2.destroy()
    
def close_acknowledgements():
    window3.destroy()

#Set up window

window = tk.Tk()
window.state("zoomed")
window.title("Numerical Methods Calculator")

#Set Title Label

title = tk.Label(window, text = "Numerical Methods Calculator",
 font = "Calibri 24 bold")
title.pack()

#Create main notebook

notebook_main = ttk.Notebook(window)

#Root finder

root_finder = tk.Frame(notebook_main)

#Create root finder notebook

root_finder_notebook = ttk.Notebook(root_finder)

#Bisection method

bisection_method = tk.Frame(root_finder_notebook)
bisection_method_label = tk.Label(bisection_method,
 text = "Bisection Method", font = "Calibri 20 bold")
bisection_method_label.pack()

#Function

bisection_method_frame1 = tk.Frame(bisection_method)
bisection_method_frame1.pack()
bisection_method_function_label = tk.Label(bisection_method_frame1,
 text = "f(x) = ", font = "Calibri 16")
bisection_method_function_label.pack(side = "left")
bisection_method_function_entry_variable = tk.StringVar()
bisection_method_function_entry = tk.Entry(bisection_method_frame1,
 font = ("Calibri 16"),
 textvariable = bisection_method_function_entry_variable)
bisection_method_function_entry.pack(side = "left")

#Interval

bisection_method_frame2 = tk.Frame(bisection_method)
bisection_method_frame2.pack()
bisection_method_interval_label1 = tk.Label(bisection_method_frame2,
 text = "[a,b] = [", font = "Calibri 16")
bisection_method_interval_label1.pack(side = "left")
bisection_method_interval_entry_variable1 = tk.DoubleVar()
bisection_method_interval_entry1 = tk.Entry(bisection_method_frame2,
 font = ("Calibri 16"),
 textvariable = bisection_method_interval_entry_variable1)
bisection_method_interval_entry1.pack(side = "left")
bisection_method_interval_label2 = tk.Label(bisection_method_frame2,
 text = ",", font = "Calibri 16")
bisection_method_interval_label2.pack(side = "left")
bisection_method_interval_entry_variable2 = tk.DoubleVar()
bisection_method_interval_entry2 = tk.Entry(bisection_method_frame2,
 font = ("Calibri 16"),
 textvariable = bisection_method_interval_entry_variable2)
bisection_method_interval_entry2.pack(side = "left")
bisection_method_interval_label3 = tk.Label(bisection_method_frame2,
 text = "]", font = "Calibri 16")
bisection_method_interval_label3.pack(side = "left")

#Tolerance

bisection_method_frame3 = tk.Frame(bisection_method)
bisection_method_frame3.pack()
bisection_method_tolerance_label = tk.Label(bisection_method_frame3,
 text = "Tolerance: ", font = "Calibri 16")
bisection_method_tolerance_label.pack(side = "left")
bisection_method_tolerance_variable = tk.DoubleVar()
bisection_method_tolerance_entry = tk.Entry(bisection_method_frame3,
 font = ("Calibri 16"),
 textvariable = bisection_method_tolerance_variable)
bisection_method_tolerance_entry.pack(side = "left")

#Submit

bisection_method_submit_button = tk.Button(bisection_method,
 text = "Submit", command = bisection_method_submit)
bisection_method_submit_button.pack()

#Output

bisection_method_output_string = tk.StringVar()
bisection_method_output_label = tk.Label(bisection_method,
 textvariable = bisection_method_output_string, font = "Calibri 16")
bisection_method_output_label.pack()

#Fixed point iteration

fixed_point_iteration = tk.Frame(root_finder_notebook)
fixed_point_iteration_label = tk.Label(fixed_point_iteration,
 text = "Fixed Point Iteration", font = "Calibri 20 bold")
fixed_point_iteration_label.pack()

#Function

fixed_point_iteration_frame1 = tk.Frame(fixed_point_iteration)
fixed_point_iteration_frame1.pack()
fixed_point_iteration_function_label = \
 tk.Label(fixed_point_iteration_frame1, text = "x = ",
 font = "Calibri 16")
fixed_point_iteration_function_label.pack(side = "left")
fixed_point_iteration_function_entry_variable = tk.StringVar()
fixed_point_iteration_function_entry = \
 tk.Entry(fixed_point_iteration_frame1, font = ("Calibri 16"),
 textvariable = fixed_point_iteration_function_entry_variable)
fixed_point_iteration_function_entry.pack(side = "left")

#Initial Value

fixed_point_iteration_frame2 = tk.Frame(fixed_point_iteration)
fixed_point_iteration_frame2.pack()
subscripts = str.maketrans("0", "")
label = "x0 = "
label = label.translate(subscripts)
fixed_point_iteration_initial_value_label = \
 tk.Label(fixed_point_iteration_frame2, text = label,
 font = "Calibri 16")
fixed_point_iteration_initial_value_label.pack(side = "left")
fixed_point_iteration_initial_value_entry_variable = tk.DoubleVar()
fixed_point_iteration_initial_value_entry = \
 tk.Entry(fixed_point_iteration_frame2, font = ("Calibri 16"),
 textvariable = fixed_point_iteration_initial_value_entry_variable)
fixed_point_iteration_initial_value_entry.pack(side = "left")

#Tolerance

fixed_point_iteration_frame3 = tk.Frame(fixed_point_iteration)
fixed_point_iteration_frame3.pack()
fixed_point_iteration_tolerance_label = \
 tk.Label(fixed_point_iteration_frame3, text = "Tolerance: ",
 font = "Calibri 16")
fixed_point_iteration_tolerance_label.pack(side = "left")
fixed_point_iteration_tolerance_variable = tk.DoubleVar()
fixed_point_iteration_tolerance_entry = \
 tk.Entry(fixed_point_iteration_frame3, font = ("Calibri 16"),
 textvariable = fixed_point_iteration_tolerance_variable)
fixed_point_iteration_tolerance_entry.pack(side = "left")

#Submit

fixed_point_iteration_submit_button = tk.Button(fixed_point_iteration,
 text = "Submit", command = fixed_point_iteration_submit)
fixed_point_iteration_submit_button.pack()

#Output

fixed_point_iteration_output_string = tk.StringVar()
fixed_point_iteration_output_label = tk.Label(fixed_point_iteration,
 textvariable = fixed_point_iteration_output_string,
 font = "Calibri 16")
fixed_point_iteration_output_label.pack()

#Newton's Method

newtons_method = tk.Frame(root_finder_notebook)
newtons_method_label = tk.Label(newtons_method,
 text = "Newton's Method", font = "Calibri 20 bold")
newtons_method_label.pack()

#Function & Derivative

newtons_method_frame1 = tk.Frame(newtons_method)
newtons_method_frame1.pack()
newtons_method_function_label1 = tk.Label(newtons_method_frame1,
 text = "f(x) = ", font = "Calibri 16")
newtons_method_function_label1.pack(side = "left")
newtons_method_function_entry_variable1 = tk.StringVar()
newtons_method_function_entry1 = tk.Entry(newtons_method_frame1,
 font = ("Calibri 16"),
 textvariable = newtons_method_function_entry_variable1)
newtons_method_function_entry1.pack(side = "left")
newtons_method_function_label2 = tk.Label(newtons_method_frame1,
 text = ", f'(x) = ", font = "Calibri 16")
newtons_method_function_label2.pack(side = "left")
newtons_method_function_entry_variable2 = tk.StringVar()
newtons_method_function_entry2 = tk.Entry(newtons_method_frame1,
 font = ("Calibri 16"),
 textvariable = newtons_method_function_entry_variable2)
newtons_method_function_entry2.pack(side = "left")

#Initial Value

newtons_method_frame2 = tk.Frame(newtons_method)
newtons_method_frame2.pack()
subscripts = str.maketrans("0", "")
label = "x0 = "
label = label.translate(subscripts)
newtons_method_initial_value_label = tk.Label(newtons_method_frame2,
 text = label, font = "Calibri 16")
newtons_method_initial_value_label.pack(side = "left")
newtons_method_initial_value_entry_variable = tk.DoubleVar()
newtons_method_initial_value_entry = tk.Entry(newtons_method_frame2,
 font = ("Calibri 16"),
 textvariable = newtons_method_initial_value_entry_variable)
newtons_method_initial_value_entry.pack(side = "left")

#Tolerance

newtons_method_frame3 = tk.Frame(newtons_method)
newtons_method_frame3.pack()
newtons_method_tolerance_label = tk.Label(newtons_method_frame3,
 text = "Tolerance: ", font = "Calibri 16")
newtons_method_tolerance_label.pack(side = "left")
newtons_method_tolerance_variable = tk.DoubleVar()
newtons_method_tolerance_entry = tk.Entry(newtons_method_frame3,
 font = ("Calibri 16"),
 textvariable = newtons_method_tolerance_variable)
newtons_method_tolerance_entry.pack(side = "left")

#Submit

newtons_method_submit_button = tk.Button(newtons_method,
 text = "Submit", command = newtons_method_submit)
newtons_method_submit_button.pack()

#Output

newtons_method_output_string = tk.StringVar()
newtons_method_output_label = tk.Label(newtons_method,
 textvariable = newtons_method_output_string, font = "Calibri 16")
newtons_method_output_label.pack()

#Secant Method

secant_method = tk.Frame(root_finder_notebook)
secant_method_label = tk.Label(secant_method, text = "Secant Method",
 font = "Calibri 20 bold")
secant_method_label.pack()

#Function

secant_method_frame1 = tk.Frame(secant_method)
secant_method_frame1.pack()
secant_method_function_label = tk.Label(secant_method_frame1,
 text = "f(x) = ", font = "Calibri 16")
secant_method_function_label.pack(side = "left")
secant_method_function_entry_variable = tk.StringVar()
secant_method_function_entry = tk.Entry(secant_method_frame1,
 font = ("Calibri 16"),
 textvariable = secant_method_function_entry_variable)
secant_method_function_entry.pack(side = "left")

#Initial values

secant_method_frame2 = tk.Frame(secant_method)
secant_method_frame2.pack()
subscripts = str.maketrans("01", "")
label1 = "x0 = "
label1 = label1.translate(subscripts)
secant_method_initial_value_label1 = tk.Label(secant_method_frame2,
 text = label1, font = "Calibri 16")
secant_method_initial_value_label1.pack(side = "left")
secant_method_initial_value_entry_variable1 = tk.DoubleVar()
secant_method_initial_value_entry1 = tk.Entry(secant_method_frame2,
 font = ("Calibri 16"),
 textvariable = secant_method_initial_value_entry_variable1)
secant_method_initial_value_entry1.pack(side = "left")
label2 = ", x1 = "
label2 = label2.translate(subscripts)
secant_method_initial_value_label2 = tk.Label(secant_method_frame2,
 text = label2, font = "Calibri 16")
secant_method_initial_value_label2.pack(side = "left")
secant_method_initial_value_entry_variable2 = tk.DoubleVar()
secant_method_initial_value_entry2 = tk.Entry(secant_method_frame2,
 font = ("Calibri 16"),
 textvariable = secant_method_initial_value_entry_variable2)
secant_method_initial_value_entry2.pack(side = "left")

#Tolerance

secant_method_frame3 = tk.Frame(secant_method)
secant_method_frame3.pack()
secant_method_tolerance_label = tk.Label(secant_method_frame3,
 text = "Tolerance: ", font = "Calibri 16")
secant_method_tolerance_label.pack(side = "left")
secant_method_tolerance_variable = tk.DoubleVar()
secant_method_tolerance_entry = tk.Entry(secant_method_frame3,
 font = ("Calibri 16"),
 textvariable = secant_method_tolerance_variable)
secant_method_tolerance_entry.pack(side = "left")

#Submit

secant_method_submit_button = tk.Button(secant_method, text = "Submit",
 command = secant_method_submit)
secant_method_submit_button.pack()

#Output

secant_method_output_string = tk.StringVar()
secant_method_output_label = tk.Label(secant_method,
 textvariable = secant_method_output_string, font = "Calibri 16")
secant_method_output_label.pack()

#Method of False Position

method_of_false_position = tk.Frame(root_finder_notebook)
method_of_false_position_label = tk.Label(method_of_false_position,
 text = "Method of False Position", font = "Calibri 20 bold")
method_of_false_position_label.pack()

#Function

method_of_false_position_frame1 = tk.Frame(method_of_false_position)
method_of_false_position_frame1.pack()
method_of_false_position_function_label = \
 tk.Label(method_of_false_position_frame1, text = "f(x) = ",
 font = "Calibri 16")
method_of_false_position_function_label.pack(side = "left")
method_of_false_position_function_entry_variable = tk.StringVar()
method_of_false_position_function_entry = \
 tk.Entry(method_of_false_position_frame1, font = ("Calibri 16"),
 textvariable = method_of_false_position_function_entry_variable)
method_of_false_position_function_entry.pack(side = "left")

#Interval

method_of_false_position_frame2 = tk.Frame(method_of_false_position)
method_of_false_position_frame2.pack()
method_of_false_position_interval_label1 = \
 tk.Label(method_of_false_position_frame2, text = "[a,b] = [",
 font = "Calibri 16")
method_of_false_position_interval_label1.pack(side = "left")
method_of_false_position_interval_entry_variable1 = tk.DoubleVar()
method_of_false_position_interval_entry1 = \
 tk.Entry(method_of_false_position_frame2, font = ("Calibri 16"),
 textvariable = method_of_false_position_interval_entry_variable1)
method_of_false_position_interval_entry1.pack(side = "left")
method_of_false_position_interval_label2 = \
 tk.Label(method_of_false_position_frame2, text = ",",
 font = "Calibri 16")
method_of_false_position_interval_label2.pack(side = "left")
method_of_false_position_interval_entry_variable2 = tk.DoubleVar()
method_of_false_position_interval_entry2 = \
 tk.Entry(method_of_false_position_frame2, font = ("Calibri 16"),
 textvariable = method_of_false_position_interval_entry_variable2)
method_of_false_position_interval_entry2.pack(side = "left")
method_of_false_position_interval_label3 = \
 tk.Label(method_of_false_position_frame2, text = "]",
 font = "Calibri 16")
method_of_false_position_interval_label3.pack(side = "left")

#Tolerance

method_of_false_position_frame3 = tk.Frame(method_of_false_position)
method_of_false_position_frame3.pack()
method_of_false_position_tolerance_label = \
 tk.Label(method_of_false_position_frame3, text = "Tolerance: ",
 font = "Calibri 16")
method_of_false_position_tolerance_label.pack(side = "left")
method_of_false_position_tolerance_variable = tk.DoubleVar()
method_of_false_position_tolerance_entry = \
 tk.Entry(method_of_false_position_frame3,
 font = ("Calibri 16"),
 textvariable = method_of_false_position_tolerance_variable)
method_of_false_position_tolerance_entry.pack(side = "left")

#Submit

method_of_false_position_submit_button = \
 tk.Button(method_of_false_position, text = "Submit",
 command = method_of_false_position_submit)
method_of_false_position_submit_button.pack()

#Output

method_of_false_position_output_string = tk.StringVar()
method_of_false_position_output_label = \
 tk.Label(method_of_false_position,
 textvariable = method_of_false_position_output_string,
 font = "Calibri 16")
method_of_false_position_output_label.pack()

#Modified Newton's Method

modified_newtons_method = tk.Frame(root_finder_notebook)
modified_newtons_method_label = tk.Label(modified_newtons_method,
 text = "Modified Newton's Method", font = "Calibri 20 bold")
modified_newtons_method_label.pack()

#Function & Derivatives

modified_newtons_method_frame1 = tk.Frame(modified_newtons_method)
modified_newtons_method_frame1.pack()
modified_newtons_method_function_label1 = \
 tk.Label(modified_newtons_method_frame1, text = "f(x) = ",
 font = "Calibri 16")
modified_newtons_method_function_label1.pack(side = "left")
modified_newtons_method_function_entry_variable1 = tk.StringVar()
modified_newtons_method_function_entry1 = \
 tk.Entry(modified_newtons_method_frame1, font = ("Calibri 16"),
 textvariable = modified_newtons_method_function_entry_variable1)
modified_newtons_method_function_entry1.pack(side = "left")
modified_newtons_method_function_label2 = \
 tk.Label(modified_newtons_method_frame1, text = ", f'(x) = ",
 font = "Calibri 16")
modified_newtons_method_function_label2.pack(side = "left")
modified_newtons_method_function_entry_variable2 = tk.StringVar()
modified_newtons_method_function_entry2 = \
 tk.Entry(modified_newtons_method_frame1, font = ("Calibri 16"),
 textvariable = modified_newtons_method_function_entry_variable2)
modified_newtons_method_function_entry2.pack(side = "left")
modified_newtons_method_function_label3 = \
 tk.Label(modified_newtons_method_frame1, text = ", f''(x) = ",
 font = "Calibri 16")
modified_newtons_method_function_label3.pack(side = "left")
modified_newtons_method_function_entry_variable3 = tk.StringVar()
modified_newtons_method_function_entry3 = \
 tk.Entry(modified_newtons_method_frame1, font = ("Calibri 16"),
 textvariable = modified_newtons_method_function_entry_variable3)
modified_newtons_method_function_entry3.pack(side = "left")

#Initial Value

modified_newtons_method_frame2 = tk.Frame(modified_newtons_method)
modified_newtons_method_frame2.pack()
subscripts = str.maketrans("0", "")
label = "x0 = "
label = label.translate(subscripts)
modified_newtons_method_initial_value_label = \
 tk.Label(modified_newtons_method_frame2, text = label,
 font = "Calibri 16")
modified_newtons_method_initial_value_label.pack(side = "left")
modified_newtons_method_initial_value_entry_variable = tk.DoubleVar()
modified_newtons_method_initial_value_entry = \
 tk.Entry(modified_newtons_method_frame2, font = ("Calibri 16"),
 textvariable = modified_newtons_method_initial_value_entry_variable)
modified_newtons_method_initial_value_entry.pack(side = "left")

#Tolerance

modified_newtons_method_frame3 = tk.Frame(modified_newtons_method)
modified_newtons_method_frame3.pack()
modified_newtons_method_tolerance_label = \
 tk.Label(modified_newtons_method_frame3, text = "Tolerance: ",
 font = "Calibri 16")
modified_newtons_method_tolerance_label.pack(side = "left")
modified_newtons_method_tolerance_variable = tk.DoubleVar()
modified_newtons_method_tolerance_entry = \
 tk.Entry(modified_newtons_method_frame3, font = ("Calibri 16"),
 textvariable = modified_newtons_method_tolerance_variable)
modified_newtons_method_tolerance_entry.pack(side = "left")

#Submit

modified_newtons_method_submit_button = \
 tk.Button(modified_newtons_method, text = "Submit",
 command = modified_newtons_method_submit)
modified_newtons_method_submit_button.pack()

#Output

modified_newtons_method_output_string = tk.StringVar()
modified_newtons_method_output_label = \
 tk.Label(modified_newtons_method,
 textvariable = modified_newtons_method_output_string,
 font = "Calibri 16")
modified_newtons_method_output_label.pack()

#Steffensen's Method

steffensens_method = tk.Frame(root_finder_notebook)
steffensens_method_label = tk.Label(steffensens_method,
 text = "Steffensen's Method", font = "Calibri 20 bold")
steffensens_method_label.pack()

#Function

steffensens_method_frame1 = tk.Frame(steffensens_method)
steffensens_method_frame1.pack()
steffensens_method_function_label = \
 tk.Label(steffensens_method_frame1, text = "x = ",
 font = "Calibri 16")
steffensens_method_function_label.pack(side = "left")
steffensens_method_function_entry_variable = tk.StringVar()
steffensens_method_function_entry = \
 tk.Entry(steffensens_method_frame1, font = ("Calibri 16"),
 textvariable = steffensens_method_function_entry_variable)
steffensens_method_function_entry.pack(side = "left")

#Initial Value

steffensens_method_frame2 = tk.Frame(steffensens_method)
steffensens_method_frame2.pack()
subscripts = str.maketrans("0", "")
label = "x0 = "
label = label.translate(subscripts)
steffensens_method_initial_value_label = \
 tk.Label(steffensens_method_frame2, text = label, font = "Calibri 16")
steffensens_method_initial_value_label.pack(side = "left")
steffensens_method_initial_value_entry_variable = tk.DoubleVar()
steffensens_method_initial_value_entry = \
 tk.Entry(steffensens_method_frame2, font = ("Calibri 16"),
 textvariable = steffensens_method_initial_value_entry_variable)
steffensens_method_initial_value_entry.pack(side = "left")

#Tolerance

steffensens_method_frame3 = tk.Frame(steffensens_method)
steffensens_method_frame3.pack()
steffensens_method_tolerance_label = \
 tk.Label(steffensens_method_frame3, text = "Tolerance: ",
 font = "Calibri 16")
steffensens_method_tolerance_label.pack(side = "left")
steffensens_method_tolerance_variable = tk.DoubleVar()
steffensens_method_tolerance_entry = \
 tk.Entry(steffensens_method_frame3, font = ("Calibri 16"),
 textvariable = steffensens_method_tolerance_variable)
steffensens_method_tolerance_entry.pack(side = "left")

#Submit

steffensens_method_submit_button = tk.Button(steffensens_method,
 text = "Submit", command = steffensens_method_submit)
steffensens_method_submit_button.pack()

#Output

steffensens_method_output_string = tk.StringVar()
steffensens_method_output_label = tk.Label(steffensens_method,
 textvariable = steffensens_method_output_string, font = "Calibri 16")
steffensens_method_output_label.pack()

#Muller's Method

mullers_method = tk.Frame(root_finder_notebook)
mullers_method_label = tk.Label(mullers_method,
 text = "Mllers Method", font = "Calibri 20 bold")
mullers_method_label.pack()

#Function

mullers_method_frame1 = tk.Frame(mullers_method)
mullers_method_frame1.pack()
mullers_method_function_label = tk.Label(mullers_method_frame1,
 text = "f(x) = ", font = "Calibri 16")
mullers_method_function_label.pack(side = "left")
mullers_method_function_entry_variable = tk.StringVar()
mullers_method_function_entry = tk.Entry(mullers_method_frame1,
 font = ("Calibri 16"),
 textvariable = mullers_method_function_entry_variable)
mullers_method_function_entry.pack(side = "left")

#Initial values

mullers_method_frame2 = tk.Frame(mullers_method)
mullers_method_frame2.pack()
subscripts = str.maketrans("012", "")
label1 = "x0 = "
label1 = label1.translate(subscripts)
mullers_method_initial_value_label1 = tk.Label(mullers_method_frame2,
 text = label1, font = "Calibri 16")
mullers_method_initial_value_label1.pack(side = "left")
mullers_method_initial_value_entry_variable1 = tk.DoubleVar()
mullers_method_initial_value_entry1 = tk.Entry(mullers_method_frame2,
 font = ("Calibri 16"),
 textvariable = mullers_method_initial_value_entry_variable1)
mullers_method_initial_value_entry1.pack(side = "left")
label2 = ", x1 = "
label2 = label2.translate(subscripts)
mullers_method_initial_value_label2 = tk.Label(mullers_method_frame2,
 text = label2, font = "Calibri 16")
mullers_method_initial_value_label2.pack(side = "left")
mullers_method_initial_value_entry_variable2 = tk.DoubleVar()
mullers_method_initial_value_entry2 = tk.Entry(mullers_method_frame2,
 font = ("Calibri 16"),
 textvariable = mullers_method_initial_value_entry_variable2)
mullers_method_initial_value_entry2.pack(side = "left")
label3 = ", x2 = "
label3 = label3.translate(subscripts)
mullers_method_initial_value_label3 = tk.Label(mullers_method_frame2,
 text = label3, font = "Calibri 16")
mullers_method_initial_value_label3.pack(side = "left")
mullers_method_initial_value_entry_variable3 = tk.DoubleVar()
mullers_method_initial_value_entry3 = tk.Entry(mullers_method_frame2,
 font = ("Calibri 16"),
 textvariable = mullers_method_initial_value_entry_variable3)
mullers_method_initial_value_entry3.pack(side = "left")

#Tolerance

mullers_method_frame3 = tk.Frame(mullers_method)
mullers_method_frame3.pack()
mullers_method_tolerance_label = tk.Label(mullers_method_frame3,
 text = "Tolerance: ", font = "Calibri 16")
mullers_method_tolerance_label.pack(side = "left")
mullers_method_tolerance_variable = tk.DoubleVar()
mullers_method_tolerance_entry = tk.Entry(mullers_method_frame3,
 font = ("Calibri 16"),
 textvariable = mullers_method_tolerance_variable)
mullers_method_tolerance_entry.pack(side = "left")

#Submit

mullers_method_submit_button = tk.Button(mullers_method,
 text = "Submit", command = mullers_method_submit)
mullers_method_submit_button.pack()

#Output

mullers_method_output_string = tk.StringVar()
mullers_method_output_label = tk.Label(mullers_method,
 textvariable = mullers_method_output_string, font = "Calibri 16")
mullers_method_output_label.pack()

#Create root finder notebook tabs

root_finder_notebook.add(bisection_method, text = "Bisection Method")
root_finder_notebook.add(fixed_point_iteration,
 text = "Fixed Point Iteration")
root_finder_notebook.add(newtons_method, text = "Newton's Method")
root_finder_notebook.add(secant_method, text = "Secant Method")
root_finder_notebook.add(method_of_false_position,
 text = "Method of False Position")
root_finder_notebook.add(modified_newtons_method,
 text = "Modified Newton's Method")
root_finder_notebook.add(steffensens_method,
 text = "Steffensen's Method")
root_finder_notebook.add(mullers_method, text = "Mllers Method")
root_finder_notebook.pack()

#Interpolation Calculator

interpolation_calculator = tk.Frame(notebook_main)

#Create interpolation calculator notebook

interpolation_calculator_notebook = \
 ttk.Notebook(interpolation_calculator)

#Lagrange interpolating polynomials

lagrange_interpolating_polynomials = \
 tk.Frame(interpolation_calculator_notebook)
lagrange_interpolating_polynomials_label = \
 tk.Label(lagrange_interpolating_polynomials,
 text = "Lagrange Interpolating Polynomials",
 font = "Calibri 20 bold")
lagrange_interpolating_polynomials_label.pack()

#x values

lagrange_interpolating_polynomials_frame1 = \
 tk.Frame(lagrange_interpolating_polynomials)
lagrange_interpolating_polynomials_frame1.pack()
lagrange_interpolating_polynomials_x_values_label = \
 tk.Label(lagrange_interpolating_polynomials_frame1,
 text = "x-values: ", font = "Calibri 16")
lagrange_interpolating_polynomials_x_values_label.pack(side = "left")
lagrange_interpolating_polynomials_x_values_entry_variable = \
 tk.StringVar()
lagrange_interpolating_polynomials_x_values_entry = \
 tk.Entry(lagrange_interpolating_polynomials_frame1,
 font = ("Calibri 16"),
 textvariable = \
 lagrange_interpolating_polynomials_x_values_entry_variable)
lagrange_interpolating_polynomials_x_values_entry.pack(side = "left")

#y values

lagrange_interpolating_polynomials_frame2 = \
 tk.Frame(lagrange_interpolating_polynomials)
lagrange_interpolating_polynomials_frame2.pack()
lagrange_interpolating_polynomials_y_values_label = \
 tk.Label(lagrange_interpolating_polynomials_frame2,
 text = "y-values: ", font = "Calibri 16")
lagrange_interpolating_polynomials_y_values_label.pack(side = "left")
lagrange_interpolating_polynomials_y_values_entry_variable = \
 tk.StringVar()
lagrange_interpolating_polynomials_y_values_entry = \
 tk.Entry(lagrange_interpolating_polynomials_frame2,
 font = ("Calibri 16"),
 textvariable = \
 lagrange_interpolating_polynomials_y_values_entry_variable)
lagrange_interpolating_polynomials_y_values_entry.pack(side = "left")

#Interpolation point

lagrange_interpolating_polynomials_frame3 = \
 tk.Frame(lagrange_interpolating_polynomials)
lagrange_interpolating_polynomials_frame3.pack()
lagrange_interpolating_polynomials_interpolation_point_label = \
 tk.Label(lagrange_interpolating_polynomials_frame3, text = "x = ",
 font = "Calibri 16")
lagrange_interpolating_polynomials_interpolation_point_label.pack(side = "left")
lagrange_interpolating_polynomials_interpolation_point_entry_variable = \
 tk.DoubleVar()
lagrange_interpolating_polynomials_interpolation_point_entry = \
 tk.Entry(lagrange_interpolating_polynomials_frame3,
 font = ("Calibri 16"),
 textvariable = \
 lagrange_interpolating_polynomials_interpolation_point_entry_variable)
lagrange_interpolating_polynomials_interpolation_point_entry.pack(side = "left")

#Submit

lagrange_interpolating_polynomials_submit_button = \
 tk.Button(lagrange_interpolating_polynomials, text = "Submit",
 command = lagrange_interpolating_polynomials_submit)
lagrange_interpolating_polynomials_submit_button.pack()

#output

lagrange_interpolating_polynomials_output_string = tk.StringVar()
lagrange_interpolating_polynomials_output_label = \
 tk.Label(lagrange_interpolating_polynomials,
 textvariable = lagrange_interpolating_polynomials_output_string,
 font = "Calibri 16")
lagrange_interpolating_polynomials_output_label.pack()

#Neville's method

nevilles_method = tk.Frame(interpolation_calculator_notebook)
nevilles_method_label = tk.Label(nevilles_method,
 text = "Neville's Method", font = "Calibri 20 bold")
nevilles_method_label.pack()

#x values

nevilles_method_frame1 = tk.Frame(nevilles_method)
nevilles_method_frame1.pack()
nevilles_method_x_values_label = tk.Label(nevilles_method_frame1,
 text = "x-values: ", font = "Calibri 16")
nevilles_method_x_values_label.pack(side = "left")
nevilles_method_x_values_entry_variable = tk.StringVar()
nevilles_method_x_values_entry = tk.Entry(nevilles_method_frame1,
 font = ("Calibri 16"),
 textvariable = nevilles_method_x_values_entry_variable)
nevilles_method_x_values_entry.pack(side = "left")

#y values

nevilles_method_frame2 = tk.Frame(nevilles_method)
nevilles_method_frame2.pack()
nevilles_method_y_values_label = tk.Label(nevilles_method_frame2,
 text = "y-values: ", font = "Calibri 16")
nevilles_method_y_values_label.pack(side = "left")
nevilles_method_y_values_entry_variable = tk.StringVar()
nevilles_method_y_values_entry = tk.Entry(nevilles_method_frame2,
 font = ("Calibri 16"),
 textvariable = nevilles_method_y_values_entry_variable)
nevilles_method_y_values_entry.pack(side = "left")

#Interpolation point

nevilles_method_frame3 = tk.Frame(nevilles_method)
nevilles_method_frame3.pack()
nevilles_method_interpolation_point_label = \
 tk.Label(nevilles_method_frame3, text = "x = ", font = "Calibri 16")
nevilles_method_interpolation_point_label.pack(side = "left")
nevilles_method_interpolation_point_entry_variable = tk.DoubleVar()
nevilles_method_interpolation_point_entry = \
 tk.Entry(nevilles_method_frame3, font = ("Calibri 16"),
 textvariable = nevilles_method_interpolation_point_entry_variable)
nevilles_method_interpolation_point_entry.pack(side = "left")

#Submit

nevilles_method_submit_button = tk.Button(nevilles_method,
 text = "Submit", command = nevilles_method_submit)
nevilles_method_submit_button.pack()

#output

nevilles_method_output_string = tk.StringVar()
nevilles_method_output_label = tk.Label(nevilles_method,
 textvariable = nevilles_method_output_string, font = "Calibri 16")
nevilles_method_output_label.pack()

#Newton's Divided Difference Method

newtons_divided_difference_method = tk.Frame(interpolation_calculator_notebook)
newtons_divided_difference_method_label = tk.Label(newtons_divided_difference_method,
                        text = "Newton's Divided Difference Method",
                        font = "Calibri 20 bold")
newtons_divided_difference_method_label.pack()

#x values

newtons_divided_difference_method_frame1 = \
 tk.Frame(newtons_divided_difference_method)
newtons_divided_difference_method_frame1.pack()
newtons_divided_difference_method_x_values_label = \
 tk.Label(newtons_divided_difference_method_frame1,
 text = "x-values: ", font = "Calibri 16")
newtons_divided_difference_method_x_values_label.pack(side = "left")
newtons_divided_difference_method_x_values_entry_variable = \
 tk.StringVar()
newtons_divided_difference_method_x_values_entry = \
 tk.Entry(newtons_divided_difference_method_frame1,
 font = ("Calibri 16"),
 textvariable = \
 newtons_divided_difference_method_x_values_entry_variable)
newtons_divided_difference_method_x_values_entry.pack(side = "left")

#y values

newtons_divided_difference_method_frame2 = \
 tk.Frame(newtons_divided_difference_method)
newtons_divided_difference_method_frame2.pack()
newtons_divided_difference_method_y_values_label = \
 tk.Label(newtons_divided_difference_method_frame2,
 text = "y-values: ", font = "Calibri 16")
newtons_divided_difference_method_y_values_label.pack(side = "left")
newtons_divided_difference_method_y_values_entry_variable = \
 tk.StringVar()
newtons_divided_difference_method_y_values_entry = \
 tk.Entry(newtons_divided_difference_method_frame2,
 font = ("Calibri 16"),
 textvariable = \
 newtons_divided_difference_method_y_values_entry_variable)
newtons_divided_difference_method_y_values_entry.pack(side = "left")

#Interpolation point

newtons_divided_difference_method_frame3 = \
 tk.Frame(newtons_divided_difference_method)
newtons_divided_difference_method_frame3.pack()
newtons_divided_difference_method_interpolation_point_label = \
 tk.Label(newtons_divided_difference_method_frame3, text = "x = ",
 font = "Calibri 16")
newtons_divided_difference_method_interpolation_point_label.pack(side = "left")
newtons_divided_difference_method_interpolation_point_entry_variable = \
 tk.DoubleVar()
newtons_divided_difference_method_interpolation_point_entry = \
 tk.Entry(newtons_divided_difference_method_frame3,
 font = ("Calibri 16"),
 textvariable = \
 newtons_divided_difference_method_interpolation_point_entry_variable)
newtons_divided_difference_method_interpolation_point_entry.pack(side = "left")

#Submit

newtons_divided_difference_method_submit_button = \
 tk.Button(newtons_divided_difference_method,
 text = "Submit", command = newtons_divided_difference_method_submit)
newtons_divided_difference_method_submit_button.pack()

#output

newtons_divided_difference_method_output_string = tk.StringVar()
newtons_divided_difference_method_output_label = \
 tk.Label(newtons_divided_difference_method,
 textvariable = newtons_divided_difference_method_output_string,
 font = "Calibri 16")
newtons_divided_difference_method_output_label.pack()

#Hermite interpolation

hermite_interpolation = tk.Frame(interpolation_calculator_notebook)
hermite_interpolation_label = tk.Label(hermite_interpolation,
 text = "Hermite Interpolation", font = "Calibri 20 bold")
hermite_interpolation_label.pack()

#x values

hermite_interpolation_frame1 = tk.Frame(hermite_interpolation)
hermite_interpolation_frame1.pack()
hermite_interpolation_x_values_label = \
 tk.Label(hermite_interpolation_frame1, text = "x-values: ",
 font = "Calibri 16")
hermite_interpolation_x_values_label.pack(side = "left")
hermite_interpolation_x_values_entry_variable = tk.StringVar()
hermite_interpolation_x_values_entry = \
 tk.Entry(hermite_interpolation_frame1,
 font = ("Calibri 16"),
 textvariable = hermite_interpolation_x_values_entry_variable)
hermite_interpolation_x_values_entry.pack(side = "left")

#y and y' values

hermite_interpolation_frame2 = tk.Frame(hermite_interpolation)
hermite_interpolation_frame2.pack()
hermite_interpolation_y_values_label = \
 tk.Label(hermite_interpolation_frame2, text = "y-values: ",
 font = "Calibri 16")
hermite_interpolation_y_values_label.pack(side = "left")
hermite_interpolation_y_values_entry_variable = tk.StringVar()
hermite_interpolation_y_values_entry = \
 tk.Entry(hermite_interpolation_frame2, font = ("Calibri 16"),
 textvariable = hermite_interpolation_y_values_entry_variable)
hermite_interpolation_y_values_entry.pack(side = "left")
hermite_interpolation_y_prime_values_label = \
 tk.Label(hermite_interpolation_frame2, text = ", y'-values: ",
 font = "Calibri 16")
hermite_interpolation_y_prime_values_label.pack(side = "left")
hermite_interpolation_y_prime_values_entry_variable = tk.StringVar()
hermite_interpolation_y_prime_values_entry = \
 tk.Entry(hermite_interpolation_frame2, font = ("Calibri 16"),
 textvariable = hermite_interpolation_y_prime_values_entry_variable)
hermite_interpolation_y_prime_values_entry.pack(side = "left")

#Interpolation point

hermite_interpolation_frame3 = tk.Frame(hermite_interpolation)
hermite_interpolation_frame3.pack()
hermite_interpolation_interpolation_point_label = \
 tk.Label(hermite_interpolation_frame3, text = "x = ",
 font = "Calibri 16")
hermite_interpolation_interpolation_point_label.pack(side = "left")
hermite_interpolation_interpolation_point_entry_variable = \
 tk.DoubleVar()
hermite_interpolation_interpolation_point_entry = \
 tk.Entry(hermite_interpolation_frame3,
 font = ("Calibri 16"),
 textvariable = \
 hermite_interpolation_interpolation_point_entry_variable)
hermite_interpolation_interpolation_point_entry.pack(side = "left")

#Submit

hermite_interpolation_submit_button = tk.Button(hermite_interpolation,
 text = "Submit", command = hermite_interpolation_submit)
hermite_interpolation_submit_button.pack()

#output

hermite_interpolation_output_string = tk.StringVar()
hermite_interpolation_output_label = tk.Label(hermite_interpolation,
 textvariable = hermite_interpolation_output_string,
 font = "Calibri 16")
hermite_interpolation_output_label.pack()

#Natural Cubic Spline

natural_cubic_spline = tk.Frame(interpolation_calculator_notebook)
natural_cubic_spline_label = tk.Label(natural_cubic_spline,
 text = "Natural Cubic Spline", font = "Calibri 20 bold")
natural_cubic_spline_label.pack()

#x values

natural_cubic_spline_frame1 = tk.Frame(natural_cubic_spline)
natural_cubic_spline_frame1.pack()
natural_cubic_spline_x_values_label = \
 tk.Label(natural_cubic_spline_frame1, text = "x-values: ",
 font = "Calibri 16")
natural_cubic_spline_x_values_label.pack(side = "left")
natural_cubic_spline_x_values_entry_variable = tk.StringVar()
natural_cubic_spline_x_values_entry = \
 tk.Entry(natural_cubic_spline_frame1, font = ("Calibri 16"),
 textvariable = natural_cubic_spline_x_values_entry_variable)
natural_cubic_spline_x_values_entry.pack(side = "left")

#y values

natural_cubic_spline_frame2 = tk.Frame(natural_cubic_spline)
natural_cubic_spline_frame2.pack()
natural_cubic_spline_y_values_label = \
 tk.Label(natural_cubic_spline_frame2, text = "y-values: ",
 font = "Calibri 16")
natural_cubic_spline_y_values_label.pack(side = "left")
natural_cubic_spline_y_values_entry_variable = tk.StringVar()
natural_cubic_spline_y_values_entry = \
 tk.Entry(natural_cubic_spline_frame2, font = ("Calibri 16"),
 textvariable = natural_cubic_spline_y_values_entry_variable)
natural_cubic_spline_y_values_entry.pack(side = "left")

#Interpolation point

natural_cubic_spline_frame3 = tk.Frame(natural_cubic_spline)
natural_cubic_spline_frame3.pack()
natural_cubic_spline_interpolation_point_label = \
 tk.Label(natural_cubic_spline_frame3, text = "x = ",
 font = "Calibri 16")
natural_cubic_spline_interpolation_point_label.pack(side = "left")
natural_cubic_spline_interpolation_point_entry_variable = tk.DoubleVar()
natural_cubic_spline_interpolation_point_entry = \
 tk.Entry(natural_cubic_spline_frame3, font = ("Calibri 16"),
 textvariable = \
 natural_cubic_spline_interpolation_point_entry_variable)
natural_cubic_spline_interpolation_point_entry.pack(side = "left")

#Submit

natural_cubic_spline_submit_button = tk.Button(natural_cubic_spline,
 text = "Submit", command = natural_cubic_spline_submit)
natural_cubic_spline_submit_button.pack()

#output

natural_cubic_spline_output_string = tk.StringVar()
natural_cubic_spline_output_label = tk.Label(natural_cubic_spline,
 textvariable = natural_cubic_spline_output_string,
 font = "Calibri 16")
natural_cubic_spline_output_label.pack()

#Clamped Cubic Spline

clamped_cubic_spline = tk.Frame(interpolation_calculator_notebook)
clamped_cubic_spline_label = tk.Label(clamped_cubic_spline,
 text = "Clamped Cubic Spline", font = "Calibri 20 bold")
clamped_cubic_spline_label.pack()

#x values

clamped_cubic_spline_frame1 = tk.Frame(clamped_cubic_spline)
clamped_cubic_spline_frame1.pack()
clamped_cubic_spline_x_values_label = \
 tk.Label(clamped_cubic_spline_frame1, text = "x-values: ",
 font = "Calibri 16")
clamped_cubic_spline_x_values_label.pack(side = "left")
clamped_cubic_spline_x_values_entry_variable = tk.StringVar()
clamped_cubic_spline_x_values_entry = \
 tk.Entry(clamped_cubic_spline_frame1, font = ("Calibri 16"),
 textvariable = clamped_cubic_spline_x_values_entry_variable)
clamped_cubic_spline_x_values_entry.pack(side = "left")

#y values

clamped_cubic_spline_frame2 = tk.Frame(clamped_cubic_spline)
clamped_cubic_spline_frame2.pack()
clamped_cubic_spline_y_values_label = \
 tk.Label(clamped_cubic_spline_frame2, text = "y-values: ",
 font = "Calibri 16")
clamped_cubic_spline_y_values_label.pack(side = "left")
clamped_cubic_spline_y_values_entry_variable = tk.StringVar()
clamped_cubic_spline_y_values_entry = \
 tk.Entry(clamped_cubic_spline_frame2, font = ("Calibri 16"),
 textvariable = clamped_cubic_spline_y_values_entry_variable)
clamped_cubic_spline_y_values_entry.pack(side = "left")

#Initial Conditions

clamped_cubic_spline_frame3 = tk.Frame(clamped_cubic_spline)
clamped_cubic_spline_frame3.pack()
clamped_cubic_spline_initial_condition_label1 = \
 tk.Label(clamped_cubic_spline_frame3, text = "f'(a) = ",
 font = "Calibri 16")
clamped_cubic_spline_initial_condition_label1.pack(side = "left")
clamped_cubic_spline_initial_condition_entry_variable1 = tk.DoubleVar()
clamped_cubic_spline_initial_condition_entry1 = \
 tk.Entry(clamped_cubic_spline_frame3, font = ("Calibri 16"),
 textvariable = clamped_cubic_spline_initial_condition_entry_variable1)
clamped_cubic_spline_initial_condition_entry1.pack(side = "left")
clamped_cubic_spline_initial_condition_label2 = \
 tk.Label(clamped_cubic_spline_frame3, text = ", f'(b) = ",
 font = "Calibri 16")
clamped_cubic_spline_initial_condition_label2.pack(side = "left")
clamped_cubic_spline_initial_condition_entry_variable2 = tk.DoubleVar()
clamped_cubic_spline_initial_condition_entry2 = \
 tk.Entry(clamped_cubic_spline_frame3, font = ("Calibri 16"),
 textvariable = clamped_cubic_spline_initial_condition_entry_variable2)
clamped_cubic_spline_initial_condition_entry2.pack(side = "left")

#Interpolation point

clamped_cubic_spline_frame4 = tk.Frame(clamped_cubic_spline)
clamped_cubic_spline_frame4.pack()
clamped_cubic_spline_interpolation_point_label = \
 tk.Label(clamped_cubic_spline_frame4, text = "x = ",
 font = "Calibri 16")
clamped_cubic_spline_interpolation_point_label.pack(side = "left")
clamped_cubic_spline_interpolation_point_entry_variable = \
 tk.DoubleVar()
clamped_cubic_spline_interpolation_point_entry = \
 tk.Entry(clamped_cubic_spline_frame4, font = ("Calibri 16"),
 textvariable = \
 clamped_cubic_spline_interpolation_point_entry_variable)
clamped_cubic_spline_interpolation_point_entry.pack(side = "left")

#Submit

clamped_cubic_spline_submit_button = tk.Button(clamped_cubic_spline,
 text = "Submit", command = clamped_cubic_spline_submit)
clamped_cubic_spline_submit_button.pack()

#output

clamped_cubic_spline_output_string = tk.StringVar()
clamped_cubic_spline_output_label = tk.Label(clamped_cubic_spline,
 textvariable = clamped_cubic_spline_output_string,
 font = "Calibri 16")
clamped_cubic_spline_output_label.pack()

#Create interpolation calculator notebook tabs

interpolation_calculator_notebook.add(lagrange_interpolating_polynomials,
 text = "Lagrange Interpolating Polynomials")
interpolation_calculator_notebook.add(nevilles_method,
 text = "Neville's Method")
interpolation_calculator_notebook.add(newtons_divided_difference_method,
 text = "Newton's Divided Difference Method")
interpolation_calculator_notebook.add(hermite_interpolation,
 text = "Hermite Interpolation")
interpolation_calculator_notebook.add(natural_cubic_spline,
 text = "Natural Cubic Spline")
interpolation_calculator_notebook.add(clamped_cubic_spline,
 text = "Clamped Cubic Spline")
interpolation_calculator_notebook.pack()

#Numerical Differentiator

numerical_differentiator = tk.Frame(notebook_main)

#Create numerical differentiator notebook

numerical_differentiator_notebook = \
 ttk.Notebook(numerical_differentiator)

#Forward Difference

forward_difference = tk.Frame(numerical_differentiator_notebook)
forward_difference_label = tk.Label(forward_difference,
 text = "Forward Difference", font = "Calibri 20 bold")
forward_difference_label.pack()

#Function

forward_difference_frame1 = tk.Frame(forward_difference)
forward_difference_frame1.pack()
forward_difference_function_label = tk.Label(forward_difference_frame1,
 text = "f(x) = ", font = "Calibri 16")
forward_difference_function_label.pack(side = "left")
forward_difference_function_entry_variable = tk.StringVar()
forward_difference_function_entry = tk.Entry(forward_difference_frame1,
 font = ("Calibri 16"),
 textvariable = forward_difference_function_entry_variable)
forward_difference_function_entry.pack(side = "left")

#x Value

forward_difference_frame2 = tk.Frame(forward_difference)
forward_difference_frame2.pack()
forward_difference_x_value_label = tk.Label(forward_difference_frame2,
 text = "x = ", font = "Calibri 16")
forward_difference_x_value_label.pack(side = "left")
forward_difference_x_value_entry_variable = tk.DoubleVar()
forward_difference_x_value_entry = tk.Entry(forward_difference_frame2,
 font = ("Calibri 16"),
 textvariable = forward_difference_x_value_entry_variable)
forward_difference_x_value_entry.pack(side = "left")

#h Value

forward_difference_frame3 = tk.Frame(forward_difference)
forward_difference_frame3.pack()
forward_difference_h_value_label = tk.Label(forward_difference_frame3,
 text = "h = ", font = "Calibri 16")
forward_difference_h_value_label.pack(side = "left")
forward_difference_h_value_entry_variable = tk.DoubleVar()
forward_difference_h_value_entry = tk.Entry(forward_difference_frame3,
 font = ("Calibri 16"),
 textvariable = forward_difference_h_value_entry_variable)
forward_difference_h_value_entry.pack(side = "left")

#Submit

forward_difference_submit_button = tk.Button(forward_difference,
 text = "Submit", command = forward_difference_submit)
forward_difference_submit_button.pack()

#output

forward_difference_output_string = tk.StringVar()
forward_difference_output_label = tk.Label(forward_difference,
 textvariable = forward_difference_output_string, font = "Calibri 16")
forward_difference_output_label.pack()

#Backward Difference

backward_difference = tk.Frame(numerical_differentiator_notebook)
backward_difference_label = tk.Label(backward_difference,
 text = "Backward Difference", font = "Calibri 20 bold")
backward_difference_label.pack()

#Function

backward_difference_frame1 = tk.Frame(backward_difference)
backward_difference_frame1.pack()
backward_difference_function_label = \
 tk.Label(backward_difference_frame1, text = "f(x) = ",
 font = "Calibri 16")
backward_difference_function_label.pack(side = "left")
backward_difference_function_entry_variable = tk.StringVar()
backward_difference_function_entry = \
 tk.Entry(backward_difference_frame1, font = ("Calibri 16"),
 textvariable = backward_difference_function_entry_variable)
backward_difference_function_entry.pack(side = "left")

#x Value

backward_difference_frame2 = tk.Frame(backward_difference)
backward_difference_frame2.pack()
backward_difference_x_value_label = \
 tk.Label(backward_difference_frame2, text = "x = ",
 font = "Calibri 16")
backward_difference_x_value_label.pack(side = "left")
backward_difference_x_value_entry_variable = tk.DoubleVar()
backward_difference_x_value_entry = \
 tk.Entry(backward_difference_frame2, font = ("Calibri 16"),
 textvariable = backward_difference_x_value_entry_variable)
backward_difference_x_value_entry.pack(side = "left")

#h Value

backward_difference_frame3 = tk.Frame(backward_difference)
backward_difference_frame3.pack()
backward_difference_h_value_label = \
 tk.Label(backward_difference_frame3, text = "h = ",
 font = "Calibri 16")
backward_difference_h_value_label.pack(side = "left")
backward_difference_h_value_entry_variable = tk.DoubleVar()
backward_difference_h_value_entry = \
 tk.Entry(backward_difference_frame3, font = ("Calibri 16"),
 textvariable = backward_difference_h_value_entry_variable)
backward_difference_h_value_entry.pack(side = "left")

#Submit

backward_difference_submit_button = tk.Button(backward_difference,
 text = "Submit", command = backward_difference_submit)
backward_difference_submit_button.pack()

#output

backward_difference_output_string = tk.StringVar()
backward_difference_output_label = tk.Label(backward_difference,
 textvariable = backward_difference_output_string, font = "Calibri 16")
backward_difference_output_label.pack()

#Central Difference

central_difference = tk.Frame(numerical_differentiator_notebook)
central_difference_label = tk.Label(central_difference,
 text = "Central Difference", font = "Calibri 20 bold")
central_difference_label.pack()

#Function

central_difference_frame1 = tk.Frame(central_difference)
central_difference_frame1.pack()
central_difference_function_label = tk.Label(central_difference_frame1,
 text = "f(x) = ", font = "Calibri 16")
central_difference_function_label.pack(side = "left")
central_difference_function_entry_variable = tk.StringVar()
central_difference_function_entry = tk.Entry(central_difference_frame1,
 font = ("Calibri 16"),
 textvariable = central_difference_function_entry_variable)
central_difference_function_entry.pack(side = "left")

#x Value

central_difference_frame2 = tk.Frame(central_difference)
central_difference_frame2.pack()
central_difference_x_value_label = tk.Label(central_difference_frame2,
 text = "x = ", font = "Calibri 16")
central_difference_x_value_label.pack(side = "left")
central_difference_x_value_entry_variable = tk.DoubleVar()
central_difference_x_value_entry = tk.Entry(central_difference_frame2,
 font = ("Calibri 16"),
 textvariable = central_difference_x_value_entry_variable)
central_difference_x_value_entry.pack(side = "left")

#h Value

central_difference_frame3 = tk.Frame(central_difference)
central_difference_frame3.pack()
central_difference_h_value_label = tk.Label(central_difference_frame3,
 text = "h = ", font = "Calibri 16")
central_difference_h_value_label.pack(side = "left")
central_difference_h_value_entry_variable = tk.DoubleVar()
central_difference_h_value_entry = \
 tk.Entry(central_difference_frame3, font = ("Calibri 16"),
 textvariable = central_difference_h_value_entry_variable)
central_difference_h_value_entry.pack(side = "left")

#Submit

central_difference_submit_button = tk.Button(central_difference,
 text = "Submit", command = central_difference_submit)
central_difference_submit_button.pack()

#output

central_difference_output_string = tk.StringVar()
central_difference_output_label = tk.Label(central_difference,
 textvariable = central_difference_output_string, font = "Calibri 16")
central_difference_output_label.pack()

#Three-point midpoint formula

three_point_midpoint_formula = \
 tk.Frame(numerical_differentiator_notebook)
three_point_midpoint_formula_label = \
 tk.Label(three_point_midpoint_formula,
 text = "Three-Point Midpoint Formula", font = "Calibri 20 bold")
three_point_midpoint_formula_label.pack()

#Function

three_point_midpoint_formula_frame1 = \
 tk.Frame(three_point_midpoint_formula)
three_point_midpoint_formula_frame1.pack()
three_point_midpoint_formula_function_label = \
 tk.Label(three_point_midpoint_formula_frame1, text = "f(x) = ",
 font = "Calibri 16")
three_point_midpoint_formula_function_label.pack(side = "left")
three_point_midpoint_formula_function_entry_variable = tk.StringVar()
three_point_midpoint_formula_function_entry = \
 tk.Entry(three_point_midpoint_formula_frame1, font = ("Calibri 16"),
 textvariable = three_point_midpoint_formula_function_entry_variable)
three_point_midpoint_formula_function_entry.pack(side = "left")

#x Value

three_point_midpoint_formula_frame2 = \
 tk.Frame(three_point_midpoint_formula)
three_point_midpoint_formula_frame2.pack()
three_point_midpoint_formula_x_value_label = \
 tk.Label(three_point_midpoint_formula_frame2, text = "x = ",
 font = "Calibri 16")
three_point_midpoint_formula_x_value_label.pack(side = "left")
three_point_midpoint_formula_x_value_entry_variable = tk.DoubleVar()
three_point_midpoint_formula_x_value_entry = \
 tk.Entry(three_point_midpoint_formula_frame2, font = ("Calibri 16"),
 textvariable = three_point_midpoint_formula_x_value_entry_variable)
three_point_midpoint_formula_x_value_entry.pack(side = "left")

#h Value

three_point_midpoint_formula_frame3 = \
 tk.Frame(three_point_midpoint_formula)
three_point_midpoint_formula_frame3.pack()
three_point_midpoint_formula_h_value_label = \
 tk.Label(three_point_midpoint_formula_frame3, text = "h = ",
 font = "Calibri 16")
three_point_midpoint_formula_h_value_label.pack(side = "left")
three_point_midpoint_formula_h_value_entry_variable = tk.DoubleVar()
three_point_midpoint_formula_h_value_entry = \
 tk.Entry(three_point_midpoint_formula_frame3, font = ("Calibri 16"),
 textvariable = three_point_midpoint_formula_h_value_entry_variable)
three_point_midpoint_formula_h_value_entry.pack(side = "left")

#Submit

three_point_midpoint_formula_submit_button = \
 tk.Button(three_point_midpoint_formula, text = "Submit",
 command = three_point_midpoint_formula_submit)
three_point_midpoint_formula_submit_button.pack()

#output

three_point_midpoint_formula_output_string = tk.StringVar()
three_point_midpoint_formula_output_label = \
 tk.Label(three_point_midpoint_formula,
 textvariable = three_point_midpoint_formula_output_string,
 font = "Calibri 16")
three_point_midpoint_formula_output_label.pack()

#Three-point endpoint formula

three_point_endpoint_formula = \
 tk.Frame(numerical_differentiator_notebook)
three_point_endpoint_formula_label = \
 tk.Label(three_point_endpoint_formula,
 text = "Three-Point Endpoint Formula", font = "Calibri 20 bold")
three_point_endpoint_formula_label.pack()

#Function

three_point_endpoint_formula_frame1 = \
 tk.Frame(three_point_endpoint_formula)
three_point_endpoint_formula_frame1.pack()
three_point_endpoint_formula_function_label = \
 tk.Label(three_point_endpoint_formula_frame1, text = "f(x) = ",
 font = "Calibri 16")
three_point_endpoint_formula_function_label.pack(side = "left")
three_point_endpoint_formula_function_entry_variable = tk.StringVar()
three_point_endpoint_formula_function_entry = \
 tk.Entry(three_point_endpoint_formula_frame1, font = ("Calibri 16"),
 textvariable = three_point_endpoint_formula_function_entry_variable)
three_point_endpoint_formula_function_entry.pack(side = "left")

#x Value

three_point_endpoint_formula_frame2 = \
 tk.Frame(three_point_endpoint_formula)
three_point_endpoint_formula_frame2.pack()
three_point_endpoint_formula_x_value_label = \
 tk.Label(three_point_endpoint_formula_frame2, text = "x = ",
 font = "Calibri 16")
three_point_endpoint_formula_x_value_label.pack(side = "left")
three_point_endpoint_formula_x_value_entry_variable = tk.DoubleVar()
three_point_endpoint_formula_x_value_entry = \
 tk.Entry(three_point_endpoint_formula_frame2, font = ("Calibri 16"),
 textvariable = three_point_endpoint_formula_x_value_entry_variable)
three_point_endpoint_formula_x_value_entry.pack(side = "left")

#h Value

three_point_endpoint_formula_frame3 = \
 tk.Frame(three_point_endpoint_formula)
three_point_endpoint_formula_frame3.pack()
three_point_endpoint_formula_h_value_label = \
 tk.Label(three_point_endpoint_formula_frame3, text = "h = ",
 font = "Calibri 16")
three_point_endpoint_formula_h_value_label.pack(side = "left")
three_point_endpoint_formula_h_value_entry_variable = tk.DoubleVar()
three_point_endpoint_formula_h_value_entry = \
 tk.Entry(three_point_endpoint_formula_frame3, font = ("Calibri 16"),
 textvariable = three_point_endpoint_formula_h_value_entry_variable)
three_point_endpoint_formula_h_value_entry.pack(side = "left")

#Submit

three_point_endpoint_formula_submit_button = \
 tk.Button(three_point_endpoint_formula, text = "Submit",
 command = three_point_endpoint_formula_submit)
three_point_endpoint_formula_submit_button.pack()

#output

three_point_endpoint_formula_output_string = tk.StringVar()
three_point_endpoint_formula_output_label = \
 tk.Label(three_point_endpoint_formula,
 textvariable = three_point_endpoint_formula_output_string,
 font = "Calibri 16")
three_point_endpoint_formula_output_label.pack()

#Five-point midpoint formula

five_point_midpoint_formula = \
 tk.Frame(numerical_differentiator_notebook)
five_point_midpoint_formula_label = \
 tk.Label(five_point_midpoint_formula,
 text = "Five-Point Midpoint Formula", font = "Calibri 20 bold")
five_point_midpoint_formula_label.pack()

#Function

five_point_midpoint_formula_frame1 = \
 tk.Frame(five_point_midpoint_formula)
five_point_midpoint_formula_frame1.pack()
five_point_midpoint_formula_function_label = \
 tk.Label(five_point_midpoint_formula_frame1, text = "f(x) = ",
 font = "Calibri 16")
five_point_midpoint_formula_function_label.pack(side = "left")
five_point_midpoint_formula_function_entry_variable = tk.StringVar()
five_point_midpoint_formula_function_entry = \
 tk.Entry(five_point_midpoint_formula_frame1, font = ("Calibri 16"),
 textvariable = five_point_midpoint_formula_function_entry_variable)
five_point_midpoint_formula_function_entry.pack(side = "left")

#x Value

five_point_midpoint_formula_frame2 = \
 tk.Frame(five_point_midpoint_formula)
five_point_midpoint_formula_frame2.pack()
five_point_midpoint_formula_x_value_label = \
 tk.Label(five_point_midpoint_formula_frame2, text = "x = ",
 font = "Calibri 16")
five_point_midpoint_formula_x_value_label.pack(side = "left")
five_point_midpoint_formula_x_value_entry_variable = tk.DoubleVar()
five_point_midpoint_formula_x_value_entry = \
 tk.Entry(five_point_midpoint_formula_frame2, font = ("Calibri 16"),
 textvariable = five_point_midpoint_formula_x_value_entry_variable)
five_point_midpoint_formula_x_value_entry.pack(side = "left")

#h Value

five_point_midpoint_formula_frame3 = \
 tk.Frame(five_point_midpoint_formula)
five_point_midpoint_formula_frame3.pack()
five_point_midpoint_formula_h_value_label = \
 tk.Label(five_point_midpoint_formula_frame3, text = "h = ",
 font = "Calibri 16")
five_point_midpoint_formula_h_value_label.pack(side = "left")
five_point_midpoint_formula_h_value_entry_variable = tk.DoubleVar()
five_point_midpoint_formula_h_value_entry = \
 tk.Entry(five_point_midpoint_formula_frame3, font = ("Calibri 16"),
 textvariable = five_point_midpoint_formula_h_value_entry_variable)
five_point_midpoint_formula_h_value_entry.pack(side = "left")

#Submit

five_point_midpoint_formula_submit_button = \
 tk.Button(five_point_midpoint_formula, text = "Submit",
 command = five_point_midpoint_formula_submit)
five_point_midpoint_formula_submit_button.pack()

#output

five_point_midpoint_formula_output_string = tk.StringVar()
five_point_midpoint_formula_output_label = \
 tk.Label(five_point_midpoint_formula,
 textvariable = five_point_midpoint_formula_output_string,
 font = "Calibri 16")
five_point_midpoint_formula_output_label.pack()

#Five-point endpoint formula

five_point_endpoint_formula = tk.Frame(numerical_differentiator_notebook)
five_point_endpoint_formula_label = \
 tk.Label(five_point_endpoint_formula,
 text = "Five-Point Endpoint Formula", font = "Calibri 20 bold")
five_point_endpoint_formula_label.pack()

#Function

five_point_endpoint_formula_frame1 = \
 tk.Frame(five_point_endpoint_formula)
five_point_endpoint_formula_frame1.pack()
five_point_endpoint_formula_function_label = \
 tk.Label(five_point_endpoint_formula_frame1, text = "f(x) = ",
 font = "Calibri 16")
five_point_endpoint_formula_function_label.pack(side = "left")
five_point_endpoint_formula_function_entry_variable = tk.StringVar()
five_point_endpoint_formula_function_entry = \
 tk.Entry(five_point_endpoint_formula_frame1, font = ("Calibri 16"),
 textvariable = five_point_endpoint_formula_function_entry_variable)
five_point_endpoint_formula_function_entry.pack(side = "left")

#x Value

five_point_endpoint_formula_frame2 = \
 tk.Frame(five_point_endpoint_formula)
five_point_endpoint_formula_frame2.pack()
five_point_endpoint_formula_x_value_label = \
 tk.Label(five_point_endpoint_formula_frame2, text = "x = ",
 font = "Calibri 16")
five_point_endpoint_formula_x_value_label.pack(side = "left")
five_point_endpoint_formula_x_value_entry_variable = tk.DoubleVar()
five_point_endpoint_formula_x_value_entry = \
 tk.Entry(five_point_endpoint_formula_frame2, font = ("Calibri 16"),
 textvariable = five_point_endpoint_formula_x_value_entry_variable)
five_point_endpoint_formula_x_value_entry.pack(side = "left")

#h Value

five_point_endpoint_formula_frame3 = \
 tk.Frame(five_point_endpoint_formula)
five_point_endpoint_formula_frame3.pack()
five_point_endpoint_formula_h_value_label = \
 tk.Label(five_point_endpoint_formula_frame3, text = "h = ",
 font = "Calibri 16")
five_point_endpoint_formula_h_value_label.pack(side = "left")
five_point_endpoint_formula_h_value_entry_variable = tk.DoubleVar()
five_point_endpoint_formula_h_value_entry = \
 tk.Entry(five_point_endpoint_formula_frame3, font = ("Calibri 16"),
 textvariable = five_point_endpoint_formula_h_value_entry_variable)
five_point_endpoint_formula_h_value_entry.pack(side = "left")

#Submit

five_point_endpoint_formula_submit_button = \
 tk.Button(five_point_endpoint_formula, text = "Submit",
 command = five_point_endpoint_formula_submit)
five_point_endpoint_formula_submit_button.pack()

#output

five_point_endpoint_formula_output_string = tk.StringVar()
five_point_endpoint_formula_output_label = \
 tk.Label(five_point_endpoint_formula,
 textvariable = five_point_endpoint_formula_output_string,
 font = "Calibri 16")
five_point_endpoint_formula_output_label.pack()

#Create numerical differentiator notebook tabs

numerical_differentiator_notebook.add(forward_difference,
 text = "Forward Difference")
numerical_differentiator_notebook.add(backward_difference,
 text = "Backward Difference")
numerical_differentiator_notebook.add(central_difference,
 text = "Central Difference")
numerical_differentiator_notebook.add(three_point_midpoint_formula,
 text = "Three-Point Midpoint Formula")
numerical_differentiator_notebook.add(three_point_endpoint_formula,
 text = "Three-Point Endpoint Formula")
numerical_differentiator_notebook.add(five_point_midpoint_formula,
 text = "Five-Point Midpoint Formula")
numerical_differentiator_notebook.add(five_point_endpoint_formula,
 text = "Five-Point Endpoint Formula")
numerical_differentiator_notebook.pack()

#Numerical Integrator

numerical_integrator = tk.Frame(notebook_main)

#Create numerical integrator notebook

numerical_integrator_notebook = ttk.Notebook(numerical_integrator)

#Midpoint rule

midpoint_rule = tk.Frame(numerical_integrator_notebook)
midpoint_rule_label = tk.Label(midpoint_rule, text = "Midpoint Rule",
 font = "Calibri 20 bold")
midpoint_rule_label.pack()

#Function

midpoint_rule_frame1 = tk.Frame(midpoint_rule)
midpoint_rule_frame1.pack()
midpoint_rule_function_label = tk.Label(midpoint_rule_frame1,
 text = "f(x) = ", font = "Calibri 16")
midpoint_rule_function_label.pack(side = "left")
midpoint_rule_function_entry_variable = tk.StringVar()
midpoint_rule_function_entry = tk.Entry(midpoint_rule_frame1,
 font = ("Calibri 16"),
 textvariable = midpoint_rule_function_entry_variable)
midpoint_rule_function_entry.pack(side = "left")

#Interval

midpoint_rule_frame2 = tk.Frame(midpoint_rule)
midpoint_rule_frame2.pack()
midpoint_rule_interval_label1 = tk.Label(midpoint_rule_frame2,
 text = "[a,b] = [", font = "Calibri 16")
midpoint_rule_interval_label1.pack(side = "left")
midpoint_rule_interval_entry_variable1 = tk.DoubleVar()
midpoint_rule_interval_entry1 = tk.Entry(midpoint_rule_frame2,
 font = ("Calibri 16"),
 textvariable = midpoint_rule_interval_entry_variable1)
midpoint_rule_interval_entry1.pack(side = "left")
midpoint_rule_interval_label2 = tk.Label(midpoint_rule_frame2,
 text = ",", font = "Calibri 16")
midpoint_rule_interval_label2.pack(side = "left")
midpoint_rule_interval_entry_variable2 = tk.DoubleVar()
midpoint_rule_interval_entry2 = tk.Entry(midpoint_rule_frame2,
 font = ("Calibri 16"),
 textvariable = midpoint_rule_interval_entry_variable2)
midpoint_rule_interval_entry2.pack(side = "left")
midpoint_rule_interval_label3 = tk.Label(midpoint_rule_frame2,
 text = "]", font = "Calibri 16")
midpoint_rule_interval_label3.pack(side = "left")

#Number of subintervals

midpoint_rule_frame3 = tk.Frame(midpoint_rule)
midpoint_rule_frame3.pack()
midpoint_rule_n_label = tk.Label(midpoint_rule_frame3, text = "n = ",
 font = "Calibri 16")
midpoint_rule_n_label.pack(side = "left")
midpoint_rule_n_entry_variable = tk.IntVar()
midpoint_rule_n_entry = tk.Entry(midpoint_rule_frame3,
 font = ("Calibri 16"), textvariable = midpoint_rule_n_entry_variable)
midpoint_rule_n_entry.pack(side = "left")

#Submit

midpoint_rule_submit_button = tk.Button(midpoint_rule, text = "Submit",
 command = midpoint_rule_submit)
midpoint_rule_submit_button.pack()

#output

midpoint_rule_output_string = tk.StringVar()
midpoint_rule_output_label = tk.Label(midpoint_rule,
 textvariable = midpoint_rule_output_string, font = "Calibri 16")
midpoint_rule_output_label.pack()

#Trapezoid rule

trapezoid_rule = tk.Frame(numerical_integrator_notebook)
trapezoid_rule_label = tk.Label(trapezoid_rule,
 text = "Trapezoid Rule", font = "Calibri 20 bold")
trapezoid_rule_label.pack()

#Function

trapezoid_rule_frame1 = tk.Frame(trapezoid_rule)
trapezoid_rule_frame1.pack()
trapezoid_rule_function_label = tk.Label(trapezoid_rule_frame1,
 text = "f(x) = ", font = "Calibri 16")
trapezoid_rule_function_label.pack(side = "left")
trapezoid_rule_function_entry_variable = tk.StringVar()
trapezoid_rule_function_entry = tk.Entry(trapezoid_rule_frame1,
 font = ("Calibri 16"),
 textvariable = trapezoid_rule_function_entry_variable)
trapezoid_rule_function_entry.pack(side = "left")

#Interval

trapezoid_rule_frame2 = tk.Frame(trapezoid_rule)
trapezoid_rule_frame2.pack()
trapezoid_rule_interval_label1 = tk.Label(trapezoid_rule_frame2,
 text = "[a,b] = [", font = "Calibri 16")
trapezoid_rule_interval_label1.pack(side = "left")
trapezoid_rule_interval_entry_variable1 = tk.DoubleVar()
trapezoid_rule_interval_entry1 = tk.Entry(trapezoid_rule_frame2,
 font = ("Calibri 16"),
 textvariable = trapezoid_rule_interval_entry_variable1)
trapezoid_rule_interval_entry1.pack(side = "left")
trapezoid_rule_interval_label2 = tk.Label(trapezoid_rule_frame2,
 text = ",", font = "Calibri 16")
trapezoid_rule_interval_label2.pack(side = "left")
trapezoid_rule_interval_entry_variable2 = tk.DoubleVar()
trapezoid_rule_interval_entry2 = tk.Entry(trapezoid_rule_frame2,
 font = ("Calibri 16"),
 textvariable = trapezoid_rule_interval_entry_variable2)
trapezoid_rule_interval_entry2.pack(side = "left")
trapezoid_rule_interval_label3 = tk.Label(trapezoid_rule_frame2,
 text = "]", font = "Calibri 16")
trapezoid_rule_interval_label3.pack(side = "left")

#Number of subintervals

trapezoid_rule_frame3 = tk.Frame(trapezoid_rule)
trapezoid_rule_frame3.pack()
trapezoid_rule_n_label = tk.Label(trapezoid_rule_frame3, text = "n = ",
 font = "Calibri 16")
trapezoid_rule_n_label.pack(side = "left")
trapezoid_rule_n_entry_variable = tk.IntVar()
trapezoid_rule_n_entry = tk.Entry(trapezoid_rule_frame3,
 font = ("Calibri 16"), textvariable = trapezoid_rule_n_entry_variable)
trapezoid_rule_n_entry.pack(side = "left")

#Submit

trapezoid_rule_submit_button = tk.Button(trapezoid_rule,
 text = "Submit", command = trapezoid_rule_submit)
trapezoid_rule_submit_button.pack()

#output

trapezoid_rule_output_string = tk.StringVar()
trapezoid_rule_output_label = tk.Label(trapezoid_rule,
 textvariable = trapezoid_rule_output_string, font = "Calibri 16")
trapezoid_rule_output_label.pack()

#Simpson's rule

simpsons_rule = tk.Frame(numerical_integrator_notebook)
simpsons_rule_label = tk.Label(simpsons_rule, text = "Simpson's Rule",
 font = "Calibri 20 bold")
simpsons_rule_label.pack()

#Function

simpsons_rule_frame1 = tk.Frame(simpsons_rule)
simpsons_rule_frame1.pack()
simpsons_rule_function_label = tk.Label(simpsons_rule_frame1,
 text = "f(x) = ", font = "Calibri 16")
simpsons_rule_function_label.pack(side = "left")
simpsons_rule_function_entry_variable = tk.StringVar()
simpsons_rule_function_entry = tk.Entry(simpsons_rule_frame1,
 font = ("Calibri 16"),
 textvariable = simpsons_rule_function_entry_variable)
simpsons_rule_function_entry.pack(side = "left")

#Interval

simpsons_rule_frame2 = tk.Frame(simpsons_rule)
simpsons_rule_frame2.pack()
simpsons_rule_interval_label1 = tk.Label(simpsons_rule_frame2,
 text = "[a,b] = [", font = "Calibri 16")
simpsons_rule_interval_label1.pack(side = "left")
simpsons_rule_interval_entry_variable1 = tk.DoubleVar()
simpsons_rule_interval_entry1 = tk.Entry(simpsons_rule_frame2,
 font = ("Calibri 16"),
 textvariable = simpsons_rule_interval_entry_variable1)
simpsons_rule_interval_entry1.pack(side = "left")
simpsons_rule_interval_label2 = tk.Label(simpsons_rule_frame2,
 text = ",", font = "Calibri 16")
simpsons_rule_interval_label2.pack(side = "left")
simpsons_rule_interval_entry_variable2 = tk.DoubleVar()
simpsons_rule_interval_entry2 = tk.Entry(simpsons_rule_frame2,
 font = ("Calibri 16"),
 textvariable = simpsons_rule_interval_entry_variable2)
simpsons_rule_interval_entry2.pack(side = "left")
simpsons_rule_interval_label3 = tk.Label(simpsons_rule_frame2,
 text = "]", font = "Calibri 16")
simpsons_rule_interval_label3.pack(side = "left")

#Number of subintervals

simpsons_rule_frame3 = tk.Frame(simpsons_rule)
simpsons_rule_frame3.pack()
simpsons_rule_n_label = tk.Label(simpsons_rule_frame3, text = "n = ",
 font = "Calibri 16")
simpsons_rule_n_label.pack(side = "left")
simpsons_rule_n_entry_variable = tk.IntVar()
simpsons_rule_n_entry = tk.Entry(simpsons_rule_frame3,
 font = ("Calibri 16"), textvariable = simpsons_rule_n_entry_variable)
simpsons_rule_n_entry.pack(side = "left")

#Submit

simpsons_rule_submit_button = tk.Button(simpsons_rule, text = "Submit",
 command = simpsons_rule_submit)
simpsons_rule_submit_button.pack()

#output

simpsons_rule_output_string = tk.StringVar()
simpsons_rule_output_label = tk.Label(simpsons_rule,
 textvariable = simpsons_rule_output_string, font = "Calibri 16")
simpsons_rule_output_label.pack()

#Romberg Integration

romberg_integration = tk.Frame(numerical_integrator_notebook)
romberg_integration_label = tk.Label(romberg_integration,
 text = "Romberg Integration", font = "Calibri 20 bold")
romberg_integration_label.pack()

#Function

romberg_integration_frame1 = tk.Frame(romberg_integration)
romberg_integration_frame1.pack()
romberg_integration_function_label = \
 tk.Label(romberg_integration_frame1, text = "f(x) = ",
 font = "Calibri 16")
romberg_integration_function_label.pack(side = "left")
romberg_integration_function_entry_variable = tk.StringVar()
romberg_integration_function_entry = \
 tk.Entry(romberg_integration_frame1, font = ("Calibri 16"),
 textvariable = romberg_integration_function_entry_variable)
romberg_integration_function_entry.pack(side = "left")

#Interval

romberg_integration_frame2 = tk.Frame(romberg_integration)
romberg_integration_frame2.pack()
romberg_integration_interval_label1 = \
 tk.Label(romberg_integration_frame2, text = "[a,b] = [",
 font = "Calibri 16")
romberg_integration_interval_label1.pack(side = "left")
romberg_integration_interval_entry_variable1 = tk.DoubleVar()
romberg_integration_interval_entry1 = \
 tk.Entry(romberg_integration_frame2, font = ("Calibri 16"),
 textvariable = romberg_integration_interval_entry_variable1)
romberg_integration_interval_entry1.pack(side = "left")
romberg_integration_interval_label2 = \
 tk.Label(romberg_integration_frame2, text = ",", font = "Calibri 16")
romberg_integration_interval_label2.pack(side = "left")
romberg_integration_interval_entry_variable2 = tk.DoubleVar()
romberg_integration_interval_entry2 = \
 tk.Entry(romberg_integration_frame2, font = ("Calibri 16"),
 textvariable = romberg_integration_interval_entry_variable2)
romberg_integration_interval_entry2.pack(side = "left")
romberg_integration_interval_label3 = \
 tk.Label(romberg_integration_frame2, text = "]", font = "Calibri 16")
romberg_integration_interval_label3.pack(side = "left")

#Tolerance

romberg_integration_frame3 = tk.Frame(romberg_integration)
romberg_integration_frame3.pack()
romberg_integration_tolerance_label = \
 tk.Label(romberg_integration_frame3, text = "Tolerance: ",
 font = "Calibri 16")
romberg_integration_tolerance_label.pack(side = "left")
romberg_integration_tolerance_variable = tk.DoubleVar()
romberg_integration_tolerance_entry = \
 tk.Entry(romberg_integration_frame3, font = ("Calibri 16"),
 textvariable = romberg_integration_tolerance_variable)
romberg_integration_tolerance_entry.pack(side = "left")

#Submit

romberg_integration_submit_button = tk.Button(romberg_integration,
 text = "Submit", command = romberg_integration_submit)
romberg_integration_submit_button.pack()

#output

romberg_integration_output_string = tk.StringVar()
romberg_integration_output_label = tk.Label(romberg_integration,
 textvariable = romberg_integration_output_string, font = "Calibri 16")
romberg_integration_output_label.pack()

#Adaptive quadrature

adaptive_quadrature = tk.Frame(numerical_integrator_notebook)
adaptive_quadrature_label = tk.Label(adaptive_quadrature,
 text = "Adaptive Quadrature", font = "Calibri 20 bold")
adaptive_quadrature_label.pack()

#Function

adaptive_quadrature_frame1 = tk.Frame(adaptive_quadrature)
adaptive_quadrature_frame1.pack()
adaptive_quadrature_function_label = \
 tk.Label(adaptive_quadrature_frame1, text = "f(x) = ",
 font = "Calibri 16")
adaptive_quadrature_function_label.pack(side = "left")
adaptive_quadrature_function_entry_variable = tk.StringVar()
adaptive_quadrature_function_entry = \
 tk.Entry(adaptive_quadrature_frame1, font = ("Calibri 16"),
 textvariable = adaptive_quadrature_function_entry_variable)
adaptive_quadrature_function_entry.pack(side = "left")

#Interval

adaptive_quadrature_frame2 = tk.Frame(adaptive_quadrature)
adaptive_quadrature_frame2.pack()
adaptive_quadrature_interval_label1 = \
 tk.Label(adaptive_quadrature_frame2, text = "[a,b] = [",
 font = "Calibri 16")
adaptive_quadrature_interval_label1.pack(side = "left")
adaptive_quadrature_interval_entry_variable1 = tk.DoubleVar()
adaptive_quadrature_interval_entry1 = \
 tk.Entry(adaptive_quadrature_frame2, font = ("Calibri 16"),
 textvariable = adaptive_quadrature_interval_entry_variable1)
adaptive_quadrature_interval_entry1.pack(side = "left")
adaptive_quadrature_interval_label2 = \
 tk.Label(adaptive_quadrature_frame2, text = ",", font = "Calibri 16")
adaptive_quadrature_interval_label2.pack(side = "left")
adaptive_quadrature_interval_entry_variable2 = tk.DoubleVar()
adaptive_quadrature_interval_entry2 = \
 tk.Entry(adaptive_quadrature_frame2, font = ("Calibri 16"),
 textvariable = adaptive_quadrature_interval_entry_variable2)
adaptive_quadrature_interval_entry2.pack(side = "left")
adaptive_quadrature_interval_label3 = \
 tk.Label(adaptive_quadrature_frame2, text = "]", font = "Calibri 16")
adaptive_quadrature_interval_label3.pack(side = "left")

#Tolerance

adaptive_quadrature_frame3 = tk.Frame(adaptive_quadrature)
adaptive_quadrature_frame3.pack()
adaptive_quadrature_tolerance_label = \
 tk.Label(adaptive_quadrature_frame3, text = "Tolerance: ",
 font = "Calibri 16")
adaptive_quadrature_tolerance_label.pack(side = "left")
adaptive_quadrature_tolerance_variable = tk.DoubleVar()
adaptive_quadrature_tolerance_entry = \
 tk.Entry(adaptive_quadrature_frame3, font = ("Calibri 16"),
 textvariable = adaptive_quadrature_tolerance_variable)
adaptive_quadrature_tolerance_entry.pack(side = "left")

#Submit

adaptive_quadrature_submit_button = tk.Button(adaptive_quadrature,
 text = "Submit", command = adaptive_quadrature_submit)
adaptive_quadrature_submit_button.pack()

#output

adaptive_quadrature_output_string = tk.StringVar()
adaptive_quadrature_output_label = tk.Label(adaptive_quadrature,
 textvariable = adaptive_quadrature_output_string, font = "Calibri 16")
adaptive_quadrature_output_label.pack()

#Gaussian quadrature

gaussian_quadrature = tk.Frame(numerical_integrator_notebook)
gaussian_quadrature_label = tk.Label(gaussian_quadrature,
 text = "Gaussian Quadrature", font = "Calibri 20 bold")
gaussian_quadrature_label.pack()

#Function

gaussian_quadrature_frame1 = tk.Frame(gaussian_quadrature)
gaussian_quadrature_frame1.pack()
gaussian_quadrature_function_label = \
 tk.Label(gaussian_quadrature_frame1, text = "f(x) = ",
 font = "Calibri 16")
gaussian_quadrature_function_label.pack(side = "left")
gaussian_quadrature_function_entry_variable = tk.StringVar()
gaussian_quadrature_function_entry = \
 tk.Entry(gaussian_quadrature_frame1, font = ("Calibri 16"),
 textvariable = gaussian_quadrature_function_entry_variable)
gaussian_quadrature_function_entry.pack(side = "left")

#Number of weights and points

gaussian_quadrature_frame3 = tk.Frame(gaussian_quadrature)
gaussian_quadrature_frame3.pack()
gaussian_quadrature_n_label = \
 tk.Label(gaussian_quadrature_frame3, text = "n = ",
 font = "Calibri 16")
gaussian_quadrature_n_label.pack(side = "left")
gaussian_quadrature_n_entry_variable = tk.IntVar()
gaussian_quadrature_n_entry = tk.Entry(gaussian_quadrature_frame3,
 font = ("Calibri 16"),
 textvariable = gaussian_quadrature_n_entry_variable)
gaussian_quadrature_n_entry.pack(side = "left")

#Submit

gaussian_quadrature_submit_button = tk.Button(gaussian_quadrature,
 text = "Submit", command = gaussian_quadrature_submit)
gaussian_quadrature_submit_button.pack()

#output

gaussian_quadrature_output_string = tk.StringVar()
gaussian_quadrature_output_label = tk.Label(gaussian_quadrature,
 textvariable = gaussian_quadrature_output_string, font = "Calibri 16")
gaussian_quadrature_output_label.pack()

#Create numerical integrator notebook tabs

numerical_integrator_notebook.add(midpoint_rule,
 text = "Midpoint Rule")
numerical_integrator_notebook.add(trapezoid_rule,
 text = "Trapezoid Rule")
numerical_integrator_notebook.add(simpsons_rule,
 text = "Simpson's Rule")
numerical_integrator_notebook.add(romberg_integration,
 text = "Romberg Integration")
numerical_integrator_notebook.add(adaptive_quadrature,
 text = "Adaptive Quadrature")
numerical_integrator_notebook.add(gaussian_quadrature,
 text = "Gaussian Quadrature")
numerical_integrator_notebook.pack()

#Numerical Solver

numerical_solver = tk.Frame(notebook_main)

#Create numerical solver notebook

numerical_solver_notebook = ttk.Notebook(numerical_solver)

#Euler's Method

eulers_method = tk.Frame(numerical_solver_notebook)
eulers_method_label = tk.Label(eulers_method, text = "Euler's Method",
 font = "Calibri 20 bold")
eulers_method_label.pack()

#Function

eulers_method_frame1 = tk.Frame(eulers_method)
eulers_method_frame1.pack()
eulers_method_function_label = tk.Label(eulers_method_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
eulers_method_function_label.pack(side = "left")
eulers_method_function_entry_variable = tk.StringVar()
eulers_method_function_entry = tk.Entry(eulers_method_frame1,
 font = ("Calibri 16"),
 textvariable = eulers_method_function_entry_variable)
eulers_method_function_entry.pack(side = "left")

#Interval

eulers_method_frame2 = tk.Frame(eulers_method)
eulers_method_frame2.pack()
eulers_method_interval_label1 = tk.Label(eulers_method_frame2,
 text = "[a,b] = [", font = "Calibri 16")
eulers_method_interval_label1.pack(side = "left")
eulers_method_interval_entry_variable1 = tk.DoubleVar()
eulers_method_interval_entry1 = tk.Entry(eulers_method_frame2,
 font = ("Calibri 16"),
 textvariable = eulers_method_interval_entry_variable1)
eulers_method_interval_entry1.pack(side = "left")
eulers_method_interval_label2 = tk.Label(eulers_method_frame2,
 text = ",", font = "Calibri 16")
eulers_method_interval_label2.pack(side = "left")
eulers_method_interval_entry_variable2 = tk.DoubleVar()
eulers_method_interval_entry2 = tk.Entry(eulers_method_frame2,
 font = ("Calibri 16"),
 textvariable = eulers_method_interval_entry_variable2)
eulers_method_interval_entry2.pack(side = "left")
eulers_method_interval_label3 = tk.Label(eulers_method_frame2,
 text = "]", font = "Calibri 16")
eulers_method_interval_label3.pack(side = "left")

#Initial Condition

eulers_method_frame3 = tk.Frame(eulers_method)
eulers_method_frame3.pack()
eulers_method_initial_condition_label = \
 tk.Label(eulers_method_frame3, text = "y(a) = ", font = "Calibri 16")
eulers_method_initial_condition_label.pack(side = "left")
eulers_method_initial_condition_entry_variable = tk.DoubleVar()
eulers_method_initial_condition_entry = \
 tk.Entry(eulers_method_frame3, font = ("Calibri 16"),
 textvariable = eulers_method_initial_condition_entry_variable)
eulers_method_initial_condition_entry.pack(side = "left")

#Number of steps

eulers_method_frame4 = tk.Frame(eulers_method)
eulers_method_frame4.pack()
eulers_method_n_label = tk.Label(eulers_method_frame4, text = "N = ",
 font = "Calibri 16")
eulers_method_n_label.pack(side = "left")
eulers_method_n_entry_variable = tk.IntVar()
eulers_method_n_entry = tk.Entry(eulers_method_frame4,
 font = ("Calibri 16"), textvariable = eulers_method_n_entry_variable)
eulers_method_n_entry.pack(side = "left")

#Submit

eulers_method_submit_button = tk.Button(eulers_method, text = "Submit",
 command = eulers_method_submit)
eulers_method_submit_button.pack()

#output

eulers_method_output_string = tk.StringVar()
eulers_method_output_label = tk.Label(eulers_method,
 textvariable = eulers_method_output_string, font = "Calibri 16")
eulers_method_output_label.pack()

#Midpoint Method

midpoint_method = tk.Frame(numerical_solver_notebook)
midpoint_method_label = tk.Label(midpoint_method,
 text = "Midpoint Method", font = "Calibri 20 bold")
midpoint_method_label.pack()

#Function

midpoint_method_frame1 = tk.Frame(midpoint_method)
midpoint_method_frame1.pack()
midpoint_method_function_label = tk.Label(midpoint_method_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
midpoint_method_function_label.pack(side = "left")
midpoint_method_function_entry_variable = tk.StringVar()
midpoint_method_function_entry = tk.Entry(midpoint_method_frame1,
 font = ("Calibri 16"),
 textvariable = midpoint_method_function_entry_variable)
midpoint_method_function_entry.pack(side = "left")

#Interval

midpoint_method_frame2 = tk.Frame(midpoint_method)
midpoint_method_frame2.pack()
midpoint_method_interval_label1 = tk.Label(midpoint_method_frame2,
 text = "[a,b] = [", font = "Calibri 16")
midpoint_method_interval_label1.pack(side = "left")
midpoint_method_interval_entry_variable1 = tk.DoubleVar()
midpoint_method_interval_entry1 = tk.Entry(midpoint_method_frame2,
 font = ("Calibri 16"),
 textvariable = midpoint_method_interval_entry_variable1)
midpoint_method_interval_entry1.pack(side = "left")
midpoint_method_interval_label2 = tk.Label(midpoint_method_frame2,
 text = ",", font = "Calibri 16")
midpoint_method_interval_label2.pack(side = "left")
midpoint_method_interval_entry_variable2 = tk.DoubleVar()
midpoint_method_interval_entry2 = tk.Entry(midpoint_method_frame2,
 font = ("Calibri 16"),
 textvariable = midpoint_method_interval_entry_variable2)
midpoint_method_interval_entry2.pack(side = "left")
midpoint_method_interval_label3 = tk.Label(midpoint_method_frame2,
 text = "]", font = "Calibri 16")
midpoint_method_interval_label3.pack(side = "left")

#Initial Condition

midpoint_method_frame3 = tk.Frame(midpoint_method)
midpoint_method_frame3.pack()
midpoint_method_initial_condition_label = \
 tk.Label(midpoint_method_frame3, text = "y(a) = ",
 font = "Calibri 16")
midpoint_method_initial_condition_label.pack(side = "left")
midpoint_method_initial_condition_entry_variable = tk.DoubleVar()
midpoint_method_initial_condition_entry = \
 tk.Entry(midpoint_method_frame3, font = ("Calibri 16"),
 textvariable = midpoint_method_initial_condition_entry_variable)
midpoint_method_initial_condition_entry.pack(side = "left")

#Number of steps

midpoint_method_frame4 = tk.Frame(midpoint_method)
midpoint_method_frame4.pack()
midpoint_method_n_label = tk.Label(midpoint_method_frame4,
 text = "N = ", font = "Calibri 16")
midpoint_method_n_label.pack(side = "left")
midpoint_method_n_entry_variable = tk.IntVar()
midpoint_method_n_entry = tk.Entry(midpoint_method_frame4,
 font = ("Calibri 16"),
 textvariable = midpoint_method_n_entry_variable)
midpoint_method_n_entry.pack(side = "left")

#Submit

midpoint_method_submit_button = tk.Button(midpoint_method,
 text = "Submit", command = midpoint_method_submit)
midpoint_method_submit_button.pack()

#output

midpoint_method_output_string = tk.StringVar()
midpoint_method_output_label = tk.Label(midpoint_method,
 textvariable = midpoint_method_output_string, font = "Calibri 16")
midpoint_method_output_label.pack()

#Modified Euler Method

modified_euler_method = tk.Frame(numerical_solver_notebook)
modified_euler_method_label = tk.Label(modified_euler_method,
 text = "Modified Euler Method", font = "Calibri 20 bold")
modified_euler_method_label.pack()

#Function

modified_euler_method_frame1 = tk.Frame(modified_euler_method)
modified_euler_method_frame1.pack()
modified_euler_method_function_label = \
 tk.Label(modified_euler_method_frame1, text = "y' = f(x, y) = ",
 font = "Calibri 16")
modified_euler_method_function_label.pack(side = "left")
modified_euler_method_function_entry_variable = tk.StringVar()
modified_euler_method_function_entry = \
 tk.Entry(modified_euler_method_frame1, font = ("Calibri 16"),
 textvariable = modified_euler_method_function_entry_variable)
modified_euler_method_function_entry.pack(side = "left")

#Interval

modified_euler_method_frame2 = tk.Frame(modified_euler_method)
modified_euler_method_frame2.pack()
modified_euler_method_interval_label1 = \
 tk.Label(modified_euler_method_frame2, text = "[a,b] = [",
 font = "Calibri 16")
modified_euler_method_interval_label1.pack(side = "left")
modified_euler_method_interval_entry_variable1 = tk.DoubleVar()
modified_euler_method_interval_entry1 = \
 tk.Entry(modified_euler_method_frame2, font = ("Calibri 16"),
 textvariable = modified_euler_method_interval_entry_variable1)
modified_euler_method_interval_entry1.pack(side = "left")
modified_euler_method_interval_label2 = \
 tk.Label(modified_euler_method_frame2, text = ",",
 font = "Calibri 16")
modified_euler_method_interval_label2.pack(side = "left")
modified_euler_method_interval_entry_variable2 = tk.DoubleVar()
modified_euler_method_interval_entry2 = \
 tk.Entry(modified_euler_method_frame2, font = ("Calibri 16"),
 textvariable = modified_euler_method_interval_entry_variable2)
modified_euler_method_interval_entry2.pack(side = "left")
modified_euler_method_interval_label3 = \
 tk.Label(modified_euler_method_frame2, text = "]",
 font = "Calibri 16")
modified_euler_method_interval_label3.pack(side = "left")

#Initial Condition

modified_euler_method_frame3 = tk.Frame(modified_euler_method)
modified_euler_method_frame3.pack()
modified_euler_method_initial_condition_label = \
 tk.Label(modified_euler_method_frame3, text = "y(a) = ",
 font = "Calibri 16")
modified_euler_method_initial_condition_label.pack(side = "left")
modified_euler_method_initial_condition_entry_variable = tk.DoubleVar()
modified_euler_method_initial_condition_entry = \
 tk.Entry(modified_euler_method_frame3, font = ("Calibri 16"),
 textvariable = modified_euler_method_initial_condition_entry_variable)
modified_euler_method_initial_condition_entry.pack(side = "left")

#Number of steps

modified_euler_method_frame4 = tk.Frame(modified_euler_method)
modified_euler_method_frame4.pack()
modified_euler_method_n_label = tk.Label(modified_euler_method_frame4,
 text = "N = ", font = "Calibri 16")
modified_euler_method_n_label.pack(side = "left")
modified_euler_method_n_entry_variable = tk.IntVar()
modified_euler_method_n_entry = tk.Entry(modified_euler_method_frame4,
 font = ("Calibri 16"),
 textvariable = modified_euler_method_n_entry_variable)
modified_euler_method_n_entry.pack(side = "left")

#Submit

modified_euler_method_submit_button = tk.Button(modified_euler_method, text = "Submit",
 command = modified_euler_method_submit)
modified_euler_method_submit_button.pack()

#output

modified_euler_method_output_string = tk.StringVar()
modified_euler_method_output_label = tk.Label(modified_euler_method,
 textvariable = modified_euler_method_output_string,
 font = "Calibri 16")
modified_euler_method_output_label.pack()

#Heun's Method

heuns_method = tk.Frame(numerical_solver_notebook)
heuns_method_label = tk.Label(heuns_method, text = "Heun's Method",
 font = "Calibri 20 bold")
heuns_method_label.pack()

#Function

heuns_method_frame1 = tk.Frame(heuns_method)
heuns_method_frame1.pack()
heuns_method_function_label = tk.Label(heuns_method_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
heuns_method_function_label.pack(side = "left")
heuns_method_function_entry_variable = tk.StringVar()
heuns_method_function_entry = tk.Entry(heuns_method_frame1,
 font = ("Calibri 16"),
 textvariable = heuns_method_function_entry_variable)
heuns_method_function_entry.pack(side = "left")

#Interval

heuns_method_frame2 = tk.Frame(heuns_method)
heuns_method_frame2.pack()
heuns_method_interval_label1 = tk.Label(heuns_method_frame2,
 text = "[a,b] = [", font = "Calibri 16")
heuns_method_interval_label1.pack(side = "left")
heuns_method_interval_entry_variable1 = tk.DoubleVar()
heuns_method_interval_entry1 = tk.Entry(heuns_method_frame2,
 font = ("Calibri 16"),
 textvariable = heuns_method_interval_entry_variable1)
heuns_method_interval_entry1.pack(side = "left")
heuns_method_interval_label2 = tk.Label(heuns_method_frame2,
 text = ",", font = "Calibri 16")
heuns_method_interval_label2.pack(side = "left")
heuns_method_interval_entry_variable2 = tk.DoubleVar()
heuns_method_interval_entry2 = tk.Entry(heuns_method_frame2,
 font = ("Calibri 16"),
 textvariable = heuns_method_interval_entry_variable2)
heuns_method_interval_entry2.pack(side = "left")
heuns_method_interval_label3 = tk.Label(heuns_method_frame2,
 text = "]", font = "Calibri 16")
heuns_method_interval_label3.pack(side = "left")

#Initial Condition

heuns_method_frame3 = tk.Frame(heuns_method)
heuns_method_frame3.pack()
heuns_method_initial_condition_label = tk.Label(heuns_method_frame3,
 text = "y(a) = ", font = "Calibri 16")
heuns_method_initial_condition_label.pack(side = "left")
heuns_method_initial_condition_entry_variable = tk.DoubleVar()
heuns_method_initial_condition_entry = tk.Entry(heuns_method_frame3,
 font = ("Calibri 16"),
 textvariable = heuns_method_initial_condition_entry_variable)
heuns_method_initial_condition_entry.pack(side = "left")

#Number of steps

heuns_method_frame4 = tk.Frame(heuns_method)
heuns_method_frame4.pack()
heuns_method_n_label = tk.Label(heuns_method_frame4, text = "N = ",
 font = "Calibri 16")
heuns_method_n_label.pack(side = "left")
heuns_method_n_entry_variable = tk.IntVar()
heuns_method_n_entry = tk.Entry(heuns_method_frame4,
 font = ("Calibri 16"), textvariable = heuns_method_n_entry_variable)
heuns_method_n_entry.pack(side = "left")

#Submit

heuns_method_submit_button = tk.Button(heuns_method, text = "Submit",
 command = heuns_method_submit)
heuns_method_submit_button.pack()

#output

heuns_method_output_string = tk.StringVar()
heuns_method_output_label = tk.Label(heuns_method,
 textvariable = heuns_method_output_string, font = "Calibri 16")
heuns_method_output_label.pack()

#Runge-Kutta Method

runge_kutta_method = tk.Frame(numerical_solver_notebook)
runge_kutta_method_label = tk.Label(runge_kutta_method,
 text = "Runge-Kutta Method", font = "Calibri 20 bold")
runge_kutta_method_label.pack()

#Function

runge_kutta_method_frame1 = tk.Frame(runge_kutta_method)
runge_kutta_method_frame1.pack()
runge_kutta_method_function_label = tk.Label(runge_kutta_method_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
runge_kutta_method_function_label.pack(side = "left")
runge_kutta_method_function_entry_variable = tk.StringVar()
runge_kutta_method_function_entry = tk.Entry(runge_kutta_method_frame1,
 font = ("Calibri 16"),
 textvariable = runge_kutta_method_function_entry_variable)
runge_kutta_method_function_entry.pack(side = "left")

#Interval

runge_kutta_method_frame2 = tk.Frame(runge_kutta_method)
runge_kutta_method_frame2.pack()
runge_kutta_method_interval_label1 = \
 tk.Label(runge_kutta_method_frame2, text = "[a,b] = [",
 font = "Calibri 16")
runge_kutta_method_interval_label1.pack(side = "left")
runge_kutta_method_interval_entry_variable1 = tk.DoubleVar()
runge_kutta_method_interval_entry1 = \
 tk.Entry(runge_kutta_method_frame2, font = ("Calibri 16"),
 textvariable = runge_kutta_method_interval_entry_variable1)
runge_kutta_method_interval_entry1.pack(side = "left")
runge_kutta_method_interval_label2 = \
 tk.Label(runge_kutta_method_frame2, text = ",", font = "Calibri 16")
runge_kutta_method_interval_label2.pack(side = "left")
runge_kutta_method_interval_entry_variable2 = tk.DoubleVar()
runge_kutta_method_interval_entry2 = \
 tk.Entry(runge_kutta_method_frame2, font = ("Calibri 16"),
 textvariable = runge_kutta_method_interval_entry_variable2)
runge_kutta_method_interval_entry2.pack(side = "left")
runge_kutta_method_interval_label3 = \
 tk.Label(runge_kutta_method_frame2, text = "]", font = "Calibri 16")
runge_kutta_method_interval_label3.pack(side = "left")

#Initial Condition

runge_kutta_method_frame3 = tk.Frame(runge_kutta_method)
runge_kutta_method_frame3.pack()
runge_kutta_method_initial_condition_label = \
 tk.Label(runge_kutta_method_frame3, text = "y(a) = ",
 font = "Calibri 16")
runge_kutta_method_initial_condition_label.pack(side = "left")
runge_kutta_method_initial_condition_entry_variable = tk.DoubleVar()
runge_kutta_method_initial_condition_entry = \
 tk.Entry(runge_kutta_method_frame3, font = ("Calibri 16"),
 textvariable = runge_kutta_method_initial_condition_entry_variable)
runge_kutta_method_initial_condition_entry.pack(side = "left")

#Number of steps

runge_kutta_method_frame4 = tk.Frame(runge_kutta_method)
runge_kutta_method_frame4.pack()
runge_kutta_method_n_label = tk.Label(runge_kutta_method_frame4,
 text = "N = ", font = "Calibri 16")
runge_kutta_method_n_label.pack(side = "left")
runge_kutta_method_n_entry_variable = tk.IntVar()
runge_kutta_method_n_entry = tk.Entry(runge_kutta_method_frame4,
 font = ("Calibri 16"),
 textvariable = runge_kutta_method_n_entry_variable)
runge_kutta_method_n_entry.pack(side = "left")

#Submit

runge_kutta_method_submit_button = tk.Button(runge_kutta_method,
 text = "Submit", command = runge_kutta_method_submit)
runge_kutta_method_submit_button.pack()

#output

runge_kutta_method_output_string = tk.StringVar()
runge_kutta_method_output_label = tk.Label(runge_kutta_method,
 textvariable = runge_kutta_method_output_string, font = "Calibri 16")
runge_kutta_method_output_label.pack()

#Runge-Kutta-Fehlberg Method

runge_kutta_fehlberg_method = tk.Frame(numerical_solver_notebook)
runge_kutta_fehlberg_method_label = \
 tk.Label(runge_kutta_fehlberg_method,
 text = "Runge-Kutta-Fehlberg Method", font = "Calibri 20 bold")
runge_kutta_fehlberg_method_label.pack()

#Function

runge_kutta_fehlberg_method_frame1 = \
 tk.Frame(runge_kutta_fehlberg_method)
runge_kutta_fehlberg_method_frame1.pack()
runge_kutta_fehlberg_method_function_label = \
 tk.Label(runge_kutta_fehlberg_method_frame1, text = "y' = f(x, y) = ",
 font = "Calibri 16")
runge_kutta_fehlberg_method_function_label.pack(side = "left")
runge_kutta_fehlberg_method_function_entry_variable = tk.StringVar()
runge_kutta_fehlberg_method_function_entry = \
 tk.Entry(runge_kutta_fehlberg_method_frame1, font = ("Calibri 16"),
 textvariable = runge_kutta_fehlberg_method_function_entry_variable)
runge_kutta_fehlberg_method_function_entry.pack(side = "left")

#Interval

runge_kutta_fehlberg_method_frame2 = \
 tk.Frame(runge_kutta_fehlberg_method)
runge_kutta_fehlberg_method_frame2.pack()
runge_kutta_fehlberg_method_interval_label1 = \
 tk.Label(runge_kutta_fehlberg_method_frame2, text = "[a,b] = [",
 font = "Calibri 16")
runge_kutta_fehlberg_method_interval_label1.pack(side = "left")
runge_kutta_fehlberg_method_interval_entry_variable1 = tk.DoubleVar()
runge_kutta_fehlberg_method_interval_entry1 = \
 tk.Entry(runge_kutta_fehlberg_method_frame2, font = ("Calibri 16"),
 textvariable = runge_kutta_fehlberg_method_interval_entry_variable1)
runge_kutta_fehlberg_method_interval_entry1.pack(side = "left")
runge_kutta_fehlberg_method_interval_label2 = \
 tk.Label(runge_kutta_fehlberg_method_frame2, text = ",",
 font = "Calibri 16")
runge_kutta_fehlberg_method_interval_label2.pack(side = "left")
runge_kutta_fehlberg_method_interval_entry_variable2 = tk.DoubleVar()
runge_kutta_fehlberg_method_interval_entry2 = \
 tk.Entry(runge_kutta_fehlberg_method_frame2, font = ("Calibri 16"),
 textvariable = runge_kutta_fehlberg_method_interval_entry_variable2)
runge_kutta_fehlberg_method_interval_entry2.pack(side = "left")
runge_kutta_fehlberg_method_interval_label3 = \
 tk.Label(runge_kutta_fehlberg_method_frame2, text = "]",
 font = "Calibri 16")
runge_kutta_fehlberg_method_interval_label3.pack(side = "left")

#Step Sizes

runge_kutta_fehlberg_method_frame3 = \
 tk.Frame(runge_kutta_fehlberg_method)
runge_kutta_fehlberg_method_frame3.pack()
subscripts = str.maketrans("aimnx", "")
label1 = "hmin = "
label1 = label1.translate(subscripts)
runge_kutta_fehlberg_method_h_value_label1 = \
 tk.Label(runge_kutta_fehlberg_method_frame3, text = label1,
 font = "Calibri 16")
runge_kutta_fehlberg_method_h_value_label1.pack(side = "left")
runge_kutta_fehlberg_method_h_value_entry_variable1 = \
 tk.DoubleVar()
runge_kutta_fehlberg_method_h_value_entry1 = \
 tk.Entry(runge_kutta_fehlberg_method_frame3, font = ("Calibri 16"),
 textvariable = \
 runge_kutta_fehlberg_method_h_value_entry_variable1)
runge_kutta_fehlberg_method_h_value_entry1.pack(side = "left")
label2 = ", hmax = "
label2 = label2.translate(subscripts)
runge_kutta_fehlberg_method_h_value_label2 = \
 tk.Label(runge_kutta_fehlberg_method_frame3, text = label2,
 font = "Calibri 16")
runge_kutta_fehlberg_method_h_value_label2.pack(side = "left")
runge_kutta_fehlberg_method_h_value_entry_variable2 = \
 tk.DoubleVar()
runge_kutta_fehlberg_method_h_value_entry2 = \
 tk.Entry(runge_kutta_fehlberg_method_frame3, font = ("Calibri 16"),
 textvariable = \
 runge_kutta_fehlberg_method_h_value_entry_variable2)
runge_kutta_fehlberg_method_h_value_entry2.pack(side = "left")

#Initial Condition

runge_kutta_fehlberg_method_frame4 = \
 tk.Frame(runge_kutta_fehlberg_method)
runge_kutta_fehlberg_method_frame4.pack()
runge_kutta_fehlberg_method_initial_condition_label = \
 tk.Label(runge_kutta_fehlberg_method_frame4, text = "y(a) = ",
 font = "Calibri 16")
runge_kutta_fehlberg_method_initial_condition_label.pack(side = "left")
runge_kutta_fehlberg_method_initial_condition_entry_variable = \
 tk.DoubleVar()
runge_kutta_fehlberg_method_initial_condition_entry = \
 tk.Entry(runge_kutta_fehlberg_method_frame4, font = ("Calibri 16"),
 textvariable = \
 runge_kutta_fehlberg_method_initial_condition_entry_variable)
runge_kutta_fehlberg_method_initial_condition_entry.pack(side = "left")

#Tolerance

runge_kutta_fehlberg_method_frame5 = \
 tk.Frame(runge_kutta_fehlberg_method)
runge_kutta_fehlberg_method_frame5.pack()
runge_kutta_fehlberg_method_tolerance_label = \
 tk.Label(runge_kutta_fehlberg_method_frame5, text = "Tolerance: ",
 font = "Calibri 16")
runge_kutta_fehlberg_method_tolerance_label.pack(side = "left")
runge_kutta_fehlberg_method_tolerance_variable = tk.DoubleVar()
runge_kutta_fehlberg_method_tolerance_entry = \
 tk.Entry(runge_kutta_fehlberg_method_frame5, font = ("Calibri 16"),
 textvariable = runge_kutta_fehlberg_method_tolerance_variable)
runge_kutta_fehlberg_method_tolerance_entry.pack(side = "left")

#Submit

runge_kutta_fehlberg_method_submit_button = \
 tk.Button(runge_kutta_fehlberg_method, text = "Submit",
 command = runge_kutta_fehlberg_method_submit)
runge_kutta_fehlberg_method_submit_button.pack()

#output

runge_kutta_fehlberg_method_output_string = tk.StringVar()
runge_kutta_fehlberg_method_output_label = \
 tk.Label(runge_kutta_fehlberg_method,
 textvariable = runge_kutta_fehlberg_method_output_string,
 font = "Calibri 16")
runge_kutta_fehlberg_method_output_label.pack()

#Adams-Bashforth Technique

adams_bashforth_technique = tk.Frame(numerical_solver_notebook)
adams_bashforth_technique_label = tk.Label(adams_bashforth_technique,
 text = "Adams-Bashforth Technique", font = "Calibri 20 bold")
adams_bashforth_technique_label.pack()

#Function

adams_bashforth_technique_frame1 = tk.Frame(adams_bashforth_technique)
adams_bashforth_technique_frame1.pack()
adams_bashforth_technique_function_label = \
 tk.Label(adams_bashforth_technique_frame1, text = "y' = f(x, y) = ",
 font = "Calibri 16")
adams_bashforth_technique_function_label.pack(side = "left")
adams_bashforth_technique_function_entry_variable = tk.StringVar()
adams_bashforth_technique_function_entry = \
 tk.Entry(adams_bashforth_technique_frame1, font = ("Calibri 16"),
 textvariable = adams_bashforth_technique_function_entry_variable)
adams_bashforth_technique_function_entry.pack(side = "left")

#Interval

adams_bashforth_technique_frame2 = tk.Frame(adams_bashforth_technique)
adams_bashforth_technique_frame2.pack()
adams_bashforth_technique_interval_label1 = \
 tk.Label(adams_bashforth_technique_frame2, text = "[a,b] = [",
 font = "Calibri 16")
adams_bashforth_technique_interval_label1.pack(side = "left")
adams_bashforth_technique_interval_entry_variable1 = tk.DoubleVar()
adams_bashforth_technique_interval_entry1 = \
 tk.Entry(adams_bashforth_technique_frame2, font = ("Calibri 16"),
 textvariable = adams_bashforth_technique_interval_entry_variable1)
adams_bashforth_technique_interval_entry1.pack(side = "left")
adams_bashforth_technique_interval_label2 = \
 tk.Label(adams_bashforth_technique_frame2, text = ",",
 font = "Calibri 16")
adams_bashforth_technique_interval_label2.pack(side = "left")
adams_bashforth_technique_interval_entry_variable2 = tk.DoubleVar()
adams_bashforth_technique_interval_entry2 = \
 tk.Entry(adams_bashforth_technique_frame2, font = ("Calibri 16"),
 textvariable = adams_bashforth_technique_interval_entry_variable2)
adams_bashforth_technique_interval_entry2.pack(side = "left")
adams_bashforth_technique_interval_label3 = \
 tk.Label(adams_bashforth_technique_frame2, text = "]",
 font = "Calibri 16")
adams_bashforth_technique_interval_label3.pack(side = "left")

#Initial Condition

adams_bashforth_technique_frame3 = tk.Frame(adams_bashforth_technique)
adams_bashforth_technique_frame3.pack()
adams_bashforth_technique_initial_condition_label = \
 tk.Label(adams_bashforth_technique_frame3, text = "y(a) = ",
 font = "Calibri 16")
adams_bashforth_technique_initial_condition_label.pack(side = "left")
adams_bashforth_technique_initial_condition_entry_variable = \
 tk.DoubleVar()
adams_bashforth_technique_initial_condition_entry = \
 tk.Entry(adams_bashforth_technique_frame3, font = ("Calibri 16"),
 textvariable = \
 adams_bashforth_technique_initial_condition_entry_variable)
adams_bashforth_technique_initial_condition_entry.pack(side = "left")

#Number of steps

adams_bashforth_technique_frame4 = \
 tk.Frame(adams_bashforth_technique)
adams_bashforth_technique_frame4.pack()
adams_bashforth_technique_n_label = \
 tk.Label(adams_bashforth_technique_frame4, text = "N = ",
 font = "Calibri 16")
adams_bashforth_technique_n_label.pack(side = "left")
adams_bashforth_technique_n_entry_variable = tk.IntVar()
adams_bashforth_technique_n_entry = \
 tk.Entry(adams_bashforth_technique_frame4, font = ("Calibri 16"),
 textvariable = adams_bashforth_technique_n_entry_variable)
adams_bashforth_technique_n_entry.pack(side = "left")

#Submit

adams_bashforth_technique_submit_button = \
 tk.Button(adams_bashforth_technique, text = "Submit",
 command = adams_bashforth_technique_submit)
adams_bashforth_technique_submit_button.pack()

#output

adams_bashforth_technique_output_string = tk.StringVar()
adams_bashforth_technique_output_label = \
 tk.Label(adams_bashforth_technique,
 textvariable = adams_bashforth_technique_output_string,
 font = "Calibri 16")
adams_bashforth_technique_output_label.pack()

#Milue's Method

milues_method = tk.Frame(numerical_solver_notebook)
milues_method_label = tk.Label(milues_method, text = "Milue's Method",
 font = "Calibri 20 bold")
milues_method_label.pack()

#Function

milues_method_frame1 = tk.Frame(milues_method)
milues_method_frame1.pack()
milues_method_function_label = tk.Label(milues_method_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
milues_method_function_label.pack(side = "left")
milues_method_function_entry_variable = tk.StringVar()
milues_method_function_entry = tk.Entry(milues_method_frame1,
 font = ("Calibri 16"),
 textvariable = milues_method_function_entry_variable)
milues_method_function_entry.pack(side = "left")

#Interval

milues_method_frame2 = tk.Frame(milues_method)
milues_method_frame2.pack()
milues_method_interval_label1 = tk.Label(milues_method_frame2,
 text = "[a,b] = [", font = "Calibri 16")
milues_method_interval_label1.pack(side = "left")
milues_method_interval_entry_variable1 = tk.DoubleVar()
milues_method_interval_entry1 = tk.Entry(milues_method_frame2,
 font = ("Calibri 16"),
 textvariable = milues_method_interval_entry_variable1)
milues_method_interval_entry1.pack(side = "left")
milues_method_interval_label2 = tk.Label(milues_method_frame2,
 text = ",", font = "Calibri 16")
milues_method_interval_label2.pack(side = "left")
milues_method_interval_entry_variable2 = tk.DoubleVar()
milues_method_interval_entry2 = tk.Entry(milues_method_frame2,
 font = ("Calibri 16"),
 textvariable = milues_method_interval_entry_variable2)
milues_method_interval_entry2.pack(side = "left")
milues_method_interval_label3 = tk.Label(milues_method_frame2,
 text = "]", font = "Calibri 16")
milues_method_interval_label3.pack(side = "left")

#Initial Condition

milues_method_frame3 = tk.Frame(milues_method)
milues_method_frame3.pack()
milues_method_initial_condition_label = tk.Label(milues_method_frame3,
 text = "y(a) = ", font = "Calibri 16")
milues_method_initial_condition_label.pack(side = "left")
milues_method_initial_condition_entry_variable = tk.DoubleVar()
milues_method_initial_condition_entry = tk.Entry(milues_method_frame3,
 font = ("Calibri 16"),
 textvariable = milues_method_initial_condition_entry_variable)
milues_method_initial_condition_entry.pack(side = "left")

#Number of steps

milues_method_frame4 = tk.Frame(milues_method)
milues_method_frame4.pack()
milues_method_n_label = tk.Label(milues_method_frame4, text = "N = ",
 font = "Calibri 16")
milues_method_n_label.pack(side = "left")
milues_method_n_entry_variable = tk.IntVar()
milues_method_n_entry = tk.Entry(milues_method_frame4,
 font = ("Calibri 16"), textvariable = milues_method_n_entry_variable)
milues_method_n_entry.pack(side = "left")

#Submit

milues_method_submit_button = tk.Button(milues_method, text = "Submit",
 command = milues_method_submit)
milues_method_submit_button.pack()

#output

milues_method_output_string = tk.StringVar()
milues_method_output_label = tk.Label(milues_method,
 textvariable = milues_method_output_string, font = "Calibri 16")
milues_method_output_label.pack()

#Extrapolation

extrapolation = tk.Frame(numerical_solver_notebook)
extrapolation_label = tk.Label(extrapolation, text = "Extrapolation",
 font = "Calibri 20 bold")
extrapolation_label.pack()

#Function

extrapolation_frame1 = tk.Frame(extrapolation)
extrapolation_frame1.pack()
extrapolation_function_label = tk.Label(extrapolation_frame1,
 text = "y' = f(x, y) = ", font = "Calibri 16")
extrapolation_function_label.pack(side = "left")
extrapolation_function_entry_variable = tk.StringVar()
extrapolation_function_entry = tk.Entry(extrapolation_frame1,
 font = ("Calibri 16"),
 textvariable = extrapolation_function_entry_variable)
extrapolation_function_entry.pack(side = "left")

#Interval

extrapolation_frame2 = tk.Frame(extrapolation)
extrapolation_frame2.pack()
extrapolation_interval_label1 = tk.Label(extrapolation_frame2,
 text = "[a,b] = [", font = "Calibri 16")
extrapolation_interval_label1.pack(side = "left")
extrapolation_interval_entry_variable1 = tk.DoubleVar()
extrapolation_interval_entry1 = tk.Entry(extrapolation_frame2,
 font = ("Calibri 16"),
 textvariable = extrapolation_interval_entry_variable1)
extrapolation_interval_entry1.pack(side = "left")
extrapolation_interval_label2 = tk.Label(extrapolation_frame2,
 text = ",", font = "Calibri 16")
extrapolation_interval_label2.pack(side = "left")
extrapolation_interval_entry_variable2 = tk.DoubleVar()
extrapolation_interval_entry2 = tk.Entry(extrapolation_frame2,
 font = ("Calibri 16"),
 textvariable = extrapolation_interval_entry_variable2)
extrapolation_interval_entry2.pack(side = "left")
extrapolation_interval_label3 = tk.Label(extrapolation_frame2,
 text = "]", font = "Calibri 16")
extrapolation_interval_label3.pack(side = "left")

#Step Sizes

extrapolation_frame3 = tk.Frame(extrapolation)
extrapolation_frame3.pack()
subscripts = str.maketrans("aimnx", "")
label1 = "hmin = "
label1 = label1.translate(subscripts)
extrapolation_h_value_label1 = tk.Label(extrapolation_frame3,
 text = label1, font = "Calibri 16")
extrapolation_h_value_label1.pack(side = "left")
extrapolation_h_value_entry_variable1 = tk.DoubleVar()
extrapolation_h_value_entry1 = tk.Entry(extrapolation_frame3,
 font = ("Calibri 16"),
 textvariable = extrapolation_h_value_entry_variable1)
extrapolation_h_value_entry1.pack(side = "left")
label2 = ", hmax = "
label2 = label2.translate(subscripts)
extrapolation_h_value_label2 = tk.Label(extrapolation_frame3,
 text = label2, font = "Calibri 16")
extrapolation_h_value_label2.pack(side = "left")
extrapolation_h_value_entry_variable2 = tk.DoubleVar()
extrapolation_h_value_entry2 = tk.Entry(extrapolation_frame3,
 font = ("Calibri 16"),
 textvariable = extrapolation_h_value_entry_variable2)
extrapolation_h_value_entry2.pack(side = "left")

#Initial Condition

extrapolation_frame4 = tk.Frame(extrapolation)
extrapolation_frame4.pack()
extrapolation_initial_condition_label = tk.Label(extrapolation_frame4,
 text = "y(a) = ", font = "Calibri 16")
extrapolation_initial_condition_label.pack(side = "left")
extrapolation_initial_condition_entry_variable = tk.DoubleVar()
extrapolation_initial_condition_entry = tk.Entry(extrapolation_frame4,
 font = ("Calibri 16"),
 textvariable = extrapolation_initial_condition_entry_variable)
extrapolation_initial_condition_entry.pack(side = "left")

#Tolerance

extrapolation_frame5 = tk.Frame(extrapolation)
extrapolation_frame5.pack()
extrapolation_tolerance_label = tk.Label(extrapolation_frame5,
 text = "Tolerance: ", font = "Calibri 16")
extrapolation_tolerance_label.pack(side = "left")
extrapolation_tolerance_variable = tk.DoubleVar()
extrapolation_tolerance_entry = tk.Entry(extrapolation_frame5,
 font = ("Calibri 16"),
 textvariable = extrapolation_tolerance_variable)
extrapolation_tolerance_entry.pack(side = "left")

#Submit

extrapolation_submit_button = tk.Button(extrapolation,
 text = "Submit", command = extrapolation_submit)
extrapolation_submit_button.pack()

#output

extrapolation_output_string = tk.StringVar()
extrapolation_output_label = tk.Label(extrapolation,
 textvariable = extrapolation_output_string, font = "Calibri 16")
extrapolation_output_label.pack()

#Create numerical integrator notebook tabs

numerical_solver_notebook.add(eulers_method, text = "Euler's Method")
numerical_solver_notebook.add(midpoint_method,
 text = "Midpoint Method")
numerical_solver_notebook.add(modified_euler_method,
 text = "Modified Euler Method")
numerical_solver_notebook.add(heuns_method, text = "Heun's Method")
numerical_solver_notebook.add(runge_kutta_method,
 text = "Runge-Kutta Method")
numerical_solver_notebook.add(runge_kutta_fehlberg_method,
 text = "Runge-Kutta-Fehlberg Method")
numerical_solver_notebook.add(adams_bashforth_technique,
 text = "Adams-Bashforth Technique")
numerical_solver_notebook.add(milues_method, text = "Milue's Method")
numerical_solver_notebook.add(extrapolation, text = "Extrapolation")
numerical_solver_notebook.pack()

#Create main notebook tabs

notebook_main.add(root_finder, text = "Root Finder")
notebook_main.add(interpolation_calculator,
 text = "Interpolation Calculator")
notebook_main.add(numerical_differentiator,
 text = "Numerical Differentiator")
notebook_main.add(numerical_integrator, text = "Numerical Integrator")
notebook_main.add(numerical_solver, text = "Numerical Solver")
notebook_main.pack()

#signature

signature_frame = tk.Frame(window)
signature_frame.pack(side = "bottom")
signature_label = tk.Label(signature_frame,
 text = "Created by Joseph W Barnett", font = "Calibri 10")
signature_label.pack()

#Help & Acknowledgements

bottom_frame = tk.Frame(window)
bottom_frame.pack(side = "bottom", fill = "x")
help_button = tk.Button(bottom_frame, text = "Help", command = help1)
help_button.pack(side = "left")
acknowledgements_button = tk.Button(bottom_frame,
 text = "Acknowledgements", command = acknowledgements)
acknowledgements_button.pack(side = "right")

#Run

window.mainloop()
```
