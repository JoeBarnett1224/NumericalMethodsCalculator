```python
import math as m
import mpmath as mp
import numpy as np
import tkinter as tk
from tkinter import ttk

class App(tk.Tk):
    def __init__(self, title, state):
        super().__init__()
        self.title(title)
        self.state(state)

        #Set Title Label

        Widgets.title_label(self, "Numerical Methods Calculator", 24)

        #Create main notebook

        notebook_main = ttk.Notebook(self)

        #Root finder

        root_finder = tk.Frame(notebook_main)

        #Create root finder notebook

        root_finder_notebook = ttk.Notebook(root_finder)

        #Bisection method

        bisection_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(bisection_method, "Bisection Method", 20)
        self.bisection_method_function_entry_variable = \
         Widgets.label_and_entry(bisection_method, "f(x) = ", "string")
        self.bisection_method_interval_entry_variable1, \
         self.bisection_method_interval_entry_variable2 = \
         Widgets.label3_and_entry2(bisection_method, "[a,b] = [", ",", "]")
        self.bisection_method_tolerance_variable = Widgets.label_and_entry(bisection_method,
         "Tolerance: ", "double")
        Widgets.submit(bisection_method, lambda: Commands.bisection_method_submit(self))
        self.bisection_method_output_string = Widgets.output(bisection_method)

        #Fixed point iteration

        fixed_point_iteration = tk.Frame(root_finder_notebook)
        Widgets.title_label(fixed_point_iteration, "Fixed Point Iteration", 20)
        self.fixed_point_iteration_function_entry_variable = \
         Widgets.label_and_entry(fixed_point_iteration, "x = ", "string")
        subscripts = str.maketrans("0", "₀")
        label = "x0 = "
        label = label.translate(subscripts)
        self.fixed_point_iteration_initial_value_entry_variable = \
         Widgets.label_and_entry(fixed_point_iteration, label, "double")
        self.fixed_point_iteration_tolerance_variable = \
         Widgets.label_and_entry(fixed_point_iteration, "Tolerance: ", "double")
        Widgets.submit(fixed_point_iteration,
         lambda: Commands.fixed_point_iteration_submit(self))
        self.fixed_point_iteration_output_string = Widgets.output(fixed_point_iteration)

        #Newton's Method

        newtons_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(newtons_method, "Newton's Method", 20)
        self.newtons_method_function_entry_variable1,\
         self.newtons_method_function_entry_variable2 = \
         Widgets.label2_and_entry2(newtons_method, "f(x) = ", ", f'(x) = ", "string")
        subscripts = str.maketrans("0", "₀")
        label = "x0 = "
        label = label.translate(subscripts)
        self.newtons_method_initial_value_entry_variable = \
         Widgets.label_and_entry(newtons_method, label, "double")
        self.newtons_method_tolerance_variable = Widgets.label_and_entry(newtons_method,
         "Tolerance: ", "double")
        Widgets.submit(newtons_method, lambda: Commands.newtons_method_submit(self))
        self.newtons_method_output_string = Widgets.output(newtons_method)

        #Secant Method

        secant_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(secant_method, "Secant Method", 20)
        self.secant_method_function_entry_variable = Widgets.label_and_entry(secant_method,
         "f(x) = ", "string")
        subscripts = str.maketrans("01", "₀₁")
        label1 = "x0 = "
        label1 = label1.translate(subscripts)
        label2 = ", x1 = "
        label2 = label2.translate(subscripts)
        self.secant_method_initial_value_entry_variable1, \
         self.secant_method_initial_value_entry_variable2 = \
         Widgets.label2_and_entry2(secant_method, label1, label2, "double")
        self.secant_method_tolerance_variable = Widgets.label_and_entry(secant_method,
         "Tolerance: ", "double")
        Widgets.submit(secant_method, lambda: Commands.secant_method_submit(self))
        self.secant_method_output_string = Widgets.output(secant_method)

        #Method of False Position

        method_of_false_position = tk.Frame(root_finder_notebook)
        Widgets.title_label(method_of_false_position, "Method of False Position", 20)
        self.method_of_false_position_function_entry_variable = \
         Widgets.label_and_entry(method_of_false_position, "f(x) = ", "string")
        self.method_of_false_position_interval_entry_variable1, \
         self.method_of_false_position_interval_entry_variable2 = \
         Widgets.label3_and_entry2(method_of_false_position, "[a,b] = [", ",", "]")
        self.method_of_false_position_tolerance_variable = \
         Widgets.label_and_entry(method_of_false_position, "Tolerance: ", "double")
        Widgets.submit(method_of_false_position,
         lambda: Commands.method_of_false_position_submit(self))
        self.method_of_false_position_output_string = \
         Widgets.output(method_of_false_position)

        #Modified Newton's Method

        modified_newtons_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(modified_newtons_method, "Modified Newton's Method", 20)
        self.modified_newtons_method_function_entry_variable1, \
         self.modified_newtons_method_function_entry_variable2, \
         self.modified_newtons_method_function_entry_variable3 = \
         Widgets.label3_and_entry3(modified_newtons_method, "f(x) = ", ", f'(x) = ",
         ", f''(x) = ", "string")
        subscripts = str.maketrans("0", "₀")
        label = "x0 = "
        label = label.translate(subscripts)
        self.modified_newtons_method_initial_value_entry_variable = \
         Widgets.label_and_entry(modified_newtons_method, label1, "double")
        self.modified_newtons_method_tolerance_variable = \
         Widgets.label_and_entry(modified_newtons_method, "Tolerance: ", "double")
        Widgets.submit(modified_newtons_method,
         lambda: Commands.modified_newtons_method_submit(self))
        self.modified_newtons_method_output_string = Widgets.output(modified_newtons_method)

        #Steffensen's Method

        steffensens_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(steffensens_method, "Steffensen's Method", 20)
        self.steffensens_method_function_entry_variable = \
         Widgets.label_and_entry(steffensens_method, "x = ", "string")
        subscripts = str.maketrans("0", "₀")
        label = "x0 = "
        label = label.translate(subscripts)
        self.steffensens_method_initial_value_entry_variable = \
         Widgets.label_and_entry(steffensens_method, label, "double")
        self.steffensens_method_tolerance_variable = \
         Widgets.label_and_entry(steffensens_method, "Tolerance: ", "double")
        Widgets.submit(steffensens_method, lambda: Commands.steffensens_method_submit(self))
        self.steffensens_method_output_string = Widgets.output(steffensens_method)

        #Muller's Method

        mullers_method = tk.Frame(root_finder_notebook)
        Widgets.title_label(mullers_method, "Müller’s Method", 20)
        self.mullers_method_function_entry_variable = Widgets.label_and_entry(mullers_method,
         "f(x) = ", "string")
        subscripts = str.maketrans("012", "₀₁₂")
        label1 = "x0 = "
        label1 = label1.translate(subscripts)
        label2 = ", x1 = "
        label2 = label2.translate(subscripts)
        label3 = ", x2 = "
        label3 = label3.translate(subscripts)
        self.mullers_method_initial_value_entry_variable1, \
         self.mullers_method_initial_value_entry_variable2, \
         self.mullers_method_initial_value_entry_variable3 = \
         Widgets.label3_and_entry3(mullers_method, label1, label2, label3, "double")
        self.mullers_method_tolerance_variable = Widgets.label_and_entry(mullers_method,
         "Tolerance: ", "double")
        Widgets.submit(mullers_method, lambda: Commands.mullers_method_submit(self))
        self.mullers_method_output_string = Widgets.output(mullers_method)

        #Create root finder notebook tabs

        root_finder_notebook.add(bisection_method, text = "Bisection Method")
        root_finder_notebook.add(fixed_point_iteration, text = "Fixed Point Iteration")
        root_finder_notebook.add(newtons_method, text = "Newton's Method")
        root_finder_notebook.add(secant_method, text = "Secant Method")
        root_finder_notebook.add(method_of_false_position, text = "Method of False Position")
        root_finder_notebook.add(modified_newtons_method, text = "Modified Newton's Method")
        root_finder_notebook.add(steffensens_method, text = "Steffensen's Method")
        root_finder_notebook.add(mullers_method, text = "Müller’s Method")
        root_finder_notebook.pack()

        #Interpolation Calculator

        interpolation_calculator = tk.Frame(notebook_main)

        #Create interpolation calculator notebook

        interpolation_calculator_notebook = ttk.Notebook(interpolation_calculator)

        #Lagrange interpolating polynomials

        lagrange_interpolating_polynomials = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(lagrange_interpolating_polynomials,
         "Lagrange Interpolating Polynomials", 20)
        self.lagrange_interpolating_polynomials_x_values_entry_variable = \
         Widgets.label_and_entry(lagrange_interpolating_polynomials, "x-values: :", "string")
        self.lagrange_interpolating_polynomials_y_values_entry_variable = \
         Widgets.label_and_entry(lagrange_interpolating_polynomials, "y-values: ", "string")
        self.lagrange_interpolating_polynomials_interpolation_point_entry_variable = \
         Widgets.label_and_entry(lagrange_interpolating_polynomials, "x = ", "double")
        Widgets.submit(lagrange_interpolating_polynomials,
         lambda: Commands.lagrange_interpolating_polynomials_submit(self))
        self.lagrange_interpolating_polynomials_output_string = \
         Widgets.output(lagrange_interpolating_polynomials)

        #Neville's method

        nevilles_method = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(nevilles_method, "Neville's Method", 20)
        self.nevilles_method_x_values_entry_variable = \
         Widgets.label_and_entry(nevilles_method, "x-values: :", "string")
        self.nevilles_method_y_values_entry_variable = \
         Widgets.label_and_entry(nevilles_method, "y-values: ", "string")
        self.nevilles_method_interpolation_point_entry_variable = \
         Widgets.label_and_entry(nevilles_method, "x = ", "double")
        Widgets.submit(nevilles_method, lambda: Commands.nevilles_method_submit(self))
        self.nevilles_method_output_string = Widgets.output(nevilles_method)

        #Newton's Divided Difference Method

        newtons_divided_difference_method = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(newtons_divided_difference_method,
         "Newton's Divided Difference Method", 20)
        self.newtons_divided_difference_method_x_values_entry_variable = \
         Widgets.label_and_entry(newtons_divided_difference_method, "x-values: :", "string")
        self.newtons_divided_difference_method_y_values_entry_variable = \
         Widgets.label_and_entry(newtons_divided_difference_method, "y-values: ", "string")
        self.newtons_divided_difference_method_interpolation_point_entry_variable = \
         Widgets.label_and_entry(newtons_divided_difference_method, "x = ", "double")
        Widgets.submit(newtons_divided_difference_method,
         lambda: Commands.newtons_divided_difference_method_submit(self))
        self.newtons_divided_difference_method_output_string = \
         Widgets.output(newtons_divided_difference_method)

        #Hermite interpolation

        hermite_interpolation = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(hermite_interpolation, "Hermite Interpolation", 20)
        self.hermite_interpolation_x_values_entry_variable = \
         Widgets.label_and_entry(hermite_interpolation, "x-values: ", "string")
        self.hermite_interpolation_y_values_entry_variable, \
         self.hermite_interpolation_y_prime_values_entry_variable = \
         Widgets.label2_and_entry2(hermite_interpolation, "y-values: ", ", y'-values: ",
         "string")
        self.hermite_interpolation_interpolation_point_entry_variable = \
         Widgets.label_and_entry(hermite_interpolation, "x = ", "double")
        Widgets.submit(hermite_interpolation,
         lambda: Commands.hermite_interpolation_submit(self))
        self.hermite_interpolation_output_string = Widgets.output(hermite_interpolation)

        #Natural Cubic Spline

        natural_cubic_spline = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(natural_cubic_spline, "Natural Cubic Spline", 20)
        self.natural_cubic_spline_x_values_entry_variable = \
         Widgets.label_and_entry(natural_cubic_spline, "x-values: :", "string")
        self.natural_cubic_spline_y_values_entry_variable = \
         Widgets.label_and_entry(natural_cubic_spline, "y-values: ", "string")
        self.natural_cubic_spline_interpolation_point_entry_variable = \
         Widgets.label_and_entry(natural_cubic_spline, "x = ", "double")
        Widgets.submit(natural_cubic_spline,
         lambda: Commands.natural_cubic_spline_submit(self))
        self.natural_cubic_spline_output_string = Widgets.output(natural_cubic_spline)

        #Clamped Cubic Spline

        clamped_cubic_spline = tk.Frame(interpolation_calculator_notebook)
        Widgets.title_label(clamped_cubic_spline, "Clamped Cubic Spline", 20)
        self.clamped_cubic_spline_x_values_entry_variable = \
         Widgets.label_and_entry(clamped_cubic_spline, "x-values: ", "string")
        self.clamped_cubic_spline_y_values_entry_variable = \
         Widgets.label_and_entry(clamped_cubic_spline, "y-values: ", "string")
        self.clamped_cubic_spline_initial_condition_entry_variable1, \
         self.clamped_cubic_spline_initial_condition_entry_variable2 = \
         Widgets.label2_and_entry2(clamped_cubic_spline, "f'(a) = ", ", f'(b) = ", "double")
        self.clamped_cubic_spline_interpolation_point_entry_variable = \
         Widgets.label_and_entry(clamped_cubic_spline, "x = ", "double")
        Widgets.submit(clamped_cubic_spline,
         lambda: Commands.clamped_cubic_spline_submit(self))
        self.clamped_cubic_spline_output_string = Widgets.output(clamped_cubic_spline)

        #Create interpolation calculator notebook tabs

        interpolation_calculator_notebook.add(lagrange_interpolating_polynomials,
         text = "Lagrange Interpolating Polynomials")
        interpolation_calculator_notebook.add(nevilles_method, text = "Neville's Method")
        interpolation_calculator_notebook.add(newtons_divided_difference_method,
         text = "Newton's Divided Difference Method")
        interpolation_calculator_notebook.add(hermite_interpolation,
         text = "Hermite Interpolation")
        interpolation_calculator_notebook.add(natural_cubic_spline,
         text = "Natural Cubic Spline")
        interpolation_calculator_notebook.add(clamped_cubic_spline,
         text = "Clamped Cubic Spline")
        interpolation_calculator_notebook.pack()

        #Numerical Differentiator

        numerical_differentiator = tk.Frame(notebook_main)

        #Create numerical differentiator notebook

        numerical_differentiator_notebook = ttk.Notebook(numerical_differentiator)

        #Forward Difference

        forward_difference = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(forward_difference, "Forward Difference", 20)
        self.forward_difference_function_entry_variable = \
         Widgets.label_and_entry(forward_difference, "f(x) = ", "string")
        self.forward_difference_x_value_entry_variable = \
         Widgets.label_and_entry(forward_difference, "x = ", "double")
        self.forward_difference_h_value_entry_variable = \
         Widgets.label_and_entry(forward_difference, "h = ", "double")
        Widgets.submit(forward_difference, lambda: Commands.forward_difference_submit(self))
        self.forward_difference_output_string = Widgets.output(forward_difference)

        #Backward Difference

        backward_difference = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(backward_difference, "Backward Difference", 20)
        self.backward_difference_function_entry_variable = \
         Widgets.label_and_entry(backward_difference, "f(x) = ", "string")
        self.backward_difference_x_value_entry_variable = \
         Widgets.label_and_entry(backward_difference, "x = ", "double")
        self.backward_difference_h_value_entry_variable = \
         Widgets.label_and_entry(backward_difference, "h = ", "double")
        Widgets.submit(backward_difference,
         lambda: Commands.backward_difference_submit(self))
        self.backward_difference_output_string = Widgets.output(backward_difference)

        #Central Difference

        central_difference = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(central_difference, "Central Difference", 20)
        self.central_difference_function_entry_variable = \
         Widgets.label_and_entry(central_difference, "f(x) = ", "string")
        self.central_difference_x_value_entry_variable = \
         Widgets.label_and_entry(central_difference, "x = ", "double")
        self.central_difference_h_value_entry_variable = \
         Widgets.label_and_entry(central_difference, "h = ", "double")
        Widgets.submit(central_difference, lambda: Commands.central_difference_submit(self))
        self.central_difference_output_string = Widgets.output(central_difference)

        #Three-point midpoint formula

        three_point_midpoint_formula = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(three_point_midpoint_formula, "Three-Point Midpoint Formula", 20)
        self.three_point_midpoint_formula_function_entry_variable = \
         Widgets.label_and_entry(three_point_midpoint_formula, "f(x) = ", "string")
        self.three_point_midpoint_formula_x_value_entry_variable = \
         Widgets.label_and_entry(three_point_midpoint_formula, "x = ", "double")
        self.three_point_midpoint_formula_h_value_entry_variable = \
         Widgets.label_and_entry(three_point_midpoint_formula, "h = ", "double")
        Widgets.submit(three_point_midpoint_formula, 
         lambda: Commands.three_point_midpoint_formula_submit(self))
        self.three_point_midpoint_formula_output_string = \
         Widgets.output(three_point_midpoint_formula)

        #Three-point endpoint formula

        three_point_endpoint_formula = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(three_point_endpoint_formula, "Three-Point Endpoint Formula", 20)
        self.three_point_endpoint_formula_function_entry_variable = \
         Widgets.label_and_entry(three_point_endpoint_formula, "f(x) = ", "string")
        self.three_point_endpoint_formula_x_value_entry_variable = \
         Widgets.label_and_entry(three_point_endpoint_formula, "x = ", "double")
        self.three_point_endpoint_formula_h_value_entry_variable = \
         Widgets.label_and_entry(three_point_endpoint_formula, "h = ", "double")
        Widgets.submit(three_point_endpoint_formula,
         lambda: Commands.three_point_endpoint_formula_submit(self))
        self.three_point_endpoint_formula_output_string = \
         Widgets.output(three_point_endpoint_formula)

        #Five-point midpoint formula

        five_point_midpoint_formula = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(five_point_midpoint_formula, "Five-Point Midpoint Formula", 20)
        self.five_point_midpoint_formula_function_entry_variable = \
         Widgets.label_and_entry(five_point_midpoint_formula, "f(x) = ", "string")
        self.five_point_midpoint_formula_x_value_entry_variable = \
         Widgets.label_and_entry(five_point_midpoint_formula, "x = ", "double")
        self.five_point_midpoint_formula_h_value_entry_variable = \
         Widgets.label_and_entry(five_point_midpoint_formula, "h = ", "double")
        Widgets.submit(five_point_midpoint_formula,
         lambda: Commands.five_point_midpoint_formula_submit(self))
        self.five_point_midpoint_formula_output_string = \
         Widgets.output(five_point_midpoint_formula)

        #Five-point endpoint formula

        five_point_endpoint_formula = tk.Frame(numerical_differentiator_notebook)
        Widgets.title_label(five_point_endpoint_formula, "Five-Point Endpoint Formula", 20)
        self.five_point_endpoint_formula_function_entry_variable = \
         Widgets.label_and_entry(five_point_endpoint_formula, "f(x) = ", "string")
        self.five_point_endpoint_formula_x_value_entry_variable = \
         Widgets.label_and_entry(five_point_endpoint_formula, "x = ", "double")
        self.five_point_endpoint_formula_h_value_entry_variable = \
         Widgets.label_and_entry(five_point_endpoint_formula, "h = ", "double")
        Widgets.submit(five_point_endpoint_formula,
         lambda: Commands.five_point_endpoint_formula_submit(self))
        self.five_point_endpoint_formula_output_string = \
         Widgets.output(five_point_endpoint_formula)

        #Create numerical differentiator notebook tabs

        numerical_differentiator_notebook.add(forward_difference,
         text = "Forward Difference")
        numerical_differentiator_notebook.add(backward_difference,
         text = "Backward Difference")
        numerical_differentiator_notebook.add(central_difference,
         text = "Central Difference")
        numerical_differentiator_notebook.add(three_point_midpoint_formula,
         text = "Three-Point Midpoint Formula")
        numerical_differentiator_notebook.add(three_point_endpoint_formula,
         text = "Three-Point Endpoint Formula")
        numerical_differentiator_notebook.add(five_point_midpoint_formula,
         text = "Five-Point Midpoint Formula")
        numerical_differentiator_notebook.add(five_point_endpoint_formula,
         text = "Five-Point Endpoint Formula")
        numerical_differentiator_notebook.pack()

        #Numerical Integrator

        numerical_integrator = tk.Frame(notebook_main)

        #Create numerical integrator notebook

        numerical_integrator_notebook = ttk.Notebook(numerical_integrator)

        #Midpoint rule

        midpoint_rule = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(midpoint_rule, "Midpoint Rule", 20)
        self.midpoint_rule_function_entry_variable = Widgets.label_and_entry(midpoint_rule,
         "f(x) = ", "string")
        self.midpoint_rule_interval_entry_variable1, \
         self.midpoint_rule_interval_entry_variable2 = \
         Widgets.label3_and_entry2(midpoint_rule, "[a,b] = [", ",", "]")
        self.midpoint_rule_n_entry_variable = Widgets.label_and_entry(midpoint_rule, "n = ",
         "integer")
        Widgets.submit(midpoint_rule, lambda: Commands.midpoint_rule_submit(self))
        self.midpoint_rule_output_string = Widgets.output(midpoint_rule)

        #Trapezoid rule

        trapezoid_rule = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(trapezoid_rule, "Trapezoid Rule", 20)
        self.trapezoid_rule_function_entry_variable = Widgets.label_and_entry(trapezoid_rule,
         "f(x) = ", "string")
        self.trapezoid_rule_interval_entry_variable1, \
         self.trapezoid_rule_interval_entry_variable2 = \
         Widgets.label3_and_entry2(trapezoid_rule, "[a,b] = [", ",", "]")
        self.trapezoid_rule_n_entry_variable = Widgets.label_and_entry(trapezoid_rule,
         "n = ", "integer")
        Widgets.submit(trapezoid_rule, lambda: Commands.trapezoid_rule_submit(self))
        self.trapezoid_rule_output_string = Widgets.output(trapezoid_rule)

        #Simpson's rule

        simpsons_rule = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(simpsons_rule, "Simpson's Rule", 20)
        self.simpsons_rule_function_entry_variable = Widgets.label_and_entry(simpsons_rule,
         "f(x) = ", "string")
        self.simpsons_rule_interval_entry_variable1, \
         self.simpsons_rule_interval_entry_variable2 = \
         Widgets.label3_and_entry2(simpsons_rule, "[a,b] = [", ",", "]")
        self.simpsons_rule_n_entry_variable = Widgets.label_and_entry(simpsons_rule, "n = ",
         "integer")
        Widgets.submit(simpsons_rule, lambda: Commands.simpsons_rule_submit(self))
        self.simpsons_rule_output_string = Widgets.output(simpsons_rule)

        #Romberg Integration

        romberg_integration = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(romberg_integration, "Romberg Integration", 20)
        self.romberg_integration_function_entry_variable = \
         Widgets.label_and_entry(romberg_integration, "f(x) = ", "string")
        self.romberg_integration_interval_entry_variable1, \
         self.romberg_integration_interval_entry_variable2 = \
         Widgets.label3_and_entry2(romberg_integration, "[a,b] = [", ",", "]")
        self.romberg_integration_tolerance_variable = \
         Widgets.label_and_entry(romberg_integration, "Tolerance: ", "double")
        Widgets.submit(romberg_integration,
         lambda: Commands.romberg_integration_submit(self))
        self.romberg_integration_output_string = Widgets.output(romberg_integration)

        #Adaptive quadrature

        adaptive_quadrature = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(adaptive_quadrature, "Adaptive Quadrature", 20)
        self.adaptive_quadrature_function_entry_variable = \
         Widgets.label_and_entry(adaptive_quadrature, "f(x) = ", "string")
        self.adaptive_quadrature_interval_entry_variable1, \
         self.adaptive_quadrature_interval_entry_variable2 = \
         Widgets.label3_and_entry2(adaptive_quadrature, "[a,b] = [", ",", "]")
        self.adaptive_quadrature_tolerance_variable = \
         Widgets.label_and_entry(adaptive_quadrature, "Tolerance: ", "double")
        Widgets.submit(adaptive_quadrature,
         lambda: Commands.adaptive_quadrature_submit(self))
        self.adaptive_quadrature_output_string = Widgets.output(adaptive_quadrature)

        #Gaussian quadrature

        gaussian_quadrature = tk.Frame(numerical_integrator_notebook)
        Widgets.title_label(gaussian_quadrature, "Gaussian Quadrature", 20)
        self.gaussian_quadrature_function_entry_variable = \
         Widgets.label_and_entry(gaussian_quadrature, "f(x) = ", "string")
        self.gaussian_quadrature_n_entry_variable = \
         Widgets.label_and_entry(gaussian_quadrature, "n = ", "integer")
        Widgets.submit(gaussian_quadrature,
         lambda: Commands.gaussian_quadrature_submit(self))
        self.gaussian_quadrature_output_string = Widgets.output(gaussian_quadrature)

        #Create numerical integrator notebook tabs

        numerical_integrator_notebook.add(midpoint_rule, text = "Midpoint Rule")
        numerical_integrator_notebook.add(trapezoid_rule, text = "Trapezoid Rule")
        numerical_integrator_notebook.add(simpsons_rule, text = "Simpson's Rule")
        numerical_integrator_notebook.add(romberg_integration, text = "Romberg Integration")
        numerical_integrator_notebook.add(adaptive_quadrature, text = "Adaptive Quadrature")
        numerical_integrator_notebook.add(gaussian_quadrature, text = "Gaussian Quadrature")
        numerical_integrator_notebook.pack()

        #Numerical Solver

        numerical_solver = tk.Frame(notebook_main)

        #Create numerical solver notebook

        numerical_solver_notebook = ttk.Notebook(numerical_solver)

        #Euler's Method

        eulers_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(eulers_method, "Euler's Method", 20)
        self.eulers_method_function_entry_variable = Widgets.label_and_entry(eulers_method,
         "y' = f(x, y) = ", "string")
        self.eulers_method_interval_entry_variable1, \
         self.eulers_method_interval_entry_variable2 = \
         Widgets.label3_and_entry2(eulers_method, "[a,b] = [", ",", "]")
        self.eulers_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(eulers_method, "y(a) = ", "double")
        self.eulers_method_n_entry_variable = Widgets.label_and_entry(eulers_method, "N = ",
         "integer")
        Widgets.submit(eulers_method, lambda: Commands.eulers_method_submit(self))
        self.eulers_method_output_string = Widgets.output(eulers_method)

        #Midpoint Method

        midpoint_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(midpoint_method, "Midpoint Method", 20)
        self.midpoint_method_function_entry_variable = \
         Widgets.label_and_entry(midpoint_method, "y' = f(x, y) = ", "string")
        self.midpoint_method_interval_entry_variable1, \
         self.midpoint_method_interval_entry_variable2 = \
         Widgets.label3_and_entry2(midpoint_method, "[a,b] = [", ",", "]")
        self.midpoint_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(midpoint_method, "y(a) = ", "double")
        self.midpoint_method_n_entry_variable = Widgets.label_and_entry(midpoint_method,
         "N = ", "integer")
        Widgets.submit(midpoint_method, lambda: Commands.midpoint_method_submit(self))
        self.midpoint_method_output_string = Widgets.output(midpoint_method)

        #Modified Euler Method

        modified_euler_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(modified_euler_method, "Modified Euler Method", 20)
        self.modified_euler_method_function_entry_variable = \
         Widgets.label_and_entry(modified_euler_method, "y' = f(x, y) = ", "string")
        self.modified_euler_method_interval_entry_variable1, \
         self.modified_euler_method_interval_entry_variable2 = \
        Widgets.label3_and_entry2(modified_euler_method, "[a,b] = [", ",", "]")
        self.modified_euler_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(modified_euler_method, "y(a) = ", "double")
        self.modified_euler_method_n_entry_variable = \
         Widgets.label_and_entry(modified_euler_method, "N = ", "integer")
        Widgets.submit(modified_euler_method,
         lambda: Commands.modified_euler_method_submit(self))
        self.modified_euler_method_output_string = Widgets.output(modified_euler_method)

        #Heun's Method

        heuns_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(heuns_method, "Heun's Method", 20)
        self.heuns_method_function_entry_variable = Widgets.label_and_entry(heuns_method,
         "y' = f(x, y) = ", "string")
        self.heuns_method_interval_entry_variable1, \
         self.heuns_method_interval_entry_variable2 = Widgets.label3_and_entry2(heuns_method,
         "[a,b] = [", ",", "]")
        self.heuns_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(heuns_method, "y(a) = ", "double")
        self.heuns_method_n_entry_variable = Widgets.label_and_entry(heuns_method, "N = ",
         "integer")
        Widgets.submit(heuns_method, lambda: Commands.heuns_method_submit(self))
        self.heuns_method_output_string = Widgets.output(heuns_method)

        #Runge-Kutta Method

        runge_kutta_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(runge_kutta_method, "Runge-Kutta Method", 20)
        self.runge_kutta_method_function_entry_variable = \
         Widgets.label_and_entry(runge_kutta_method, "y' = f(x, y) = ", "string")
        self.runge_kutta_method_interval_entry_variable1, \
         self.runge_kutta_method_interval_entry_variable2 = \
         Widgets.label3_and_entry2(runge_kutta_method, "[a,b] = [", ",", "]")
        self.runge_kutta_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(runge_kutta_method, "y(a) = ", "double")
        self.runge_kutta_method_n_entry_variable = \
         Widgets.label_and_entry(runge_kutta_method, "N = ", "integer")
        Widgets.submit(runge_kutta_method, lambda: Commands.runge_kutta_method_submit(self))
        self.runge_kutta_method_output_string = Widgets.output(runge_kutta_method)

        #Adams-Bashforth Technique

        adams_bashforth_technique = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(adams_bashforth_technique, "Adams-Bashforth Technique", 20)
        self.adams_bashforth_technique_function_entry_variable = \
         Widgets.label_and_entry(adams_bashforth_technique, "y' = f(x, y) = ", "string")
        self.adams_bashforth_technique_interval_entry_variable1, \
         self.adams_bashforth_technique_interval_entry_variable2 = \
         Widgets.label3_and_entry2(adams_bashforth_technique, "[a,b] = [", ",", "]")
        self.adams_bashforth_technique_initial_condition_entry_variable = \
         Widgets.label_and_entry(adams_bashforth_technique, "y(a) = ", "double")
        self.adams_bashforth_technique_n_entry_variable = \
         Widgets.label_and_entry(adams_bashforth_technique, "N = ", "integer")
        Widgets.submit(adams_bashforth_technique,
         lambda: Commands.adams_bashforth_technique_submit(self))
        self.adams_bashforth_technique_output_string = \
         Widgets.output(adams_bashforth_technique)

        #Milue's Method

        milues_method = tk.Frame(numerical_solver_notebook)
        Widgets.title_label(milues_method, "Milue's Method", 20)
        self.milues_method_function_entry_variable = Widgets.label_and_entry(milues_method,
         "y' = f(x, y) = ", "string")
        self.milues_method_interval_entry_variable1, \
         self.milues_method_interval_entry_variable2 = \
         Widgets.label3_and_entry2(milues_method, "[a,b] = [", ",", "]")
        self.milues_method_initial_condition_entry_variable = \
         Widgets.label_and_entry(milues_method, "y(a) = ", "double")
        self.milues_method_n_entry_variable = Widgets.label_and_entry(milues_method, "N = ",
         "integer")
        Widgets.submit(milues_method, lambda: Commands.milues_method_submit(self))
        self.milues_method_output_string = Widgets.output(milues_method)

        #Create numerical Solver notebook tabs

        numerical_solver_notebook.add(eulers_method, text = "Euler's Method")
        numerical_solver_notebook.add(midpoint_method, text = "Midpoint Method")
        numerical_solver_notebook.add(modified_euler_method, text = "Modified Euler Method")
        numerical_solver_notebook.add(heuns_method, text = "Heun's Method")
        numerical_solver_notebook.add(runge_kutta_method, text = "Runge-Kutta Method")
        numerical_solver_notebook.add(adams_bashforth_technique,
         text = "Adams-Bashforth Technique")
        numerical_solver_notebook.add(milues_method, text = "Milue's Method")
        numerical_solver_notebook.pack()

        #Create main notebook tabs

        notebook_main.add(root_finder, text = "Root Finder")
        notebook_main.add(interpolation_calculator, text = "Interpolation Calculator")
        notebook_main.add(numerical_differentiator, text = "Numerical Differentiator")
        notebook_main.add(numerical_integrator, text = "Numerical Integrator")
        notebook_main.add(numerical_solver, text = "Numerical Solver")
        notebook_main.pack()

        #signature

        signature_frame = Widgets.bottom_frame(self, "none")
        Widgets.paragraph_label(signature_frame, "Created by Joseph W Barnett", 10)

        #Help & Acknowledgements

        bottom_frame1 = Widgets.bottom_frame(self, "x")
        Widgets.window_button(bottom_frame1, "Help", lambda: Commands.help1(self), "left")
        Widgets.window_button(bottom_frame1, "Acknowledgements",
         lambda: Commands.acknowledgements(self), "right")

        #Run

        self.mainloop()
        
class Widgets:
    def title_label(frame, text1, size):
        label1 = tk.Label(frame, text = text1, font = f"Calibri {str(size)} bold")
        label1.pack()
    
    def paragraph_label(frame, text1, size):
        label1 = tk.Label(frame, text = text1, font = f"Calibri {str(size)}")
        label1.pack()

    def label_and_entry(frame, text1, var):
        frame1 = tk.Frame(frame)
        frame1.pack()
        label1 = tk.Label(frame1, text = text1, font = "Calibri 16")
        label1.pack(side = "left")
        if var == "string":
            entry_variable1 = tk.StringVar()
        if var == "double":
            entry_variable1 = tk.DoubleVar()
        if var == "integer":
            entry_variable1 = tk.IntVar()
        entry1 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable1)
        entry1.pack(side = "left")
        return entry_variable1

    def label2_and_entry2(frame, text1, text2, var):
        frame1 = tk.Frame(frame)
        frame1.pack()
        label1 = tk.Label(frame1, text = text1, font = "Calibri 16")
        label1.pack(side = "left")
        if var == "string":
            entry_variable1 = tk.StringVar()
        if var == "double":
            entry_variable1 = tk.DoubleVar()
        entry1 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable1)
        entry1.pack(side = "left")
        label2 = tk.Label(frame1, text = text2, font = "Calibri 16")
        label2.pack(side = "left")
        if var == "string":
            entry_variable2 = tk.StringVar()
        if var == "double":
            entry_variable2 = tk.DoubleVar()
        entry2 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable2)
        entry2.pack(side = "left")
        return entry_variable1, entry_variable2

    def label3_and_entry2(frame, text1, text2, text3):
        frame1 = tk.Frame(frame)
        frame1.pack()
        label1 = tk.Label(frame1, text = text1, font = "Calibri 16")
        label1.pack(side = "left")
        entry_variable1 = tk.DoubleVar()
        entry1 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable1)
        entry1.pack(side = "left")
        label2 = tk.Label(frame1, text = text2, font = "Calibri 16")
        label2.pack(side = "left")
        entry_variable2 = tk.DoubleVar()
        entry2 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable2)
        entry2.pack(side = "left")
        label3 = tk.Label(frame1, text = text3, font = "Calibri 16")
        label3.pack(side = "left")
        return entry_variable1, entry_variable2

    def label3_and_entry3(frame, text1, text2, text3, var):
        frame1 = tk.Frame(frame)
        frame1.pack()
        label1 = tk.Label(frame1, text = text1, font = "Calibri 16")
        label1.pack(side = "left")
        if var == "string":
            entry_variable1 = tk.StringVar()
        if var == "double":
            entry_variable1 = tk.DoubleVar()
        entry1 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable1)
        entry1.pack(side = "left")
        label2 = tk.Label(frame1, text = text2, font = "Calibri 16")
        label2.pack(side = "left")
        if var == "string":
            entry_variable2 = tk.StringVar()
        if var == "double":
            entry_variable2 = tk.DoubleVar()
        entry2 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable2)
        entry2.pack(side = "left")
        label3 = tk.Label(frame1, text = text3, font = "Calibri 16")
        label3.pack(side = "left")
        if var == "string":
            entry_variable3 = tk.StringVar()
        if var == "double":
            entry_variable3 = tk.DoubleVar()
        entry3 = tk.Entry(frame1, font = ("Calibri 16"), textvariable = entry_variable3)
        entry3.pack(side = "left")
        return entry_variable1, entry_variable2, entry_variable3

    def submit(frame, command1):
        submit_button = tk.Button(frame, text = "Submit", command = command1)
        submit_button.pack()

    def window_button(frame, text1, command1, side1):
        button1 = tk.Button(frame, text = text1, command = command1)
        button1.pack(side = side1)

    def bottom_frame(window, fill1):
        frame1 = tk.Frame(window)
        frame1.pack(side = "bottom", fill = fill1)
        return frame1

    def output(frame):
        output_string = tk.StringVar()
        output_label = tk.Label(frame, textvariable = output_string, font = "Calibri 16")
        output_label.pack()
        return output_string

    def scrollbar(frame):
        frame1 = tk.Frame(frame)
        frame1.pack(fill = "both", expand = True)
        canvas1 = tk.Canvas(frame1)
        canvas1.pack(side = "left", fill = "both", expand = True)
        scrollbar1 = ttk.Scrollbar(frame1, orient = "vertical",
         command = canvas1.yview)
        scrollbar1.pack(side = "right", fill = "y")
        canvas1.configure(yscrollcommand = scrollbar1.set)
        canvas1.bind("<Configure>",
         lambda e: canvas1.configure(scrollregion = \
         canvas1.bbox("all")))
        frame2 = tk.Frame(canvas1)
        canvas1.create_window((0, 0), window = frame2,
         anchor = "nw")
        return frame2

    def treeview(frame, string1, rows1):
        table1 = ttk.Treeview(frame, columns = ("column1",
         "column2", "column3"), show = "headings", height = len(rows1))
        table1.heading("column1", text = f"Mathematical {string1}")
        if string1 == "Function":
            table1.column("column1", width = 300)
        table1.heading("column2", text = "Implementation")
        table1.heading("column3", text = "Python Function")
        if string1 == "Function":
            table1.column("column3", width = 300)
        i = 0
        while i < len(rows1):
            if i == 0:
                table1.insert(parent = "", index = i, values = rows1[i])
                i += 1
            if i > 0:
                table1.insert(parent = "", index = "end", values = rows1[i])
                i += 1
        table1.pack()
        
class Commands:
    def bisection_method_submit(self):
    
        #Get and parse data

        bisection_method_function = str(self.bisection_method_function_entry_variable.get())
        bisection_method_function = Parse.parse_function(bisection_method_function)
        a = str(self.bisection_method_interval_entry_variable1.get())
        b = str(self.bisection_method_interval_entry_variable2.get())
        tolerance = self.bisection_method_tolerance_variable.get()

        #Run bisection method.

        value_at_a = float(Parse.evaluate_function(bisection_method_function, a))
        value_at_b = float(Parse.evaluate_function(bisection_method_function, b))
        if value_at_a * value_at_b > 0:
            output = "Bisection method won't work as f(a) and f(b) have the same sign."
        else:
            error = abs(float(b) - float(a))
            while error > 2 * tolerance:
                c = (float(a) + float(b)) / 2
                value_at_c = \
                 float(Parse.evaluate_function(bisection_method_function, str(c)))
                if value_at_a * value_at_c < 0:
                    b = c
                    error = abs(float(b) - float(a))
                if value_at_b * value_at_c < 0:
                    a = c
                    error = abs(float(b) - float(a))
            value = (float(a) + float(b)) / 2
            error = error / 2

            #Create and set output.

            output = f"x = {str(value)}, Error: {str(error)}."
        self.bisection_method_output_string.set(output)

    def fixed_point_iteration_submit(self):

        #Get and parse data

        fixed_point_iteration_function = \
         str(self.fixed_point_iteration_function_entry_variable.get())
        fixed_point_iteration_function = Parse.parse_function(fixed_point_iteration_function)
        x0 = str(self.fixed_point_iteration_initial_value_entry_variable.get())
        tolerance = self.fixed_point_iteration_tolerance_variable.get()

        #Run fixed point iteration

        x = Parse.evaluate_function(fixed_point_iteration_function, x0)
        error = abs(float(x0) - float(x))
        while error > tolerance:
            x0 = x
            x = Parse.evaluate_function(fixed_point_iteration_function, x0)
            error = abs(float(x0) - float(x))

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.fixed_point_iteration_output_string.set(output)

    def newtons_method_submit(self):

        #Get and parse data

        newtons_method_function1 = str(self.newtons_method_function_entry_variable1.get())
        newtons_method_function1 = Parse.parse_function(newtons_method_function1)
        newtons_method_function2 = str(self.newtons_method_function_entry_variable2.get())
        newtons_method_function2 = Parse.parse_function(newtons_method_function2)
        x0 = str(self.newtons_method_initial_value_entry_variable.get())
        tolerance = self.newtons_method_tolerance_variable.get()

        #Run Newton's Method

        f = Parse.evaluate_function(newtons_method_function1, x0)
        df = Parse.evaluate_function(newtons_method_function2, x0)
        x = float(x0) - (float(f)/float(df))
        error = abs(float(x0) - float(x))
        while error > tolerance:
            x0 = x
            f = Parse.evaluate_function(newtons_method_function1, str(x0))
            df = Parse.evaluate_function(newtons_method_function2, str(x0))
            x = x0 - (float(f)/float(df))
            error = abs(float(x0) - float(x))

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.newtons_method_output_string.set(output)

    def secant_method_submit(self):

        #Get and parse data

        secant_method_function = str(self.secant_method_function_entry_variable.get())
        secant_method_function = Parse.parse_function(secant_method_function)
        x0 = str(self.secant_method_initial_value_entry_variable1.get())
        x1 = str(self.secant_method_initial_value_entry_variable2.get())
        tolerance = self.secant_method_tolerance_variable.get()

        #Run Secant Method

        f0 = Parse.evaluate_function(secant_method_function, x0)
        f1 = Parse.evaluate_function(secant_method_function, x1)
        x = float(x0) - (float(f0) / ((float(f0) - float(f1)) / (float(x0) - float(x1))))
        error = abs(float(x0) - x)
        while error > tolerance:
            x0 = x1
            x1 = x
            f0 = Parse.evaluate_function(secant_method_function, str(x0))
            f1 = Parse.evaluate_function(secant_method_function, str(x1))
            x = float(x0) - (float(f0) / ((float(f0) - float(f1)) / (float(x0) - float(x1))))
            error = abs(float(x0) - x)

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.secant_method_output_string.set(output)

    def method_of_false_position_submit(self):

        #Get and parse data

        method_of_false_position_function = \
         str(self.method_of_false_position_function_entry_variable.get())
        method_of_false_position_function = \
         Parse.parse_function(method_of_false_position_function)
        a = str(self.method_of_false_position_interval_entry_variable1.get())
        b = str(self.method_of_false_position_interval_entry_variable2.get())
        tolerance = self.method_of_false_position_tolerance_variable.get()

        #Run method of false position.

        value_at_a = float(Parse.evaluate_function(method_of_false_position_function, a))
        value_at_b = float(Parse.evaluate_function(method_of_false_position_function, b))
        if value_at_a * value_at_b > 0:
            output = "Method of false position won't work as f(a) and f(b) have the same \
             sign."
        else:
            error = abs(float(b) - float(a))
            while error > tolerance:
                c = (float(a) * value_at_b - float(b) * value_at_a) / (value_at_b - \
                 value_at_a)
                value_at_c = float(Parse.evaluate_function(method_of_false_position_function,
                 str(c)))
                if value_at_a * value_at_c < 0:
                    b = c
                    error = abs(float(b) - float(a))
                if value_at_b * value_at_c < 0:
                    a = c
                    error = abs(float(b) - float(a))

            output = f"x = {str(c)}, Error: {str(error)}."
        self.method_of_false_position_output_string.set(output)

    def modified_newtons_method_submit(self):

        #Get and parse data

        modified_newtons_method_function1 = \
         str(self.modified_newtons_method_function_entry_variable1.get())
        modified_newtons_method_function1 = \
         Parse.parse_function(modified_newtons_method_function1)
        modified_newtons_method_function2 = \
         str(self.modified_newtons_method_function_entry_variable2.get())
        modified_newtons_method_function2 = \
         Parse.parse_function(modified_newtons_method_function2)
        modified_newtons_method_function3 = \
         str(self.modified_newtons_method_function_entry_variable3.get())
        modified_newtons_method_function3 = \
         Parse.parse_function(modified_newtons_method_function3)
        x0 = str(self.modified_newtons_method_initial_value_entry_variable.get())
        tolerance = self.modified_newtons_method_tolerance_variable.get()

        #Run Modified Newton's Method

        f = Parse.evaluate_function(modified_newtons_method_function1, x0)
        df = Parse.evaluate_function(modified_newtons_method_function2, x0)
        d2f = Parse.evaluate_function(modified_newtons_method_function3, x0)
        x = float(x0) - (float(f) * float(df))/((float(df)) ** 2 - float(f) * float(d2f))
        error = abs(float(x0) - float(x))
        while error > tolerance:
            x0 = x
            f = Parse.evaluate_function(modified_newtons_method_function1, str(x0))
            df = Parse.evaluate_function(modified_newtons_method_function2, str(x0))
            d2f = Parse.evaluate_function(modified_newtons_method_function3, str(x0))
            x = float(x0) - (float(f) * float(df)) / ((float(df)) ** 2 - float(f) * \
             float(d2f))
            error = abs(float(x0) - float(x))

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.modified_newtons_method_output_string.set(output)

    def steffensens_method_submit(self):

        #Get and parse data

        steffensens_method_function = \
         str(self.steffensens_method_function_entry_variable.get())
        steffensens_method_function = Parse.parse_function(steffensens_method_function)
        x0 = str(self.steffensens_method_initial_value_entry_variable.get())
        tolerance = self.steffensens_method_tolerance_variable.get()

        #Run Steffensen's Method

        x1 = float(x0) + float(Parse.evaluate_function(steffensens_method_function, x0))
        x = float(x0) - ((float(Parse.evaluate_function(steffensens_method_function,
         x0))) ** 2) / (float(Parse.evaluate_function(steffensens_method_function,
         str(x1))) - float(Parse.evaluate_function(steffensens_method_function, x0)))
        error = abs(float(x0) - float(x))
        while error > tolerance:
            x0 = x
            x1 = float(x0) + float(Parse.evaluate_function(steffensens_method_function,
             str(x0)))
            x = x0 - ((float(Parse.evaluate_function(steffensens_method_function,
             str(x0)))) ** 2) / (float(Parse.evaluate_function(steffensens_method_function,
             str(x1))) - float(Parse.evaluate_function(steffensens_method_function,
             str(x0))))
            error = abs(float(x0) - float(x))

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.steffensens_method_output_string.set(output)

    def mullers_method_submit(self):

        #Get and parse data

        mullers_method_function = str(self.mullers_method_function_entry_variable.get())
        mullers_method_function = Parse.parse_function(mullers_method_function)
        x0 = str(self.mullers_method_initial_value_entry_variable1.get())
        x1 = str(self.mullers_method_initial_value_entry_variable2.get())
        x2 = str(self.mullers_method_initial_value_entry_variable3.get())
        tolerance = self.mullers_method_tolerance_variable.get()

        #Run Muller's Method

        f0 = float(Parse.evaluate_function(mullers_method_function, x0))
        f1 = float(Parse.evaluate_function(mullers_method_function, x1))
        f2 = float(Parse.evaluate_function(mullers_method_function, x2))
        x0 = float(x0)
        x1 = float(x1)
        x2 = float(x2)
        a = ((x1 - x2) * (f0 - f2) - (x0 - x2) * (f1 - f2)) / ((x0 - x2) * (x1 - x2) * \
         (x0 - x1))
        b = ((x0 - x2) ** 2 * (f1 - f2) - (x1 - x2) ** 2 * (f0 - f2)) / ((x0 - x2) * (x1 - \
         x2) * (x0 - x1))
        c = f2
        x = x2 - (2 * c)/(b + (b/abs(b)) * m.sqrt(b**2 - 4 * a * c))
        error = abs(x2 - x)
        while error > tolerance:
            x0 = x1
            x1 = x2
            x2 = x
            f0 = float(Parse.evaluate_function(mullers_method_function, str(x0)))
            f1 = float(Parse.evaluate_function(mullers_method_function, str(x1)))
            f2 = float(Parse.evaluate_function(mullers_method_function, str(x2)))
            a = ((x1 - x2) * (f0 - f2) - (x0 - x2) * (f1 - f2)) / ((x0 - x2) * (x1 - x2) * \
             (x0 - x1))
            b = ((x0 - x2) ** 2 * (f1 - f2) - (x1 - x2) ** 2 * (f0 - f2)) / ((x0 - x2) * \
             (x1 - x2) * (x0 - x1))
            c = f2
            x = x2 - (2 * c)/(b + (b/abs(b)) * m.sqrt(b**2 - 4 * a * c))
            error = abs(x2 - x)

        #Create and set output.

        output = f"x = {str(x)}, Error: {str(error)}."
        self.mullers_method_output_string.set(output)

    def lagrange_interpolating_polynomials_submit(self):

        #Get and parse data

        x_list = str(self.lagrange_interpolating_polynomials_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.lagrange_interpolating_polynomials_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        x = \
         float(self.lagrange_interpolating_polynomials_interpolation_point_entry_variable.get())

        #Run Lagrange Interpolation

        y = 0
        i = 0
        while i < len(x_list):
            p = 1
            j = 0
            while j < len(x_list):
                if j != i:
                    p *= (x - x_list[j]) / (x_list[i] - x_list[j])
                j += 1
            y += y_list[i] * p
            i += 1

        #Create and set output.

        output = f"f({str(x)}) = {str(y)}."
        self.lagrange_interpolating_polynomials_output_string.set(output)

    def nevilles_method_submit(self):

        #Get and parse data

        x_list = str(self.nevilles_method_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.nevilles_method_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        x = float(self.nevilles_method_interpolation_point_entry_variable.get())

        #Run Neville's Method

        i = 1
        y = y_list
        while i < len(x_list):
            j = 0
            while j < len(x_list) - i:
                y[j] = ((x - x_list[i + j]) * y[j] + (x_list[j] - x) * y[j + 1]) / \
                 (x_list[j] - x_list[i + j])
                j += 1
            i += 1
        y = y[0]

        #Create and set output.

        output = f"f({str(x)}) = {str(y)}."
        self.nevilles_method_output_string.set(output)

    def newtons_divided_difference_method_submit(self):

        #Get and parse data

        x_list = str(self.newtons_divided_difference_method_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.newtons_divided_difference_method_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        x = \
         float(self.newtons_divided_difference_method_interpolation_point_entry_variable.get())

        #Run Newton's Divided Difference Method

        matrix = []
        i = 0
        while i < len(x_list):
            matrix.append([])
            j = 0
            while j < len(x_list):
                matrix[i].append(0)
                j += 1
            i += 1
        i = 0
        while i < len(y_list):
            matrix[i][0] = y_list[i]
            i += 1
        i = 1
        while i < len(x_list):
            j = 1
            while j <= i:
                matrix[i][j] = (matrix[i][j - 1] - matrix[i - 1][j - 1]) / (x_list[i] - \
                 x_list[i - j])
                j += 1
            i += 1
        i=0
        sum = 0
        while i < len(x_list):
            j = 0
            product = 1
            while i - j > 0:
                product *= x - x_list[i - j - 1]
                j += 1
            sum += matrix[i][i] * product
            i += 1

        #Create and set output.

        output = f"f({str(x)}) = {str(sum)}."
        self.newtons_divided_difference_method_output_string.set(output)

    def hermite_interpolation_submit(self):

        #Get and parse data

        x_list = str(self.hermite_interpolation_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.hermite_interpolation_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        y_prime_list = str(self.hermite_interpolation_y_prime_values_entry_variable.get())
        y_prime_list = Parse.string_to_list(y_prime_list)
        x = float(self.hermite_interpolation_interpolation_point_entry_variable.get())

        #Run Hermite Interpolation

        matrix = []
        i = 0
        while i < 2 * len(x_list):
            matrix.append([])
            j = 0
            while j < 2 * len(x_list):
                matrix[i].append(0)
                j += 1
            i += 1
        i = 0
        while i < 2 * len(y_list):
            matrix[i][0] = y_list[i // 2]
            i += 1
        i = 0
        x_list_temp = []
        while i < 2 * len(x_list):
            x_list_temp.append(x_list[i // 2])
            i += 1
        i = 1
        while i < 2 * len(x_list):
            j = 1
            while j <= i:
                if j == 1 and i % 2 == 1:
                    matrix[i][j] = y_prime_list[i // 2]
                else:
                    matrix[i][j] = (matrix[i][j - 1] - matrix[i - 1][j - 1]) / \
                     (x_list_temp[i] - x_list_temp[i - j])
                j += 1
            i += 1
        i = 0
        sum = 0
        while i < 2 * len(x_list):
            j = 0
            product = 1
            while i - j > 0:
                product *= x - x_list_temp[i - j - 1]
                j += 1
            sum += matrix[i][i] * product
            i += 1

        #Create and set output.

        output = f"f({str(x)}) = {str(sum)}."
        self.hermite_interpolation_output_string.set(output)

    def natural_cubic_spline_submit(self):

        #Get and parse data

        x_list = str(self.natural_cubic_spline_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.natural_cubic_spline_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        x = float(self.natural_cubic_spline_interpolation_point_entry_variable.get())

        #Evaluate natural cubic spline

        n = 4 * (len(x_list) - 1)    
        A = np.zeros((n, n))
        i = 0
        while i < n / 2:
            j = 0
            while j < n:
                if i % 2 == 0 and j > 2 * i - 1 and j < 2 * i + 4:
                    A[i][j] = x_list[int(i / 2)] ** (j % 4)
                if i % 2 == 1 and j > 2 * i - 3 and j < 2 * i + 2:
                    A[i][j] = x_list[int(i // 2 + 1)] ** (j % 4)
                j += 1
            i += 1
        i = int(n / 2)
        while i < (3 * n - 4) / 4:
            j = 0
            while j < n:
                if j > (4 * i) - (2 * n) - 1 and j < (4 * i) - (2 * n) + 8:
                    if j % 4 == 0:
                        A[i][j] = 0
                    elif i % 2 == 0 and j % 8 == 1:
                        A[i][j] = x_list[i - int(n / 2) + 1]
                    elif i % 2 == 0 and j % 8 == 2:
                        A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                    elif i % 2 == 0 and j % 8 == 3:
                        A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 0 and j % 8 == 5:
                        A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 0 and j % 8 == 6:
                        A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 0 and j % 8 == 7:
                        A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 1 and j % 8 == 5:
                        A[i][j] = x_list[i - int(n / 2) + 1]
                    elif i % 2 == 1 and j % 8 == 6:
                        A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                    elif i % 2 == 1 and j % 8 == 7:
                        A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 1 and j % 8 == 1:
                        A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 1 and j % 8 == 2:
                        A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                    else:
                        A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
                j += 1
            i += 1
        i = int((3 * n - 4) / 4)
        while i < n - 2:
            j = 0
            while j < n:
                if (j > (4 * i) - (3 * n) + 3 and j < (4 * i) - (3 * n) + 12):
                    if j % 4 == 0 or j % 4 == 1:
                        A[i][j] = 0
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 2:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 3:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 6:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 7:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 2:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 3:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 6:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 7:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 2:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 3:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 6:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 7:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 2:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 3:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 6:
                        A[i][j] = 2
                    else:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                j += 1
            i += 1
        i = n - 2
        while i < n:
            j = 0
            while j < n:
                if i % 2 == 0 and j < 4:
                    if j % 4 == 0 or j % 4 == 1:
                        A[i][j] = 0
                    elif j % 4 == 2:
                        A[i][j] = 2
                    else:
                        A[i][j] = 6 * x_list[0]
                if i % 2 == 1 and j > n - 5:
                    if j % 4 == 0 or j % 4 == 1:
                        A[i][j] = 0
                    elif j % 4 == 2:
                        A[i][j] = 2
                    else:
                        A[i][j] = 6 * x_list[-1]
                j += 1
            i += 1
        B = np.zeros(n)
        i = 0
        while i < n/2:
            if i % 2 == 0:
                B[i] = y_list[int(i / 2)]
            else:
                B[i] = y_list[int(i // 2) + 1]
            i += 1
        C = np.linalg.solve(A, B.transpose())
        x_list.sort(key = float)
        n = 0
        while x_list[n] < x:
            n += 1
        n = n - 1
        y = C[4 * n] + C[4 * n + 1] * x + C[4 * n + 2] * x ** 2 + C[4 * n + 3] * x ** 3

        #Create and set output.

        output = f"f({str(x)}) = {str(y)}."
        self.natural_cubic_spline_output_string.set(output)

    def clamped_cubic_spline_submit(self):

        #Get and parse data

        x_list = str(self.clamped_cubic_spline_x_values_entry_variable.get())
        x_list = Parse.string_to_list(x_list)
        y_list = str(self.clamped_cubic_spline_y_values_entry_variable.get())
        y_list = Parse.string_to_list(y_list)
        ic1 = float(self.clamped_cubic_spline_initial_condition_entry_variable1.get())
        ic2 = float(self.clamped_cubic_spline_initial_condition_entry_variable2.get())
        x = float(self.clamped_cubic_spline_interpolation_point_entry_variable.get())

        #Evaluate clamped cubic spline

        n = 4 * (len(x_list) - 1)    
        A = np.zeros((n, n))
        i = 0
        while i < n / 2:
            j = 0
            while j < n:
                if i % 2 == 0 and j > 2 * i - 1 and j < 2 * i + 4:
                    A[i][j] = x_list[int(i / 2)] ** (j % 4)
                if i % 2 == 1 and j > 2 * i - 3 and j < 2 * i + 2:
                    A[i][j] = x_list[int(i // 2 + 1)] ** (j % 4)
                j += 1
            i += 1
        i = int(n / 2)
        while i < (3 * n - 4) / 4:
            j = 0
            while j < n:
                if j > (4 * i) - (2 * n) - 1 and j < (4 * i) - (2 * n) + 8:
                    if j % 4 == 0:
                        A[i][j] = 0
                    elif i % 2 == 0 and j % 8 == 1:
                        A[i][j] = x_list[i - int(n / 2) + 1]
                    elif i % 2 == 0 and j % 8 == 2:
                        A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                    elif i % 2 == 0 and j % 8 == 3:
                        A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 0 and j % 8 == 5:
                        A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 0 and j % 8 == 6:
                        A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 0 and j % 8 == 7:
                        A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 1 and j % 8 == 5:
                        A[i][j] = x_list[i - int(n / 2) + 1]
                    elif i % 2 == 1 and j % 8 == 6:
                        A[i][j] = 2 * x_list[i - int(n / 2) + 1]
                    elif i % 2 == 1 and j % 8 == 7:
                        A[i][j] = 3 * (x_list[i - int(n / 2) + 1]) ** 2
                    elif i % 2 == 1 and j % 8 == 1:
                        A[i][j] = (-1) * (x_list[i - int(n / 2) + 1])
                    elif i % 2 == 1 and j % 8 == 2:
                        A[i][j] = (-2) * (x_list[i - int(n / 2) + 1])
                    else:
                        A[i][j] = (-3) * (x_list[i - int(n / 2) + 1]) ** 2
                j += 1
            i += 1
        i = int((3 * n - 4) / 4)
        while i < n - 2:
            j = 0
            while j < n:
                if (j > (4 * i) - (3 * n) + 3 and j < (4 * i) - (3 * n) + 12):
                    if j % 4 == 0 or j % 4 == 1:
                        A[i][j] = 0
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 2:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 3:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 6:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 0 and i % 2 == 0 and j % 8 == 7:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 2:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 3:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 6:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 0 and i % 2 == 1 and j % 8 == 7:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 2:
                        A[i][j] = 2
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 3:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 6:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 1 and i % 2 == 0 and j % 8 == 7:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 2:
                        A[i][j] = -2
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 3:
                        A[i][j] = (-6) * (x_list[i - int((3 * n - 4) / 4) + 1])
                    elif (n / 4) % 2 == 1 and i % 2 == 1 and j % 8 == 6:
                        A[i][j] = 2
                    else:
                        A[i][j] = 6 * x_list[i - int((3 * n - 4) / 4) + 1]
                j += 1
            i += 1
        i = n - 2
        while i < n:
            j = 0
            while j < n:
                if i % 2 == 0 and j < 4:
                    if j % 4 == 0:
                        A[i][j] = 0
                    elif j % 4 == 1:
                        A[i][j] = 1
                    elif j % 4 == 2:
                        A[i][j] = 2 * x_list[0]
                    else:
                        A[i][j] = 3 * x_list[0] ** 2
                if i % 2 == 1 and j > n - 5:
                    if j % 4 == 0:
                        A[i][j] = 0
                    elif j % 4 == 1:
                        A[i][j] = 1
                    elif j % 4 == 2:
                        A[i][j] = 2 * x_list[-1]
                    else:
                        A[i][j] = 3 * x_list[-1] ** 2
                j += 1
            i += 1
        B = np.zeros(n)
        i = 0
        while i < n/2:
            if i % 2 == 0:
                B[i] = y_list[int(i / 2)]
            else:
                B[i] = y_list[int(i // 2) + 1]
            i += 1
        B[-2] = ic1
        B[-1] = ic2
        C = np.linalg.solve(A, B.transpose())
        x_list.sort(key = float)
        n = 0
        while x_list[n] < x:
            n += 1
        n = n - 1
        y = C[4 * n] + C[4 * n + 1] * x + C[4 * n + 2] * x ** 2 + C[4 * n + 3] * x ** 3

        #Create and set output.

        output = f"f({str(x)}) = {str(y)}."
        self.clamped_cubic_spline_output_string.set(output)

    def forward_difference_submit(self):

        #Get and parse data

        forward_difference_function = \
         str(self.forward_difference_function_entry_variable.get())
        forward_difference_function = Parse.parse_function(forward_difference_function)
        x = float(self.forward_difference_x_value_entry_variable.get())
        h = float(self.forward_difference_h_value_entry_variable.get())

        #Evaluate forward difference

        df = (float(Parse.evaluate_function(forward_difference_function, str(x + h))) - \
         float(Parse.evaluate_function(forward_difference_function, str(x)))) / h
        d2f = (float(Parse.evaluate_function(forward_difference_function, str(x + 2 * \
         h))) - 2 * (float(Parse.evaluate_function(forward_difference_function, str(x + \
         h)))) + float(Parse.evaluate_function(forward_difference_function, str(x)))) / \
         (h ** 2)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.forward_difference_output_string.set(output)

    def backward_difference_submit(self):

        #Get and parse data

        backward_difference_function = \
         str(self.backward_difference_function_entry_variable.get())
        backward_difference_function = Parse.parse_function(backward_difference_function)
        x = float(self.backward_difference_x_value_entry_variable.get())
        h = float(self.backward_difference_h_value_entry_variable.get())

        #Evaluate backward difference

        df = (float(Parse.evaluate_function(backward_difference_function, str(x))) - \
         float(Parse.evaluate_function(backward_difference_function, str(x - h)))) / h
        d2f = (float(Parse.evaluate_function(backward_difference_function, str(x))) - 2 * \
         (float(Parse.evaluate_function(backward_difference_function, str(x - h)))) + \
         float(Parse.evaluate_function(backward_difference_function, str(x - 2 * h)))) / (h \
         ** 2)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.backward_difference_output_string.set(output)

    def central_difference_submit(self):

        #Get and parse data

        central_difference_function = \
         str(self.central_difference_function_entry_variable.get())
        central_difference_function = Parse.parse_function(central_difference_function)
        x = self.central_difference_x_value_entry_variable.get()
        h = self.central_difference_h_value_entry_variable.get()

        #Evaluate central difference

        df = (float(Parse.evaluate_function(central_difference_function, str(x + h))) - \
         float(Parse.evaluate_function(central_difference_function, str(x - h)))) / (2 * h)
        d2f = (float(Parse.evaluate_function(central_difference_function, str(x + h))) - \
         2 * (float(Parse.evaluate_function(central_difference_function, str(x)))) + \
         float(Parse.evaluate_function(central_difference_function, str(x - h)))) / (h ** 2)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.central_difference_output_string.set(output)

    def three_point_midpoint_formula_submit(self):

        #Get and parse data

        three_point_midpoint_formula_function = \
         str(self.three_point_midpoint_formula_function_entry_variable.get())
        three_point_midpoint_formula_function = \
         Parse.parse_function(three_point_midpoint_formula_function)
        x = float(self.three_point_midpoint_formula_x_value_entry_variable.get())
        h = float(self.three_point_midpoint_formula_h_value_entry_variable.get())

        #Evaluate three-point midpoint formula

        df = (float(Parse.evaluate_function(three_point_midpoint_formula_function, str(x + \
         h))) - float(Parse.evaluate_function(three_point_midpoint_formula_function,
         str(x - h)))) / (2 * h)
        d2f = (float(Parse.evaluate_function(three_point_midpoint_formula_function, str(x + \
         h))) - 2 * (float(Parse.evaluate_function(three_point_midpoint_formula_function,
         str(x)))) + float(Parse.evaluate_function(three_point_midpoint_formula_function,
         str(x - h)))) / (h ** 2)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.three_point_midpoint_formula_output_string.set(output)

    def three_point_endpoint_formula_submit(self):

        #Get and parse data

        three_point_endpoint_formula_function = \
         str(self.three_point_endpoint_formula_function_entry_variable.get())
        three_point_endpoint_formula_function = \
         Parse.parse_function(three_point_endpoint_formula_function)
        x = float(self.three_point_endpoint_formula_x_value_entry_variable.get())
        h = float(self.three_point_endpoint_formula_h_value_entry_variable.get())

        #Evaluate three-point endpoint formula

        df = (-3 * (float(Parse.evaluate_function(three_point_endpoint_formula_function,
         str(x)))) + 4 * \
         (float(Parse.evaluate_function(three_point_endpoint_formula_function, str(x + \
         h)))) - float(Parse.evaluate_function(three_point_endpoint_formula_function,
         str(x + 2 * h)))) / (2 * h)
        d2f = (float(Parse.evaluate_function(three_point_endpoint_formula_function,
         str(x))) - 2 * (float(Parse.evaluate_function(three_point_endpoint_formula_function,
         str(x + h)))) + float(Parse.evaluate_function(three_point_endpoint_formula_function,
         str(x + 2 * h))))/(h ** 2)

        #Create and set output.
    
        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.three_point_endpoint_formula_output_string.set(output)

    def five_point_midpoint_formula_submit(self):

        #Get and parse data

        five_point_midpoint_formula_function = \
         str(self.five_point_midpoint_formula_function_entry_variable.get())
        five_point_midpoint_formula_function = \
         Parse.parse_function(five_point_midpoint_formula_function)
        x = float(self.five_point_midpoint_formula_x_value_entry_variable.get())
        h = float(self.five_point_midpoint_formula_h_value_entry_variable.get())

        #Evaluate five-point midpoint formula

        df = (float(Parse.evaluate_function(five_point_midpoint_formula_function, str(x - \
         2 * h))) - 8 * (float(Parse.evaluate_function(five_point_midpoint_formula_function,
         str(x - h)))) + 8 * \
         (float(Parse.evaluate_function(five_point_midpoint_formula_function, str(x + \
         h)))) - float(Parse.evaluate_function(five_point_midpoint_formula_function,
         str(x + 2 * h)))) / (12 * h)
        d2f = ((-1) * (float(Parse.evaluate_function(five_point_midpoint_formula_function,
         str(x - 2 * h)))) + 16 * \
         (float(Parse.evaluate_function(five_point_midpoint_formula_function, str(x - \
         h)))) - 30 * (float(Parse.evaluate_function(five_point_midpoint_formula_function,
         str(x)))) + 16 * \
         (float(Parse.evaluate_function(five_point_midpoint_formula_function, str(x + \
         h)))) - float(Parse.evaluate_function(five_point_midpoint_formula_function,
         str(x + 2 * h))))/(12 * h ** 2)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}, f''({str(x)}) = {str(d2f)}."
        self.five_point_midpoint_formula_output_string.set(output)

    def five_point_endpoint_formula_submit(self):

        #Get and parse data

        five_point_endpoint_formula_function = \
         str(self.five_point_endpoint_formula_function_entry_variable.get())
        five_point_endpoint_formula_function = \
         Parse.parse_function(five_point_endpoint_formula_function)
        x = float(self.five_point_endpoint_formula_x_value_entry_variable.get())
        h = float(self.five_point_endpoint_formula_h_value_entry_variable.get())

        #Evaluate five-point endpoint formula

        df = ((-25) * (float(Parse.evaluate_function(five_point_endpoint_formula_function,
         str(x)))) + 48 * \
         (float(Parse.evaluate_function(five_point_endpoint_formula_function, str(x + \
         h)))) - 36 * (float(Parse.evaluate_function(five_point_endpoint_formula_function,
         str(x + 2 * h)))) + 16 * \
         (float(Parse.evaluate_function(five_point_endpoint_formula_function, str(x + 3 * \
         h)))) - 3 * float(Parse.evaluate_function(five_point_endpoint_formula_function,
         str(x + 4 * h)))) / (12 * h)

        #Create and set output.

        output = f"f'({str(x)}) = {str(df)}."
        self.five_point_endpoint_formula_output_string.set(output)

    def midpoint_rule_submit(self):

        #Get and parse data

        midpoint_rule_function = str(self.midpoint_rule_function_entry_variable.get())
        midpoint_rule_function = Parse.parse_function(midpoint_rule_function)
        a = float(self.midpoint_rule_interval_entry_variable1.get())
        b = float(self.midpoint_rule_interval_entry_variable2.get())
        n = self.midpoint_rule_n_entry_variable.get()

        #Run midpoint rule

        h = (b - a) / n
        x = a + (h / 2)
        s = 0
        while x < b:
            s += float(Parse.evaluate_function(midpoint_rule_function, str(x)))
            x += h
        s = h * s

        #Create and set output.

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(a)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(b)
        b = b.translate(superscripts)
        f = str(self.midpoint_rule_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(s)}."
        self.midpoint_rule_output_string.set(output)

    def trapezoid_rule_submit(self):

        #Get and parse data

        trapezoid_rule_function = str(self.trapezoid_rule_function_entry_variable.get())
        trapezoid_rule_function = Parse.parse_function(trapezoid_rule_function)
        a = float(self.trapezoid_rule_interval_entry_variable1.get())
        b = float(self.trapezoid_rule_interval_entry_variable2.get())
        n = self.trapezoid_rule_n_entry_variable.get()

        #Run trapezoid rule

        h = (b - a) / n
        x = a + h
        s = 0
        while x < b:
            s += 2 * float(Parse.evaluate_function(trapezoid_rule_function, str(x)))
            x += h
        s = s + float(Parse.evaluate_function(trapezoid_rule_function, str(a)))
        s = s + float(Parse.evaluate_function(trapezoid_rule_function, str(b)))
        s = (h / 2) * s

        #Create and set output.

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(a)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(b)
        b = b.translate(superscripts)
        f = str(self.trapezoid_rule_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(s)}."
        self.trapezoid_rule_output_string.set(output)

    def simpsons_rule_submit(self):

        #Get and parse data

        simpsons_rule_function = str(self.simpsons_rule_function_entry_variable.get())
        simpsons_rule_function = Parse.parse_function(simpsons_rule_function)
        a = float(self.simpsons_rule_interval_entry_variable1.get())
        b = float(self.simpsons_rule_interval_entry_variable2.get())
        n = self.simpsons_rule_n_entry_variable.get()

        #Run Simpson's rule

        h = (b - a) / n
        x = a + h
        i = 0
        s = 0
        while x < b:
            i += 1
            if i % 2 == 0:
                s += 2 * float(Parse.evaluate_function(simpsons_rule_function, str(x)))
            else:
                s += 4 * float(Parse.evaluate_function(simpsons_rule_function, str(x)))
            x += h
        s = s + float(Parse.evaluate_function(simpsons_rule_function, str(a)))
        s = s + float(Parse.evaluate_function(simpsons_rule_function, str(b)))
        s = (h / 3) * s

        #Create and set output.

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(a)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(b)
        b = b.translate(superscripts)
        f = str(self.simpsons_rule_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(s)}."
        self.simpsons_rule_output_string.set(output)

    def romberg_integration_submit(self):

        #Get and parse data

        romberg_integration_function = \
         str(self.romberg_integration_function_entry_variable.get())
        romberg_integration_function = Parse.parse_function(romberg_integration_function)
        a = float(self.romberg_integration_interval_entry_variable1.get())
        b = float(self.romberg_integration_interval_entry_variable2.get())
        tolerance = self.romberg_integration_tolerance_variable.get()

        #Run Romberg Integration

        matrix = [[0, 0], [0, 0]]
        matrix[0][0] = ((b - a) / 2) * \
         (float(Parse.evaluate_function(romberg_integration_function, str(a))) + \
         float(Parse.evaluate_function(romberg_integration_function, str(b))))
        matrix[1][0] = ((b - a) / 4) * \
         (float(Parse.evaluate_function(romberg_integration_function, str(a))) + 2 * \
         float(Parse.evaluate_function(romberg_integration_function, str((a + b) / 2))) + \
         float(Parse.evaluate_function(romberg_integration_function, str(b))))
        matrix[1][1] = matrix[1][0] + (1 / 3) * (matrix[1][0] - matrix[0][0])
        error = abs(matrix[1][1] - matrix[0][0])
        while error > tolerance:
            i = 0
            while i < len(matrix):
                matrix[i].append(0)
                i += 1
            matrix.append([])
            i = len(matrix) - 1
            j = 0
            while j < len(matrix):
                if j == 0:
                    n = 2 ** i
                    h = (b - a) / n
                    x = a + h
                    s = 0
                    while x < b:
                        s += 2 * float(Parse.evaluate_function(romberg_integration_function,
                         str(x)))
                        x += h
                    s = s + float(Parse.evaluate_function(romberg_integration_function,
                     str(a)))
                    s = s + float(Parse.evaluate_function(romberg_integration_function,
                     str(b)))
                    s = (h / 2) * s
                else:
                    s = matrix[i][j - 1] + (1 / ((4 ** j) - 1)) * (matrix[i][j - 1] - \
                     matrix[i - 1][j - 1])
                matrix[i].append(s)
                j += 1
            error = abs(matrix[i][i] - matrix[i - 1][i - 1])
        I = matrix[-1][-1]

        #Create and set output

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(a)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(b)
        b = b.translate(superscripts)
        f = str(self.romberg_integration_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(I)}, Error: {str(error)}."
        self.romberg_integration_output_string.set(output)

    def adaptive_quadrature_submit(self):

        #Get and parse data

        adaptive_quadrature_function = \
         str(self.adaptive_quadrature_function_entry_variable.get())
        adaptive_quadrature_function = Parse.parse_function(adaptive_quadrature_function)
        a = float(self.adaptive_quadrature_interval_entry_variable1.get())
        b = float(self.adaptive_quadrature_interval_entry_variable2.get())
        tolerance = self.adaptive_quadrature_tolerance_variable.get()

        #Run adaptive quadrature

        I = 0
        e = 0
        point_list = [a, b]
        while len(point_list) > 0:
            I1 = ((b - a) / 6) * (float(Parse.evaluate_function(adaptive_quadrature_function,
             str(point_list[0]))) + 4 * \
             (float(Parse.evaluate_function(adaptive_quadrature_function,
             str((point_list[0] + point_list[1]) / 2)))) + \
             float(Parse.evaluate_function(adaptive_quadrature_function,
             str(point_list[1]))))
            I2 = ((b - a) / 12) * \
             (float(Parse.evaluate_function(adaptive_quadrature_function,
             str(point_list[0]))) + 4 * \
             (float(Parse.evaluate_function(adaptive_quadrature_function,
             str((point_list[0] + point_list[1]) / 4)))) + \
             float(Parse.evaluate_function(adaptive_quadrature_function,
             str((point_list[0] + point_list[1]) / 2))))
            I3 = ((b - a) / 12) * \
             (float(Parse.evaluate_function(adaptive_quadrature_function,
             str((point_list[0] + point_list[1]) / 2))) + 4 * \
             (float(Parse.evaluate_function(adaptive_quadrature_function, str((3 * \
             (point_list[0] + point_list[1])) / 4)))) + \
             float(Parse.evaluate_function(adaptive_quadrature_function,
             str(point_list[1]))))
            error = (abs(I1 - I2 - I3)) / 15
            if error < tolerance:
                I += I2
                I += I3
                e += error
                point_list.pop(0)
                point_list.pop(0)
            else:
                tolerance = tolerance / 2
                m = (point_list[0] + point_list[1]) / 2
                first = point_list.pop(0)
                point_list.insert(0, m)
                point_list.insert(0, m)
                point_list.insert(0, first)

        #Create and set output

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(a)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(b)
        b = b.translate(superscripts)
        f = str(self.adaptive_quadrature_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(I)}, Error: {str(e)}."
        self.adaptive_quadrature_output_string.set(output)

    def gaussian_quadrature_submit(self):

        #Get and parse data

        gaussian_quadrature_function = \
         str(self.gaussian_quadrature_function_entry_variable.get())
        gaussian_quadrature_function = Parse.parse_function(gaussian_quadrature_function)
        n = self.gaussian_quadrature_n_entry_variable.get()

        #Run Gaussian Quadrature

        nodes = mp.taylor(lambda x: mp.legendre(n, x), 0, n)
        nodes.reverse()
        nodes = mp.polyroots(nodes)
        coefficients = mp.taylor(lambda x: mp.legendre(n - 1, x), 0, n - 1)
        coefficients.reverse()
        weights = []
        i = 0
        while i < len(nodes):
            weight = (2 * (1 - (nodes[i]) ** 2)) / ((n * (mp.polyval(coefficients,
             nodes[i]))) ** 2)
            weights.append(weight)
            i += 1
        i = 0
        I = 0
        while i < len(nodes):
            I += weights[i] * float(Parse.evaluate_function(gaussian_quadrature_function,
             str(nodes[i])))
            i += 1

        #Create and set output

        subscripts = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
        a = str(-1)
        a = a.translate(subscripts)
        superscripts = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")
        b = str(1)
        b = b.translate(superscripts)
        f = str(self.gaussian_quadrature_function_entry_variable.get())
        output = f"∫{a}{b}{f}dx = {str(I)}."
        self.gaussian_quadrature_output_string.set(output)

    def eulers_method_submit(self):

        #Get and parse data

        eulers_method_function = str(self.eulers_method_function_entry_variable.get())
        eulers_method_function = Parse.multivariate_parse(eulers_method_function)
        a = float(self.eulers_method_interval_entry_variable1.get())
        b = float(self.eulers_method_interval_entry_variable2.get())
        ic = float(self.eulers_method_initial_condition_entry_variable.get())
        n = self.eulers_method_n_entry_variable.get()

        #Run Euler's Method

        h = (b - a) / n
        x = a
        y = ic
        while x < b:
            df = float(Parse.evaluate_multivariate(eulers_method_function, str(x), str(y)))
            y = y + h * df
            x = x + h

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.eulers_method_output_string.set(output)
        
    def midpoint_method_submit(self):

        #Get and parse data

        midpoint_method_function = str(self.midpoint_method_function_entry_variable.get())
        midpoint_method_function = Parse.multivariate_parse(midpoint_method_function)
        a = float(self.midpoint_method_interval_entry_variable1.get())
        b = float(self.midpoint_method_interval_entry_variable2.get())
        ic = float(self.midpoint_method_initial_condition_entry_variable.get())
        n = self.midpoint_method_n_entry_variable.get()

        #Run Midpoint Method

        h = (b - a) / n
        x = a
        y = ic
        while x < b - h:
            df = float(Parse.evaluate_multivariate(midpoint_method_function, str(x), str(y)))
            y = y + h * float(Parse.evaluate_multivariate(midpoint_method_function,
             str(x + (h / 2)), str(y + ((h / 2) * df))))
            x = x + h

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.midpoint_method_output_string.set(output)

    def modified_euler_method_submit(self):

        #Get and parse data

        modified_euler_method_function = \
         str(self.modified_euler_method_function_entry_variable.get())
        modified_euler_method_function = \
         Parse.multivariate_parse(modified_euler_method_function)
        a = float(self.modified_euler_method_interval_entry_variable1.get())
        b = float(self.modified_euler_method_interval_entry_variable2.get())
        ic = float(self.modified_euler_method_initial_condition_entry_variable.get())
        n = self.modified_euler_method_n_entry_variable.get()

        #Run Modified Euler Method

        h = (b - a) / n
        x = a
        y = ic
        while x < b - h:
            df = float(Parse.evaluate_multivariate(modified_euler_method_function, str(x),
             str(y)))
            y = y + (h / 2) * float(df + \
             float(Parse.evaluate_multivariate(modified_euler_method_function, str(x + h),
             str(y + (h * df)))))
            x = x + h

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.modified_euler_method_output_string.set(output)

    def heuns_method_submit(self):

        #Get and parse data

        heuns_method_function = str(self.heuns_method_function_entry_variable.get())
        heuns_method_function = Parse.multivariate_parse(heuns_method_function)
        a = float(self.heuns_method_interval_entry_variable1.get())
        b = float(self.heuns_method_interval_entry_variable2.get())
        ic = float(self.heuns_method_initial_condition_entry_variable.get())
        n = self.heuns_method_n_entry_variable.get()

        #Run Heun's Method

        h = (b - a) / n
        x = a
        y = ic
        while x < b - h:
            df = float(Parse.evaluate_multivariate(heuns_method_function, str(x), str(y)))
            y = y + (h / 4) * float(df + 3 * \
             float(Parse.evaluate_multivariate(heuns_method_function, str(x + ((2 * h) / 3)),
             str(y + ((2 * h) / 3) * float(Parse.evaluate_multivariate(heuns_method_function,
             str(x + (h / 3)), str(y + (h / 3) * df)))))))
            x = x + h

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.heuns_method_output_string.set(output)

    def runge_kutta_method_submit(self):

        #Get and parse data

        runge_kutta_method_function = \
         str(self.runge_kutta_method_function_entry_variable.get())
        runge_kutta_method_function = Parse.multivariate_parse(runge_kutta_method_function)
        a = float(self.runge_kutta_method_interval_entry_variable1.get())
        b = float(self.runge_kutta_method_interval_entry_variable2.get())
        ic = float(self.runge_kutta_method_initial_condition_entry_variable.get())
        n = self.runge_kutta_method_n_entry_variable.get()

        #Run Runge Kutta Method

        h = (b - a) / n
        x = a
        y = ic
        while x < b - h:
            k1 = h * float(Parse.evaluate_multivariate(runge_kutta_method_function, str(x),
             str(y)))
            k2 = h * float(Parse.evaluate_multivariate(runge_kutta_method_function, str(x + \
             (h / 2)), str(y + (k1 / 2))))
            k3 = h * float(Parse.evaluate_multivariate(runge_kutta_method_function, str(x + \
             (h / 2)), str(y + (k2 / 2))))
            k4 = h * float(Parse.evaluate_multivariate(runge_kutta_method_function, str(x + \
             h), str(y + k3)))
            y = y + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
            x = x + h

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.runge_kutta_method_output_string.set(output)

    def adams_bashforth_technique_submit(self):

        #Get and parse data

        adams_bashforth_technique_function = \
         str(self.adams_bashforth_technique_function_entry_variable.get())
        adams_bashforth_technique_function = \
         Parse.multivariate_parse(adams_bashforth_technique_function)
        a = float(self.adams_bashforth_technique_interval_entry_variable1.get())
        b = float(self.adams_bashforth_technique_interval_entry_variable2.get())
        ic = float(self.adams_bashforth_technique_initial_condition_entry_variable.get())
        n = self.adams_bashforth_technique_n_entry_variable.get()

        #Run Adams-Bashforth Technique

        h = (b - a) / n
        x = a
        y = ic
        i = 0
        y_list = []
        while i < n + 1:
            if i == 0:
                y_list.append(y)
            elif i < 4:
                k1 = h * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x), str(y_list[-1])))
                k2 = h * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x + (h / 2)), str(y_list[-1] + (k1 / 2))))
                k3 = h * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x + (h / 2)), str(y_list[-1] + (k2 / 2))))
                k4 = h * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x + h), str(y_list[-1] + k3)))
                y = y_list[-1] + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
                y_list.append(y)
            else:
                y = y_list[3] + (h / 24) * ((55 * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x), str(y_list[3])))) - (59 * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x - h), str(y_list[2])))) + (37 * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x - (2 * h)), str(y_list[1])))) - (9 * \
                 float(Parse.evaluate_multivariate(adams_bashforth_technique_function,
                 str(x - (3 * h)), str(y_list[0])))))
                y_list.pop(0)
                y_list.append(y)
            if i > 0:
                x = x + h
            i += 1
        y = y_list[-1]

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.adams_bashforth_technique_output_string.set(output)

    def milues_method_submit(self):

        #Get and parse data

        milues_method_function = str(self.milues_method_function_entry_variable.get())
        milues_method_function = Parse.multivariate_parse(milues_method_function)
        a = float(self.milues_method_interval_entry_variable1.get())
        b = float(self.milues_method_interval_entry_variable2.get())
        ic = float(self.milues_method_initial_condition_entry_variable.get())
        n = self.milues_method_n_entry_variable.get()

        #Run Milues Method

        h = (b - a) / n
        x = a
        y = ic
        i = 0
        y_list = []
        while i < n + 1:
            if i == 0:
                y_list.append(y)
            elif i < 4:
                k1 = h * float(Parse.evaluate_multivariate(milues_method_function, str(x),
                 str(y_list[-1])))
                k2 = h * float(Parse.evaluate_multivariate(milues_method_function, str(x + \
                 (h / 2)), str(y_list[-1] + (k1 / 2))))
                k3 = h * float(Parse.evaluate_multivariate(milues_method_function, str(x + \
                 (h / 2)), str(y_list[-1] + (k2 / 2))))
                k4 = h * float(Parse.evaluate_multivariate(milues_method_function, str(x + \
                 h), str(y_list[-1] + k3)))
                y = y_list[-1] + (1 / 6) * (k1 + (2 * k2) + (2 * k3) + k4)
                y_list.append(y)
            else:
                y = y_list[0] + ((4 * h) / 3) * ((2 * \
                 float(Parse.evaluate_multivariate(milues_method_function, str(x),
                 str(y_list[3])))) - \
                 (float(Parse.evaluate_multivariate(milues_method_function, str(x - h),
                 str(y_list[2])))) + (2 * \
                 float(Parse.evaluate_multivariate(milues_method_function, str(x - (2 * h)),
                 str(y_list[1])))))
                y_list.pop(0)
                y_list.append(y)
            if i > 0:
                x = x + h
            i += 1
        y = y_list[-1]

        #Create and set output

        output = f"y({str(b)}) = {str(y)}."
        self.milues_method_output_string.set(output)

    def help1(self):

        #Set up window

        self.window1 = tk.Toplevel()
        self.window1.title("Help")
        self.window1.geometry("1250x500")

        #Widgets

        Widgets.title_label(self.window1, "Help", 24)
        help_frame = Widgets.scrollbar(self.window1)
        label1 = '''Below are instructions on how to use this app, including installation,
         supported functions, and implemented numerical methods.'''
        Widgets.paragraph_label(help_frame, label1, 14)
        Widgets.title_label(help_frame, "Required Packages", 20)
        Widgets.paragraph_label(help_frame, "math", 14)
        Widgets.paragraph_label(help_frame, "mpmath", 14)
        Widgets.paragraph_label(help_frame, "numpy", 14)
        Widgets.paragraph_label(help_frame, "tkinter", 14)
        Widgets.paragraph_label(help_frame, "ttk", 14)
        label2 = '''All of these packages can be installed via the terminal in Jupyter
         Notebook using the pip install command.'''
        Widgets.paragraph_label(help_frame, label2, 14)
        Widgets.title_label(help_frame, "Supported Constants and Functions", 20)
        label3 = '''Many of the numerical methods implemented in this app require the user
         to input one or more functions, which are interpreted as Python strings.

         The supported mathematical operations are addition (+), subtraction (-),
         multiplication (*), division (/), and exponentiation (^). The app follows the order
         of operations taught in grade school.

         This app implements various mathematical constants and functions. See the Constants
         and Functions sections for more information on which constants and functions are
         supported and how they are implemented. When using a mathematical function be sure
         to enclose the argument(s) in parenthesis. For example, use sin(x), not sinx.'''
        Widgets.paragraph_label(help_frame, label3, 14)
        Widgets.title_label(help_frame, "Constants", 18)
        Widgets.treeview(help_frame, "Constant", [("Pi", "pi", "math.pi"), ("Euler's Number",
         "e^(1)", "math.exp(1)"), ("Golden Ratio", "phi", "mpmath.phi"),
         ("Euler-Mascheroni Constant", "gamma", "mpmath.euler"),
         ("Glaisher-Kinkelin Constant", "A", "mpmath.glaisher"), ("Catalan's Constant",
         "(Phi(-1,2,1/2))/4", "(mpmath.lerchphi(-1,2,1/2))/4"), ("Infinity", "infinity",
         "math.inf")])
        Widgets.title_label(help_frame, "Functions", 18)
        Widgets.treeview(help_frame, "Function", [("Absolute Value", "|x|", "math.fabs(x)"),
         ("Exponential", "e^(x)", "math.exp(x)"), ("Natural Logarithm", "ln(x)",
         "math.log(x)"), ("Common Logarithm", "log(x)", "math.log(x,10)"),
         ("General Logarithm", "log(x,a)", "math.log(x,a)"), ("Square Root", "sqrt(x)",
         "math.sqrt(x)"), ("Cube Root", "cbrt(x)", "mpmath.cbrt(x)"), ("N-th Root",
         "root(x,n)", "mpmath.root(x,n)"), ("Lambert W Function", "W(x)",
         "mpmath.lambertw(x)"), ("Sine", "sin(x)", "math.sin(x)"), ("Inverse Sine",
         "arcsin(x)", "math.asin(x)"), ("Hyperbolic Sine", "sinh(x)", "math.sinh(x)"),
         ("Inverse Hyperbolic Sine", "arsinh(x)", "math.asinh(x)"), ("Cosine", "cos(x)",
         "math.cos(x)"), ("Inverse Cosine", "arccos(x)", "math.acos(x)"),
         ("Hyperbolic Cosine", "cosh(x)", "math.cosh(x)"), ("Inverse Hyperbolic Cosine",
         "arcosh(x)", "math.acosh(x)"), ("Tangent", "tan(x)", "math.tan(x)"),
         ("Inverse Tangent", "arctan(x)", "math.atan(x)"), ("Hyperbolic Tangent", "tanh(x)",
         "math.tanh(x)"), ("Inverse Hyperbolic Tangent", "artanh(x)", "math.atanh(x)"),
         ("Secant", "sec(x)", "mpmath.sec(x)"), ("Inverse Secant", "arcsec(x)",
         "mpmath.asec(x)"), ("Hyperbolic Secant", "sech(x)", "mpmath.sech(x)"),
         ("Inverse Hyperbolic Secant", "arsech(x)", "mpmath.asech(x)"), ("Cosecant",
         "csc(x)", "mpmath.csc(x)"), ("Inverse Cosecant", "arccsc(x)", "mpmath.acsc(x)"),
         ("Hyperbolic Cosecant", "csch(x)", "mpmath.csch(x)"),
         ("Inverse Hyperbolic Cosecant", "arcsch(x)", "mpmath.acsch(x)"), ("Cotangent",
         "cot(x)", "mpmath.cot(x)"), ("Inverse Cotangent", "arccot(x)", "mpmath.acot(x)"),
         ("Hyperbolic Cotangent", "coth(x)", "mpmath.coth(x)"),
         ("Inverse Hyperbolic Cotangent", "arcoth(x)", "mpmath.acoth(x)"), ("Gamma Function",
         "Gamma(x)", "math.gamma(x)"), ("Generalized Incomplete Gamma Function",
         "Gamma(x,a,b)", "mpmath.gammainc(x,a,b)"), ("Digamma Function", "psi(x)",
         "mpmath.digamma(x)"), ("Polygamma Function", "psi(n,x)", "mpmath.psi(n,x)"),
         ("Barnes G Function", "G(x)", "mpmath.barnesg(x)"),
         ("Exponential Integral Function", "Ei(x)", "mpmath.ei(x)"),
         ("Generalized Exponential Integral Function", "En(n,x)", "mpmath.expint(n,x)"),
         ("Logarithmic Integral Function", "li(x)", "mpmath.li(x)"),
         ("Cosine Integral Function", "Ci(x)", "mpmath.ci(x)"),
         ("Hyperbolic Cosine Integral Function", "Chi(x)", "mpmath.chi(x)"),
         ("Sine Integral Function", "Si(x)", "mpmath.si(x)"),
         ("Hyperbolic Sine Integral Function", "Shi(x)", "mpmath.shi(x)"), ("Error Function",
         "erf(x)", "math.erf(x)"), ("Complementary Error Function", "erfc(x)",
         "math.erfc(x)"), ("Imaginary Error Function", "erfi(x)", "mpmath.erfi(x)"),
         ("Fresnel Sine Integral Function", "S(x)", "mpmath.fresnels(x)"),
         ("Fresnel Cosine Integral Function", "C(x)", "mpmath.fresnelc(x)"),
         ("Bessel Function of the First Kind", "J(n,x)", "mpmath.besselj(n,x)"),
         ("Bessel Function of the Second Kind", "Y(n,x)", "mpmath.bessely(n,x)"),
         ("Modified Bessel Function of the First Kind", "I(n,x)", "mpmath.besseli(n,x)"),
         ("Modified Bessel Function of the Second Kind", "K(n,x)", "mpmath.besselk(n,x)"),
         ("Struve Function", "H(n,x)", "mpmath.struveh(n,x)"), ("Modified Struve Function",
         "L(n,x)", "mpmath.struvel(n,x)"), ("Weber Function", "E(v,x)",
         "mpmath.webere(v,x)"), ("Lommel Function", "s(u,v,x)", "mpmath.lommels1(u,v,x)"),
         ("Airy Ai Function", "Ai(x)", "mpmath.airyai(x)"), ("Airy Bi Function", "Bi(x)",
         "mpmath.airybi(x)"), ("Scorer Gi Function", "Gi(x)", "mpmath.scorergi(x)"),
         ("Scorer Hi Function", "Hi(x)", "mpmath.scorerhi(x)"),
         ("Generalized Hypergeometric Function", "pFq([a_1,...,a_p],[b_1,...,b_q],x)",
         "mpmath.hyper([a_1,...,a_p],[b_1,...,b_q],x)"),
         ("Bilateral Hypergeometric Function", "AHB([a_1,...,a_A],[b_1,...,b_B],x)",
         "mpmath.bihyper([a_1,...,a_A],[b_1,...,b_B],x)"), ("Dirichlet Eta Function",
         "eta(x)", "mpmath.altzeta(x)"), ("Jacobi Theta Function", "theta(n,x,q)",
         "mpmath.jtheta(n,x,q)"), ("Riemann Zeta Function", "zeta(x)", "mpmath.zeta(x)"),
         ("Hurwitz Zeta Function", "zeta(x,a)", "mpmath.zeta(x,a)"),
         ("Lerch Transcendent Function", "Phi(x,s,a)", "mpmath.lerchphi(x,s,a)"),
         ("Polylogarithm", "Li(s,x)", "mpmath.polylog(s,x)")])
        Widgets.title_label(help_frame, "Numerical Algorithms", 20)
        label4 = '''This app implements many of the most commonly used numerical methods. It
         includes a root finder, an interpolation calculator, a numerical differentiator, a
         numerical integrator, and a numerical solver. See the respective sections for
         instructions on how to use these tools.'''
        Widgets.paragraph_label(help_frame, label4, 14)
        Widgets.title_label(help_frame, "Root Finder", 18)
        label5 = '''Root finding is the process of finding solutions, also called roots or
         zeros, of the equation f(x) = 0. A similar procedure is determining fixed points of
         a function, or values of x which satisfy the equation f(x) = x. Some equations,
         such as x^2 - 7x + 12 = 0, can be easily solved analytically. However, equations
         like tanx = x cannot be solved analytically. In these cases, numerical techniques
         must be used to approximate a solution. The techniques include bracketing methods
         and iterative methods.

         Bracketing methods start with a closed interval [a,b] such that the function is
         ontinuous over the interval and f(a) and f(b) have opposite signs. By the
         Intermediate Value Theorem, it is then known that a root exists between a and b.
         Bracketing methods shorten the size of this interval while still maintaining the
         conditions of the Intermediate Value Theorem, and proceeds until the interval is
         small enough to approximate the root within a given tolerance. Bracketing methods
         implemented in this app include the bisection method and the method of false
         position.

         Iterative methods begin with one or more initial approximations. These data are
         then used to derive a better approximation which is then used to derive an even
         better approximation. The process continues until aroot is found within a given
         tolerance. Fixed-point iteration, Newton's method, the secant method, modified
         Newton's method, Steffensen's method, and Muller's method are the iterative methods
         implemented in this app.

         See the respective sections for instructions on how to use the various root finding
         methods.'''
        Widgets.paragraph_label(help_frame, label5, 14)
        Widgets.title_label(help_frame, "Bisection Method", 16)
        label6 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Left endpoint of interval. The function value at f(a) must have a
         different sign than at f(b).

         b (float): Right endpoint of interval. The function value at f(b) must have a
         different sign than at f(a).

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via the bisection
         method, or message that the conditions of the Intermediate Value Theorem are not
         met.'''
        Widgets.paragraph_label(help_frame, label6, 14)
        Widgets.title_label(help_frame, "Fixed Point Iteration", 16)
        label7 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x_0 (float): Initial approximation for fixed point.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = x and estimated error derived via fixed-point
         interation.'''
        Widgets.paragraph_label(help_frame, label7, 14)
        Widgets.title_label(help_frame, "Newton's Method", 16)
        label8 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         Derivative (string): Derivative of given function with respect to x. See the
         Supported Constants and Functions section.

         x_0 (float): Initial approximation for root.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via Newton's method.'''
        Widgets.paragraph_label(help_frame, label8, 14)
        Widgets.title_label(help_frame, "Secant Method", 16)
        label9 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions sectipn.

         x_0 (float): First initial approximation for root.

         x_1 (float): Second initial approximation for root.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via the secant
         method.'''
        Widgets.paragraph_label(help_frame, label9, 14)
        Widgets.title_label(help_frame, "Method of False Position", 16)
        label10 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Left endpoint of interval. The function value at f(a) must have a
         different sign than at f(b).

         b (float): Right endpoint of interval. The function value at f(b) must have a
         different sign than at f(a).

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via the method of
         false position, or message that the conditions of the Intermediate Value Theorem
         are not met.'''
        Widgets.paragraph_label(help_frame, label10, 14)
        Widgets.title_label(help_frame, "Modified Newton's Method", 16)
        label11 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         Derivative (string): First derivative of given function with respect to x. See the
         Supported Constants and Functions section.

         Second Derivative (string): Second derivative of given function with respect to x.
         See the Supported Constants and Functions section.

         x_0 (float): Initial approximation for root.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via modified Newton's
         method.'''
        Widgets.paragraph_label(help_frame, label11, 14)
        Widgets.title_label(help_frame, "Steffensen's Method", 16)
        label12 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x_0 (float): Initial approximation for fixed point.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = x and estimated error derived via Steffensen's
         method.'''
        Widgets.paragraph_label(help_frame, label12, 14)
        Widgets.title_label(help_frame, "Müller’s Method", 16)
        label13 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x_0 (float): First initial approximation for root.

         x_1 (float): Second initial approximation for root.

         x_2 (float): Third initial approximation for root.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of f(x) = 0 and estimated error derived via Muller's method.'''
        Widgets.paragraph_label(help_frame, label13, 14)
        Widgets.title_label(help_frame, "Interpolation Calculator", 18)
        label14 = '''Given any set of data there will be gaps in the data. As a result,
         there are often missing data points for which information needs to be determined.
         The process of estimating these missing data points based on given data is called
         interpolation. Included in the techniques of interpolation are polynomial
         interpolation and spline interpolation.

         Polynomial interpolation is based on the idea that given n data points there is a
         unique polynomial of maximum degree n - 1 that agrees with the data. Polynomial
         interpolation determines this polynomial and evaluates it at the point(s) to be
         determined. Four methods of polynomial interpolation implemented in this app
         include the use of Lagrange interpolating polynomials, Neville's method, Newton's
         divided difference method, and Hermite Interpolation.

         Given n data points, spline interpolation fits a polynomial to the data in each of
         the n - 1 subintervals. This app uses cubic splines, meaning each subinterval
         approximates the data using a cubic polynomial. These polynomials agree at each of
         the data points and have continuous first and second order derivatives at the
         interior data points. In order to obtain a unique solution, conditions must be
         given at the two boundary points which are called boundary conditions. For a
         natural cubic spline, the second derivative is equal to zero at the boundary points.
         In a clamped cubic spline, the value of the first derivative is specified at each
         boundary point. Natural cubic splines and clamped cubic splines are implemented in
         this app.

         See the respective sections for instructions on how to use the various methods of
         interpolation.'''
        Widgets.paragraph_label(help_frame, label14, 14)
        Widgets.title_label(help_frame, "Lagrange Interpolating Polynomials", 16)
        label15 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via Lagrange
         interpolating polynomial.'''
        Widgets.paragraph_label(help_frame, label15, 14)
        Widgets.title_label(help_frame, "Neville's Method", 16)
        label16 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via Neville's
         method.'''
        Widgets.paragraph_label(help_frame, label16, 14)
        Widgets.title_label(help_frame, "Newton's Divided Difference Method", 16)
        label17 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via Newton's divided
         difference method.'''
        Widgets.paragraph_label(help_frame, label17, 14)
        Widgets.title_label(help_frame, "Hermite Interpolation", 16)
        label18 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         y'-values (string): Derivative values at the x-coordinates of data set entered as
         floats separated by commas.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via Hermite
         interpolation.'''
        Widgets.paragraph_label(help_frame, label18, 14)
        Widgets.title_label(help_frame, "Natural Cubic Spline", 16)
        label19 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via natural cubic
         spline.'''
        Widgets.paragraph_label(help_frame, label19, 14)
        Widgets.title_label(help_frame, "Clamped Cubic Spline", 16)
        label20 = '''Inputs:

         x-values (string): x-coordinates of data set entered as floats separated by commas.

         y-values (string): y-coordinates of data set entered as floats separated by commas.

         f'(a) (float): Value of first derivative at the minimum x-value.

         f'(b) (float): Value of first derivative at the maximum x-value.

         x (float): Interpolation point.

         Output:

         Estimated value of y-coordinate at interpolation point derived via clamped cubic
         spline.'''
        Widgets.paragraph_label(help_frame, label20, 14)
        Widgets.title_label(help_frame, "Numerical Differentiator", 18)
        label21 = '''Differentiation, along with integration, is one of the two major
         operators taught in calculus. The chain rule, product rule, and quotient rule are
         among the techniques used to compute derivatives analytically. Due to the wide
         applicability of the techniques the set of elementary functions is closed under
         differentiation, that is every elementary function has an elementary derivative.
         Even though any derivative can be found analytically, the more complicated the
         function becomes the less practical it becomes to compute the derivative by hand.
         In these cases, it is more practical to compute the derivative numerically. The
         finite difference methods, such as those included in this app, are the most common
         methods used to compute derivatives numerically.

         Finite difference methods use a given step size and the given function evaluated
         at one or more points to approximate the derivative at a given point. The smaller
         the step size and the more points used in the evaluation, the more accurate the
         approximation. The definition of the derivative in terms of the difference quotient
         is used to derive the forward, backward, and central difference formulas. Higher
         order methods, including the three and five point formulas can be derived using
         Taylor series. The forward difference method, the backward difference method, the
         central difference method, the three-point midpoint formula, the three-point
         endpoint formula, the five-point midpoint formula, and the five-point endpoint
         formula are the finite difference methods implemented in this app.

         See the respective sections for instructions on how to use the various methods of
         numerical differentiation.'''
        Widgets.paragraph_label(help_frame, label21, 14)
        Widgets.title_label(help_frame, "Forward Difference", 16)
        label22 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         forward difference method.'''
        Widgets.paragraph_label(help_frame, label22, 14)
        Widgets.title_label(help_frame, "Backward Difference", 16)
        label23 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         backward difference method.'''
        Widgets.paragraph_label(help_frame, label23, 14)
        Widgets.title_label(help_frame, "Central Difference", 16)
        label24 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         central difference method.'''
        Widgets.paragraph_label(help_frame, label24, 14)
        Widgets.title_label(help_frame, "Three-Point Midpoint Formula", 16)
        label25 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         three-point midpoint formula.'''
        Widgets.paragraph_label(help_frame, label25, 14)
        Widgets.title_label(help_frame, "Three-Point Endpoint Formula", 16)
        label26 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         three-point endpoint formula.'''
        Widgets.paragraph_label(help_frame, label26, 14)
        Widgets.title_label(help_frame, "Five-Point Midpoint Formula", 16)
        label27 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximate values of the first and second derivatives at x derived via the
         five-point midpoint formula.'''
        Widgets.paragraph_label(help_frame, label27, 14)
        Widgets.title_label(help_frame, "Five-Point Endpoint Formula", 16)
        label28 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         x (float): Point where the derivative is approximated.

         h (float): Step size. The smaller the step size, the more accurate the
         approximation will be.

         Output:

         The approximated value of the first derivative at x derived via the five-point
         endpoint formula.'''
        Widgets.paragraph_label(help_frame, label28, 14)
        Widgets.title_label(help_frame, "Numerical Integrator", 18)
        label29 = '''Integration, along with differentiation, is one of the two major
         operators taught in calculus. Substitution, integration by parts, and partial
         fraction decomposition are just a few of the many techniques used to compute
         integrals. Techniques such as substitution and integration by parts are general
         techniques and can be applied in many circumstances. However, techniques such as
         partial fraction decomposition and trigonometric substitution are more specialized
         and apply in specific cases. As a result, the set of elementary functions is not
         closed under integration, that is not every elementary function has an elementary
         antiderivative. In cases where an integrand does not have an elementary
         antiderivative, more sophisticated techniques such as Taylor series,
         differentiation under the integral sign, and contour integration can on occasion be
         used to obtain a solution analytically. For these reasons, integration is often a
         very difficult process, if even possible, and it is often more practical to use
         numerical methods to approximate integrals. These methods include the Newton-Cotes 
         formulas, extrapolation, and quadrature. All of these techniques make use of the
         fact that the integral is just the signed area under a curve.

         The Newton-Cotes formulas break up the interval of integration into subintervals,
         and then uses polynomial interpolation and equally spaced points to approximate the
         area over each subinterval. It then sums up all these areas to get the total area.
         The more subintervals, the more accurate the approximation. The Newton-Cotes
         formulas implemented in this app include the midpoint rule, the trapezoid rule, and
         Simpson's rule.

         The combination of a Newton-Cotes formula along with the technique of extrapolation
         is called Romberg integration. An advantage of this method is that it can be used
         to approximate an integral within a given tolerance. Romberg integration is
         implemented in this app.

         Quadrature is similar to the use of the Newton-Cotes formulas, but instead of using
         predetermined weights and having the user specify the number of subintervals, it
         chooses the grid spacing and the weights at each point such that the integral is
         approximated as accurately as possible. Adaptive quadrature and Gaussian
         quadrature are implemented in this app.

         See the respective sections for instructions on how to use the various methods of
         numerical integration.'''
        Widgets.paragraph_label(help_frame, label29, 14)
        Widgets.title_label(help_frame, "Midpoint Rule", 16)
        label30 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Lower limit of integration.

         b (float): Upper limit of integration.

         n (integer): Number of subintervals. The larger n is, the better the approximation.

         Output:

         Approximate solution of the integral derived via the midpoint rule.'''
        Widgets.paragraph_label(help_frame, label30, 14)
        Widgets.title_label(help_frame, "Trapezoid Rule", 16)
        label31 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions.

         a (float): Lower limit of integration.

         b (float): Upper limit of integration.

         n (integer): Number of subintervals. The larger n is, the better the approximation.

         Output:

         Approximate solution of the integral derived via the trapezoid rule.'''
        Widgets.paragraph_label(help_frame, label31, 14)
        Widgets.title_label(help_frame, "Simpson's Rule", 16)
        label32 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Lower limit of integration.

         b (float): Upper limit of integration.

         n (integer): Number of subintervals. The larger n is, the better the approximation.
         The value of n must be even.

         Output:

         Approximate solution of the integral derived via Simpson's rule.'''
        Widgets.paragraph_label(help_frame, label32, 14)
        Widgets.title_label(help_frame, "Romberg Integration", 16)
        label33 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Lower limit of integration.

         b (float): Upper limit of integration.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of the integral and estimated error derived via Romberg
         integration.

         Note: This algorithm uses the trapezoid rule as the base method.'''
        Widgets.paragraph_label(help_frame, label33, 14)
        Widgets.title_label(help_frame, "Adaptive Quadrature", 16)
        label34 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         a (float): Lower limit of integration.

         b (float): Upper limit of integration.

         Tolerance (float): Acceptable error for approximation.

         Output:

         Approximate solution of the integral and estimated error derived via adaptive
         quadrature.

         Note: This algorithm uses Simpson's rule as the base method.'''
        Widgets.paragraph_label(help_frame, label34, 14)
        Widgets.title_label(help_frame, "Gaussian Quadrature", 16)
        label35 = '''Inputs:

         Function (string): Function with independent variable x. See the Supported
         Constants and Functions section.

         n (integer): Number of weights/nodes. The larger n is, the better the approximation.

         Output:

         Approximate solution of the integral derived via Gaussian quadrature.

         Note: Interval of integration is from -1 to 1.'''
        Widgets.paragraph_label(help_frame, label35, 14)
        Widgets.title_label(help_frame, "Numerical Solver", 18)
        label36 = '''Differential equations are equations which contain one or more
         independent variables, a function and one or more of its derivatives. An ordinary
         differential equation (ODE) contains only ordinary derivatives, whereas a partial
         differential equation (PDE) contains partial derivatives. The order of a
         differential equation is the order of its highest order derivative. Analytical
         solution techniques for solving first order ODEs include direct integration,
         integrating factors, separation of variables, potential functions, and substitution.
         However, these techniques can only be applied in very specific cases. In the vast
         majority of cases, one must resort to the use of numerical methods to approximate
         solutions to ODEs. This app implements various techniques for approximating
         solutions of first order ODEs including linear multistep methods and Runge-Kutta
         methods.

         Linear multistep methods use a given step size to estimate the value of the
         solution to an ODE until it achieves an estimate for the solution at the desired
         point. At each step, an estimate for the solution is obtained using the value of
         the function and the derivative at previous points. Euler's method, Adams
         Bashforth technique, and Milue's method are the linear multistep methods
         implemented in this app.

         Runge-Kutta methods are similar to the linear multistep methods except the Rung
         Kutta methods also use the values of the function and derivative at half step sizes
         to compute an estimate at each point. The Runge-Kutta methods implemented in this
         app are the midpoint method, the modified Euler method, Heun's method, and the
         Runge Kutta method.

         See the respective sections for instructions on how to use the various numerical
         methods for solving first order ODEs.'''
        Widgets.paragraph_label(help_frame, label36, 14)
        Widgets.title_label(help_frame, "Euler's Method", 16)
        label37 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via Euler's method.'''
        Widgets.paragraph_label(help_frame, label37, 14)
        Widgets.title_label(help_frame, "Midpoint Method", 16)
        label38 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via the midpoint method.'''
        Widgets.paragraph_label(help_frame, label38, 14)
        Widgets.title_label(help_frame, "Modified Euler Method", 16)
        label39 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via the modified Euler method.'''
        Widgets.paragraph_label(help_frame, label39, 14)
        Widgets.title_label(help_frame, "Heun's Method", 16)
        label40 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via Heun's method.'''
        Widgets.paragraph_label(help_frame, label40, 14)
        Widgets.title_label(help_frame, "Runge-Kutta Method", 16)
        label41 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via the Runge-Kutta method.'''
        Widgets.paragraph_label(help_frame, label41, 14)
        Widgets.title_label(help_frame, "Adams-Bashforth Technique", 16)
        label42 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via the Adams-Bashforth technique.

         Note: This algorithm uses the Runge-Kutta method as the base method.'''
        Widgets.paragraph_label(help_frame, label42, 14)
        Widgets.title_label(help_frame, "Milue's Method", 16)
        label43 = '''Inputs:

         Function (string): Function with independent variables x and y. See the Supported
         Constants and Functions section.

         a (float): The x-value where the initial condition is specified.

         b (float): The x-value where the solution of the ODE is approximated.

         y(a) (float): Value of function where the initial condition is specified.

         N (integer): Number of steps. The larger n is, the better the approximation.

         Output:

         Approximate solution of the ODE at x = b derived via Milue's method.

         Note: This algorithm uses the Runge-Kutta method as the base method.'''
        Widgets.paragraph_label(help_frame, label43, 14)
        Widgets.window_button(self.window1, "Close", lambda: Commands.close_help(self), "top")

    def acknowledgements(self):

        #Set up window

        self.window2 = tk.Toplevel()
        self.window2.title("Acknowledgements")
        self.window2.geometry("1250x500")

        #Widgets

        Widgets.title_label(self.window2, "Acknowledgements", 24)
        acknowledgements_frame = Widgets.scrollbar(self.window2)
        label = '''I acknowledge the extensive use of the math and mpmath Python modules in
         this project and am grateful for the work done by the creators of these packages.
         Without these libraries this project would have been much harder.

         The book Numerical Analysis by Burden & Faires, 9th edition was used as a reference
         over the course of the project, and I thank the authors of this book which served
         as a valuable resource.

         Furthermore, Wolfram MathWorld allowed me to learn about the various special
         functions I did not know about prior to this project. I thank the creators of this
         website, for the material I referred to during this project. As well as serving as
         a great resource for nearly every area of mathematics.

         I thank the owner of the YouTube channel Clear Code. His tkinter tutorial helped
         me learn how to create user interfaces in Python. Including the one for this
         project.

         I thank Liu Zuo Lin for his Medium article "Another Python Question That Took Me
         Days To Solve As A Beginner" on how to evaluate Python strings as mathematical
         expressions. This helped me avoid the use of the eval and exec functions, which
         have been shown to be insecure.'''
        Widgets.paragraph_label(acknowledgements_frame, label, 14)
        Widgets.window_button(self.window2, "Close",
         lambda: Commands.close_acknowledgements(self), "top")

    def close_help(self):
        self.window1.destroy()

    def close_acknowledgements(self):
        self.window2.destroy()
        
class Parse:
    def parse_function(function):

        #Remove spaces

        function = function.replace(" ", "")

        #Multiplication

        i = 48
        number_list = []
        while i < 58:
            number_list.append(chr(i))
            i += 1
        i = 65
        letter_list = []
        while i < 91:
            letter_list.append(chr(i))
            i += 1
        i = 97
        while i < 123:
            letter_list.append(chr(i))
            i += 1
        i = 0
        index_list = []
        while i < (len(function) - 1):
            if (function[i] in number_list and function[i + 1] in letter_list):
                index_list.append(i)
            i += 1
        i = 0
        j = 0
        break_function = []
        while i < len(function):
            if i in index_list:
                break_function.append(function[j:i + 1])
                j = i + 1
            i += 1
        break_function.append(function[j:])
        i = 0
        while i < len(break_function):
            if i == 0:
                function = break_function[i]
            else:
                function = f"{function}*{break_function[i]}"
            i += 1
        function = function.replace("x(", "x*(")
        i = 0
        while i < 10:
            search_string = f"{str(i)}("
            function = function.replace(search_string, f"{str(i)}*(")
            i += 1
        function = function.replace(")(", ")*(")

        #Exponentiation

        function = function.replace("^", "**")

        #Absolute value

        split_function = function.split("|")
        i = 0
        temp_function = ""
        if "|" in function:
            while i < len(split_function):
                if i % 2 == 0 and (i != len(split_function) - 1):
                    temp_function = f"{temp_function}{split_function[i]}abs("
                if i % 2 == 1:
                    temp_function = f"{temp_function}{split_function[i]})"
                if i == (len(split_function) - 1):
                    temp_function = f"{temp_function}{split_function[i]}"
                i += 1
            function = temp_function
        function = function.replace("xabs", "x*abs")
        i = 0
        while i < 10:
            search_string = f"{str(i)}abs"
            function = function.replace(search_string, f"{str(i)}*abs")
            i += 1
        function = function.replace(")abs", ")*abs")

        #Exponential and logarithmic function

        function = function.replace("e**", "m.exp")
        function = function.replace("log", "m.log")
        i = 0
        argument_list = []
        while i < len(function) - 2:
            if function[i:i+3] == "log":
                j = i + 3
                log_argument = ""
                while function[j] != ")":
                    log_argument = f"{log_argument}{function[j]}"
                    j += 1
                log_argument = f"{log_argument})"
                argument_list.append(log_argument)
            i += 1
        i = 0
        while i < len(argument_list):
            if "," not in argument_list[i]:
                function = function.replace(argument_list[i],
                 f"{argument_list[i][0:-1]},10)")
            i += 1
        function = function.replace("ln", "m.log")
        function = function.replace("sqrt", "m.sqrt")
        function = function.replace("cbrt", "mp.cbrt")
        function = function.replace("root", "mp.root")
        function = function.replace("W", "mp.lambertw")

        #Trigonometric and hyperbolic functions

        function = function.replace("sin", "m.sin")
        function = function.replace("arcm.sin", "m.asin")
        function = function.replace("arm.sinh", "m.asinh")
        function = function.replace("cos", "m.cos")
        function = function.replace("arcm.cos", "m.acos")
        function = function.replace("arm.cosh", "m.acosh")
        function = function.replace("tan", "m.tan")
        function = function.replace("arcm.tan", "m.atan")
        function = function.replace("arm.tanh", "m.atanh")
        function = function.replace("sec", "mp.sec")
        function = function.replace("arcmp.sec", "mp.asec")
        function = function.replace("armp.sech", "mp.asech")
        function = function.replace("csc", "mp.csc")
        function = function.replace("arcmp.csc", "mp.acsc")
        function = function.replace("armp.csch", "mp.acsch")
        function = function.replace("cot", "mp.cot")
        function = function.replace("arcmp.cot", "mp.acot")
        function = function.replace("armp.coth", "mp.acoth")

        #Special functions

        i = 0
        psi_function_list = []
        while i < len(function) - 2:
            if function[i:i+3] == "psi":
                j = i + 3
                psi_argument = ""
                while function[j] != ")":
                    psi_argument = f"{psi_argument}{function[j]}"
                    j += 1
                psi_argument = f"{psi_argument})"
                psi_function = f"psi{psi_argument}"
                psi_function_list.append(psi_function)
            i += 1
        i = 0
        while i < len(psi_function_list):
            if "," in psi_function_list[i]:
                function = function.replace(psi_function_list[i],
                 f"mp.{psi_function_list[i]})")
            else:
                function = function.replace(psi_function_list[i],
                 f"mp.digamma{psi_function_list[i]}")
            i += 1
        function = function.replace("digammapsi", "digamma")
        i = 0
        gamma_function_list = []
        while i < len(function) - 4:
            if function[i:i+5] == "Gamma":
                j = i + 5
                gamma_argument = ""
                while function[j] != ")":
                    gamma_argument = f"{gamma_argument}{function[j]}"
                    j += 1
                gamma_argument = f"{gamma_argument})"
                gamma_function = f"Gamma{gamma_argument}"
                gamma_function_list.append(gamma_function)
            i += 1
        i = 0
        while i < len(gamma_function_list):
            if "," not in gamma_function_list[i]:
                function = function.replace(gamma_function_list[i],
                 f"m.g{gamma_function_list[i]})")
            else:
                function = function.replace(gamma_function_list[i],
                 f"mp.gammainc{gamma_function_list[i]})")
            i += 1
        function = function.replace("gGamma", "gamma")
        function = function.replace("gammaincGamma", "gammainc")
        function = function.replace("G", "mp.barnesg")
        function = function.replace("Ei", "mp.ei")
        function = function.replace("En", "mp.exp-int")
        function = function.replace("li", "mp.li")
        function = function.replace("Ci", "mp.ci")
        function = function.replace("Chi", "mp.chi")
        function = function.replace("Si", "mp.si")
        function = function.replace("Shi", "mp.shi")
        function = function.replace("erf","m.erf")
        function = function.replace("m.erfi","mp.erfi")
        function = function.replace("S", "mp.fresnels")
        function = function.replace("C", "mp.fresnelc")
        function = function.replace("J", "mp.besselj")
        function = function.replace("Y", "mp.bessely")
        function = function.replace("I", "mp.besseli")
        function = function.replace("K", "mp.besselk")
        function = function.replace("H", "mp.struveh")
        function = function.replace("L", "mp.struvel")
        function = function.replace("E", "mp.webere")
        function = function.replace("s(", "mp.lommels1(")
        function = function.replace("abmp.lommels1(", "abs(")
        function = function.replace("comp.lommels1(", "cos(")
        function = function.replace("fresnelmp.lommels1(", "fresnels(")
        function = function.replace("Ai", "mp.airyai")
        function = function.replace("Bi", "mp.airybi")
        function = function.replace("barnesgi", "scorergi")
        function = function.replace("struvehi", "scorerhi")
        function = function.replace("pFq", "mp.hyper")
        function = function.replace("Amp.struvehB", "mp.bihyper")
        function = function.replace("eta", "mp.altzeta")
        function = function.replace("thmp.altzeta", "mp.jtheta")
        function = function.replace("zmp.altzeta", "mp.zeta")
        function = function.replace("Phi", "mp.lerchPhi")
        function = function.replace("mp.struveli", "mp.polylog")

        #Constants

        function = function.replace("pi", "m.pi")
        function = function.replace("pix", "pi*x")
        function = function.replace("piabs", "pi*abs")
        function = function.replace("pim.", "pi*m.")
        function = function.replace("pi(", "pi*(")
        function = function.replace("phi", "mp.phi")
        function = function.replace("phix", "phi*x")
        function = function.replace("phiabs", "phi*abs")
        function = function.replace("phim.", "phi*m.")
        function = function.replace("phi(", "phi*(")
        i = 0
        while i < len(function) - 4:
            if (function[i - 1] != "." and function[i-2:i] != "di" and
             function[i:i + 5] == "gamma"):
                function = f"{function[:i]}mp.euler{function[i + 5:]}"
            i += 1
        function = function.replace("eulerx", "euler*x")
        function = function.replace("eulerabs", "euler*abs")
        function = function.replace("eulerm.", "euler*m.")
        function = function.replace("euler(", "euler*(")
        function = function.replace("A", "mp.glaisher")
        function = function.replace("glaisherx", "glaisher*x")
        function = function.replace("glaisherabs", "glaisher*abs")
        function = function.replace("glaisherm.", "glaisher*m.")
        function = function.replace("glaisher(", "glaisher*(")
        function = function.replace("infinity", "m.inf")

        #Multiplication 2

        function = function.replace("xm.", "x*m.")
        function = function.replace(")m.", ")*m.")
        function = function.replace("xmp.", "x*mp.")
        function = function.replace(")mp.", ")*mp.")
        function = function.replace("pimp.", "pi*mp.")
        function = function.replace("phimp.", "phi*mp.")
        function = function.replace("eulermp.", "euler*mp.")
        function = function.replace("glaishermp.", "glaisher*mp.")

        #Corrections

        function = function.replace("mp.exp-int", "mp.expint")
        function = function.replace("mp.lerchPhi", "mp.lerchphi")

        #Return parsed functiom

        return function

    def split_by(function, operations):
        output_list = []
        string_segment = ""
        i = 0
        while i < len(function):
            if function[i] in operations:
                output_list.append(string_segment)
                output_list.append(function[i])
                string_segment = ""
            else:
                string_segment = f"{string_segment}{function[i]}"
            i += 1
        output_list.append(string_segment)
        i = 0
        while i < len(output_list):
            if output_list[i] == "":
                output_list.pop(i)
            i += 1
        return output_list

    def evaluate_expression(function):
        expression_value = Parse.split_by(function, "+-*/")
        i = 0
        while i < len(expression_value):
            if expression_value[i][-1] == "e":
                expression_value[i:i + 3] = ["".join(expression_value[i:i + 3])]
            i += 1
        i = 0
        while i < len(expression_value):
            if (expression_value[i - 1] == "*" and expression_value[i] == "-"):
                expression_value[i:i + 2] = ["".join(expression_value[i:i + 2])]
            i += 1

        #Evaluate mathemetical constants

        constant_dictionary = {"m.pi": m.pi, "mp.phi": mp.phi, 'mp.euler': mp.euler,
         "mp.glaisher": mp.glaisher, "m.inf": m.inf}
        i = 0
        while i < len(expression_value):
            if expression_value[i] in constant_dictionary:
                expression_value[i] = str(constant_dictionary[expression_value[i]])
            i += 1

        #Evaluate mathematical functions

        function_dictionary = {"abs": m.fabs, "m.exp": m.exp, "m.log": m.log,
         "m.sqrt": m.sqrt, "mp.cbrt": mp.cbrt, "mp.root": mp.root,
         "mp.lambertw": mp.lambertw, "m.sin": m.sin, "m.asin": m.asin, "m.sinh": m.sinh,
         "m.asinh": m.asinh, "m.cos": m.cos, "m.acos": m.acos, "m.cosh": m.cosh,
         "m.acosh": m.acosh, "m.tan": m.tan, "m.atan": m.atan, "m.tanh": m.tanh,
         "m.atanh": m.atanh, "mp.sec": mp.sec, "mp.asec": mp.asec, "mp.sech": mp.sech,
         "mp.asech": mp.asech, "mp.csc": mp.csc, "mp.acsc": mp.acsc, "mp.csch": mp.csch,
         "mp.acsch": mp.acsch, "mp.cot": mp.cot, "mp.acot": mp.acot, "mp.coth": mp.coth,
         "mp.acoth": mp.acoth, "mp.digamma": mp.digamma, "mp.psi": mp.psi,
         "m.gamma": m.gamma, "mp.gammainc": mp.gammainc, "mp.barnesg": mp.barnesg,
         "mp.ei": mp.ei, "mp.expint": mp.expint, "mp.li": mp.li, "mp.ci": mp.ci,
         "mp.chi": mp.chi, "mp.si": mp.si, "mp.shi": mp.shi, "m.erf": m.erf,
         "m.erfc": m.erfc, "mp.erfi": mp.erfi, "mp.fresnels": mp.fresnels,
         "mp.fresnelc": mp.fresnelc, "mp.besselj": mp.besselj, "mp.bessely": mp.bessely,
         "mp.besseli": mp.besseli, "mp.besselk": mp.besselk, "mp.struveh": mp.struveh,
         "mp.struvel": mp.struvel, "mp.webere": mp.webere, "mp.lommels1": mp.lommels1,
         "mp.airyai": mp.airyai, "mp.airybi": mp.airybi, "mp.scorergi": mp.scorergi,
         "mp.scorerhi": mp.scorerhi, "mp.hyper": mp.hyper, "mp.bihyper": mp.bihyper,
         "mp.altzeta": mp.altzeta, "mp.jtheta": mp.jtheta, "mp.zeta": mp.zeta,
         "mp.lerchphi": mp.lerchphi, "mp.polylog": mp.polylog}
        i = 0
        while i < len(expression_value):
            expression_value[i] = expression_value[i].replace("negative", "-")
            if ("(" in expression_value[i] and "[" not in expression_value[i]):
                if "," not in expression_value[i]:
                    sub_list = Parse.split_by(expression_value[i], "()")
                    value = function_dictionary[sub_list[0]](float(sub_list[2]))
                    expression_value[i] = str(value)
                else:
                    sub_list = Parse.split_by(expression_value[i], "(,)")
                    if len(sub_list) == 6:
                        value = function_dictionary[sub_list[0]](float(sub_list[2]),
                         float(sub_list[4]))
                        expression_value[i] = str(value)
                    if len(sub_list) == 8:
                        value = mp.re(function_dictionary[sub_list[0]](float(sub_list[2]),
                         float(sub_list[4]), float(sub_list[6])))
                        expression_value[i] = str(value)
            if "(" in expression_value[i] and "[" in expression_value[i]:
                sub_list = Parse.split_by(expression_value[i], "()[],")
                j = 0
                index_values = []
                while j < len(sub_list):
                    if sub_list[j] == "[" or sub_list[j] == "]":
                        index_values.append(j)
                    j += 1
                j = 0
                list1 = []
                list2 = []
                while j < len(sub_list):
                    if (j > index_values[0] and j < index_values[1] and sub_list[j] != ","):
                        list1.append(float(sub_list[j]))
                    if (j > index_values[2] and j < index_values[3] and sub_list[j] != ","):
                        list2.append(float(sub_list[j]))
                    j += 1
                value = function_dictionary[sub_list[0]](list1, list2, float(sub_list[-2]))
                expression_value[i] = str(value)
            i += 1

        #Exponents

        i = 0
        while i < len(expression_value) - 2:
            if expression_value[i:i + 2] == ["*", "*"]:
                expression_value[i:i + 2] = ["". join(expression_value[i:i + 2])]
            i += 1
        i = 0
        while i < len(expression_value) - 1:
            if expression_value[i] == "**":
                value = float(expression_value[i - 1]) ** float(expression_value[i + 1])
                expression_value[i - 1:i + 2] = ["".join(expression_value[i - 1:i + 2])]
                expression_value[i - 1] = str(value)
            i += 1

        #Multiplication & Division

        i = 0
        while i < len(expression_value) - 1:
            if expression_value[i] == "*":
                value = float(expression_value[i - 1]) * float(expression_value[i + 1])
                expression_value[i - 1:i + 2] = ["".join(expression_value[i - 1:i + 2])]
                expression_value[i - 1] = str(value)
                i = i
            elif expression_value[i] == "/":
                value = float(expression_value[i - 1]) / float(expression_value[i + 1])
                expression_value[i - 1:i + 2] = ["".join(expression_value[i - 1:i + 2])]
                expression_value[i - 1] = str(value)
                i = i
            else:
                i += 1

        #Addition & Subtraction

        i = 0
        while i < len(expression_value) - 1:
            if expression_value[i] == "+":
                value = float(expression_value[i - 1]) + float(expression_value[i + 1])
                expression_value[i - 1:i + 2] = ["".join(expression_value[i - 1:i + 2])]
                expression_value[i - 1] = str(value)
                i = i
            elif (expression_value[i] == "-" and i != 0 and
             ("e" not in expression_value[i])):
                value = float(expression_value[i - 1]) - float(expression_value[i + 1])
                expression_value[i - 1:i + 2] = ["".join(expression_value[i - 1:i + 2])]
                expression_value[i - 1] = str(value)
                i = i
            elif expression_value[i] == "-" and i == 0:
                expression_value[i:i + 2] = ["".join(expression_value[i:i + 2])]
                i += 1
            else:
                i += 1
                                            
        #Return evaluated expression

        return expression_value[0]

    def evaluate_function(function, number):

        #Preliminary steps

        function = f"({function})"
        function = function.replace("x", number)
        exception = f"e{str(number)}p"
        function = function.replace("--", "")
        function = function.replace(exception, "exp")

        #Split at ( and )

        evaluation_list = Parse.split_by(function, "(,)")

        #Evaluate parentheses

        i = 0
        j = len(evaluation_list)
        while j > 1:
            i = 0
            while i < j:
                evaluation_list[i] = evaluation_list[i].replace("start", "(")
                evaluation_list[i] = evaluation_list[i].replace("end", ")")
                evaluation_list[i] = evaluation_list[i].replace("comma", ",")
                i += 1
            i = 0
            while i < j - 2:
                if ((evaluation_list[i] == "(" and evaluation_list[i + 2] == ")") or
                 (evaluation_list[i] == "(" and evaluation_list[i + 2] == ",") or
                 (evaluation_list[i] == "comma" and evaluation_list[i + 2] == ",") or
                 (evaluation_list[i] == "comma" and evaluation_list[i + 2] == ")")):
                    evaluation_list[i + 1] = \
                     Parse.evaluate_expression(evaluation_list[i + 1])
                    if (evaluation_list[i - 1][-1] == "+" or
                     evaluation_list[i - 1][-1] == "-" or
                     evaluation_list[i - 1][-1] == "*" or
                     evaluation_list[i - 1][-1] == "/" or evaluation_list[i - 1][-1] == "("):
                        evaluation_list.pop(i)
                        evaluation_list.pop(i + 1)
                        evaluation_list.append("")
                        evaluation_list.append("")
                    elif (evaluation_list[i] == "(" and evaluation_list[i + 2] == ","):
                        evaluation_list[i] = "start"
                        evaluation_list[i + 2] = "comma"
                    elif (evaluation_list[i] == "comma" and evaluation_list[i + 2] == ","):
                        evaluation_list[i + 2] = "comma"
                    elif (evaluation_list[i] == "comma" and evaluation_list[i + 2] == ")"):
                        evaluation_list[i + 2] = "end"
                    else:
                        evaluation_list[i] = "start"
                        evaluation_list[i + 2] = "end"
                i += 1
            i = 0
            function = ""
            while i < j:
                function = f"{function}{evaluation_list[i]}"
                i += 1
            function = function.replace("start-", "startnegative")
            function = function.replace("comma-", "commanegative")
            evaluation_list = Parse.split_by(function, "()")
            j = len(evaluation_list)
        function = function.replace("negative", "-")
        function = function.replace("start", "")
        function = function.replace("end", "")

        #Return value

        return function

    def string_to_list(string):
        string = string.replace(" ", "")
        string_list = string.split(",")
        i = 0
        while i < len(string_list):
            string_list[i] = float(string_list[i])
            i += 1
        return string_list

    def multivariate_parse(function):

        function = Parse.parse_function(function)

        #Multiplication

        function = function.replace("xy", "x*y")
        function = function.replace("y(", "y*(")
        function = function.replace("yabs", "y*abs")
        function = function.replace("piy", "pi*y")
        function = function.replace("phiy", "phi*y")
        function = function.replace("eulery", "euler*y")
        function = function.replace("glaishery", "glaisher*y")
        function = function.replace("ym.", "y*m.")
        function = function.replace("ymp.", "y*mp.")

        #Corrections

        function = function.replace("bessely*(", "bessely(")

        #Return parsed function

        return function

    def evaluate_multivariate(function, number1, number2):

        #Preliminary steps

        function = function.replace("y", number2)
        exception = f"bessel{str(number2)}"
        function = function.replace(exception, "bessely")

        #Evaluate Function

        function = Parse.evaluate_function(function, number1)

        #Return value

        return function
        
App("Numerical Methods Calculator", "zoomed")
```
